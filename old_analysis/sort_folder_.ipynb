```python
import os
import re
from datetime import datetime

# Path to your data_validation folder
folder_path = "/home/azureuser/mycode/gitrepos/AzureMachineLearning/ASA_pool_aml/data_validation"

# Regex to extract date from folder name
date_pattern = re.compile(r"(\d{2})_(\d{2})_(\d{4})")

# List all sub-folders
folders = [f for f in os.listdir(folder_path) if os.path.isdir(os.path.join(folder_path, f))]

# Extract dates and sort
dated_folders = []
for folder in folders:
    match = date_pattern.search(folder)
    if match:
        day, month, year = match.groups()
        date = datetime.strptime(f"{day}_{month}_{year}", "%d_%m_%Y")
        dated_folders.append((folder, date))

# Sort from latest to oldest
sorted_folders = sorted(dated_folders, key=lambda x: x[1], reverse=False)

# Print sorted folder names
for folder, date in sorted_folders:
    print(folder)
```
