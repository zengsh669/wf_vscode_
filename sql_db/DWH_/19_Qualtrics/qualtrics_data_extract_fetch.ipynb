# Fetch functions
Same fetch mechanics as `qualtrics_fetch.ipynb` (initiate export → poll → download zip → extract CSV), copied here as an independent function — not imported.

```python
def fetch_survey_responses(survey_id: str, label: str) -> pd.DataFrame:
    # Step 1: Initiate export
    resp = requests.post(
        f"{BASE_URL}/surveys/{survey_id}/export-responses",
        headers=headers,
        json={"format": "csv", "useLabels": True}
    )
    resp.raise_for_status()
    progress_id = resp.json()["result"]["progressId"]
    print(f"[{label}] Export started — progressId: {progress_id}")

    # Step 2: Poll until complete
    file_id = None
    while True:
        poll = requests.get(
            f"{BASE_URL}/surveys/{survey_id}/export-responses/{progress_id}",
            headers=headers
        )
        poll.raise_for_status()
        result = poll.json()["result"]
        status = result["status"]
        if status == "complete":
            file_id = result["fileId"]
            break
        elif status == "failed":
            raise RuntimeError(f"[{label}] Export failed")
        time.sleep(2)

    print(f"[{label}] Done — fileId: {file_id}")

    # Step 3: Download zip → extract CSV → raw DataFrame
    download = requests.get(
        f"{BASE_URL}/surveys/{survey_id}/export-responses/{file_id}/file",
        headers=headers
    )
    download.raise_for_status()

    with zipfile.ZipFile(io.BytesIO(download.content)) as z:
        csv_name = z.namelist()[0]
        # Qualtrics CSVs: row 0 = col names, row 1 = question text, row 2 = ImportId
        df = pd.read_csv(z.open(csv_name), skiprows=[1, 2], low_memory=False)

    print(f"[{label}] Raw rows: {len(df):,}  |  Columns: {len(df.columns)}")
    return df


def _split_timestamp(raw):
    # Qlik: Timestamp(timestamp#(x, 'YYYY-MM-DD hh:mm:ss')) / Date(date#(left(x,10),...)) /
    # Time(time#(mid(x,12,8),...)) — all three derive from the same source string. A blank/
    # missing value maps to Null() in Qlik; a present-but-unparseable value ALSO fails to
    # parse in Qlik (silent Null), so both cases must degrade to (NaT, NaT, None) here rather
    # than raising — pd.to_datetime(..., errors="coerce") can return NaT even for a truthy
    # input string, and NaT has no .normalize()/.time() methods, so those must be guarded too.
    ts = pd.to_datetime(raw, errors="coerce") if raw else pd.NaT
    date_part = ts.normalize() if pd.notna(ts) else pd.NaT
    time_part = ts.time() if pd.notna(ts) else None
    return ts, date_part, time_part


# Column order/schema fetch_surveys_metadata() must always return, even when the Surveys API
# returns zero elements — pd.DataFrame(rows) with rows=[] produces a DataFrame with NO columns
# at all (shape (0,0)), which would make any downstream Qualtrics_Surveys["Survey_id"] etc.
# raise a KeyError instead of just being an empty-but-well-shaped table.
_SURVEYS_METADATA_COLUMNS = [
    "Survey_id", "Survey_name", "Survey_ownerId", "Survey_lastModified",
    "Survey_lastModified_timestamp", "Survey_lastModified_date", "Survey_lastModified_time",
    "Survey_creationDate", "Survey_creationDate_timestamp", "Survey_creationDate_date",
    "Survey_creationDate_time", "Survey_isActive",
]


def fetch_surveys_metadata() -> pd.DataFrame:
    # Qlik: Qualtrics_Surveys (lines 198-218) — LOAD id, name, ownerId, lastModified,
    # creationDate, isActive FROM Surveys, PLUS 6 derived timestamp/date/time columns
    # split out of lastModified/creationDate (each is an ISO-ish 'YYYY-MM-DDThh:mm:ssZ'
    # string; Qlik takes left(x,10) for the date part and mid(x,12,8) for the time part).
    resp = requests.get(f"{BASE_URL}/surveys", headers=headers)
    resp.raise_for_status()
    elements = resp.json()["result"]["elements"]

    rows = []
    for s in elements:
        last_modified = s.get("lastModified")
        creation_date = s.get("creationDate")
        lm_ts, lm_date, lm_time = _split_timestamp(last_modified)
        cd_ts, cd_date, cd_time = _split_timestamp(creation_date)
        rows.append({
            "Survey_id": s.get("id"),
            "Survey_name": s.get("name"),
            "Survey_ownerId": s.get("ownerId"),
            "Survey_lastModified": last_modified,
            "Survey_lastModified_timestamp": lm_ts,
            "Survey_lastModified_date": lm_date,
            "Survey_lastModified_time": lm_time,
            "Survey_creationDate": creation_date,
            "Survey_creationDate_timestamp": cd_ts,
            "Survey_creationDate_date": cd_date,
            "Survey_creationDate_time": cd_time,
            "Survey_isActive": s.get("isActive"),
        })
    # If the API returns zero surveys, pd.DataFrame([]) would have no columns at all; force
    # the expected schema so downstream column access degrades to an empty-but-valid table
    # instead of a KeyError.
    return pd.DataFrame(rows, columns=_SURVEYS_METADATA_COLUMNS)


def fetch_question_definitions(survey_id: str) -> pd.DataFrame:
    # Qlik: Qualtrics_Questions — SELECT QuestionID, QuestionText FROM Questions WITH PROPERTIES (surveyId=...)
    resp = requests.get(
        f"{BASE_URL}/survey-definitions/{survey_id}/questions",
        headers=headers
    )
    resp.raise_for_status()
    elements = resp.json()["result"]["elements"]

    rows = []
    for q in elements:
        rows.append({
            "QuestionID": q.get("QuestionID"),
            "QuestionText": q.get("QuestionText"),
            "DataExportTag": q.get("DataExportTag"),  # kept for cross-referencing against Q-numbered export columns
            "QuestionType": q.get("QuestionType"),
            "Selector": q.get("Selector"),
        })
    return pd.DataFrame(rows)
```

```python
# Qlik: Qualtrics_Questions — For Each vSurveyID IN $(vSurveys) (all 8 IDs), concatenates each
# survey's question definitions into one table, tagged with Survey_id / [Survey Name]. This is
# ALL_VSURVEYS (8 IDs), not just the 6-survey SURVEYS dict used for response data — round 7 fix:
# the previous version only looped SURVEYS and silently dropped the "90 Day NPS" / "10 Day NPS -
# New member survey" question definitions that Qlik's vSurveys loop does include.
#
# Qlik: [Survey Name] = ApplyMap('Survey_MAP', vSurveyID) — Survey_MAP is a MAPPING LOAD built
# RESIDENT Qualtrics_Surveys (the just-fetched, live survey metadata), so [Survey Name] always
# reflects the survey's CURRENT name in Qualtrics, not a hardcoded string. Round 11 fix: this
# notebook previously used the static ALL_VSURVEYS_NAMES dict here — now uses SURVEY_NAME_MAP
# (built from the live Qualtrics_Surveys DataFrame fetched in the previous cell) to match.
SURVEY_NAME_MAP = Qualtrics_Surveys.set_index("Survey_id")["Survey_name"].to_dict()

questions_parts = []
for label, survey_id in ALL_VSURVEYS.items():
    part = question_defs[label][["QuestionID", "QuestionText"]].copy()
    part.insert(0, "Survey_id", survey_id)
    part.insert(1, "Survey Name", SURVEY_NAME_MAP.get(survey_id))
    questions_parts.append(part)

Qualtrics_Questions = pd.concat(questions_parts, ignore_index=True)
print(f"Qualtrics_Questions: {len(Qualtrics_Questions):,} rows")
Qualtrics_Questions.head()
```

# Shared helper functions
Copied (not imported) from `qualtrics_fetch.ipynb` — already verified against `Qualtrics_Data_Extract.md` across 8 rounds of independent review. Kept identical here so the two notebooks can diverge safely without affecting each other.

```python
def notnull_or_blank(df: pd.DataFrame, col: str) -> pd.Series:
    # Always return a Series aligned to df's index (never a bare scalar), so downstream
    # helpers (blank_to_null, combine_first, etc.) can rely on Series semantics even when
    # the source column is entirely absent from this survey's export.
    if col in df.columns:
        return df[col]
    return pd.Series(np.nan, index=df.index)

def blank_to_null(series: pd.Series) -> pd.Series:
    # Qlik: If(Len(Trim(x))=0, Null(), x) — blank/whitespace-only string treated as Null,
    # not just true NaN
    is_blank = series.isna() | (series.astype(str).str.strip() == "")
    return series.where(~is_blank, np.nan)

def constant_col(value, df: pd.DataFrame) -> pd.Series:
    # A bare scalar assigned to a brand-new column on an empty DataFrame does not
    # broadcast reliably in pandas — build an explicit Series aligned to df's index instead.
    return pd.Series([value] * len(df), index=df.index)

def blank_fallback(primary: pd.Series, fallback: pd.Series) -> pd.Series:
    # Qlik: If(Len(Trim(primary))=0, fallback, primary)
    return blank_to_null(primary).combine_first(fallback)

def qlik_date_only(series: pd.Series) -> pd.Series:
    # Qlik: Date(date#(left(x, 10), 'YYYY-MM-DD')) — truncates to the date portion only,
    # discarding any time-of-day component. A blank/missing/malformed date string silently
    # degrades to Null() in Qlik (date#() failing to parse never raises, never halts the
    # LOAD) — errors="coerce" is required here to match that row-by-row fault tolerance;
    # without it, a single bad row would crash pd.to_datetime() for the whole survey's
    # DataFrame instead of just leaving that one row's date blank. Round 11 fix — this
    # notebook's own _split_timestamp() already established this same principle for
    # Qualtrics_Surveys' timestamp columns; qlik_date_only() had not been brought in line.
    return pd.to_datetime(series, errors="coerce").dt.normalize()

def product_with_extras_fallback(df: pd.DataFrame, product_col: str, extras_col: str = "Extras"):
    # Qlik: If(Len(Trim([Hospital Product]))>0, Trim([Hospital Product]), Trim(Extras)) as Product
    if product_col not in df.columns:
        return pd.Series(np.nan, index=df.index)
    product = df[product_col]
    extras = notnull_or_blank(df, extras_col)
    is_blank = product.isna() | (product.astype(str).str.strip() == "")
    result = product.where(~is_blank, extras)
    return result.astype(str).str.strip().where(result.notna(), result)

def hwfl_substitute(df: pd.DataFrame, col: str = "Survey") -> pd.Series:
    # Qlik: If(Survey='HWFL','Healthy Weight for Life',Survey)
    survey = notnull_or_blank(df, col)
    return survey.where(survey != "HWFL", "Healthy Weight for Life")

# Qlik: ProgramProviderId_Map — a hardcoded `LOAD * Inline` table (Qualtrics_Data_Extract.md
# lines 57-66), NOT sourced from a Paragon BRONZE table, so it's in scope to replicate.
PROGRAM_PROVIDER_ID_MAP = {
    "360 Med Care": "F200017B",
    "Hospital Care at Home": "F300006H",
    "Rehabilitation at Home": "F300006H",
    "MindStep": "F300006H",
    "HWFL": "G200010Y",
}

def program_provider_no(key_series: pd.Series) -> pd.Series:
    # Qlik: ApplyMap('ProgramProviderId_Map', <key>) — unmatched keys map to Null()
    return key_series.map(PROGRAM_PROVIDER_ID_MAP)

# Qlik: ProgramProvider_Map — ALSO a hardcoded `LOAD * Inline` table (Qualtrics_Data_Extract.md
# lines 68-75), the inverse of ProgramProviderId_Map (keyed on provider_number_id, returning the
# provider display name). Round 7 finding: this is used for `[Provider Group]` in
# HealthServices/MentalHealth/Wellbeing/WellbeingV2 (`ApplyMap('ProgramProvider_Map',ProviderNo)
# as [Provider Group]`) and, exactly like ProgramProviderId_Map, is a static inline table with
# NO Paragon BRONZE dependency — the commented-out lines 76-78 show it USED TO be BRONZE-sourced
# but was replaced with the hardcoded inline version. It was previously mis-classified as
# out-of-scope alongside the genuinely-BRONZE-dependent [Provider Type]/[Provider Address]/etc.
# fields; fixed to be replicated like ProviderNo already is.
PROGRAM_PROVIDER_MAP = {
    "F200017B": "360 Med Care",
    "F300006H": "Remedy at Home",
    "G200010Y": "Prima Health Solutions Pty ltd",
}

def program_provider_group(provider_no_series: pd.Series) -> pd.Series:
    # Qlik: ApplyMap('ProgramProvider_Map', ProviderNo) as [Provider Group] — unmatched keys
    # (including Null/NaN ProviderNo) map to Null() (no 3rd default-value argument given).
    return provider_no_series.map(PROGRAM_PROVIDER_MAP)

# Qlik: Pick(Match(SubField(Interaction,' ',-2), 'Eye','Dental','Care','Call','Contact'),
#             'Eye Care','Dental','Care Centre','Contact Centre','Contact Centre') as Interaction
_CSAT_INTERACTION_MAP = {"Eye": "Eye Care", "Dental": "Dental", "Care": "Care Centre",
                          "Call": "Contact Centre", "Contact": "Contact Centre"}

def csat_interaction(raw: str):
    if pd.isna(raw) or not str(raw).strip():
        return np.nan
    words = str(raw).strip().split(" ")
    if len(words) < 2:
        return np.nan  # SubField(x,' ',-2) has no second-to-last token -> Match() fails -> Null()
    second_last = words[-2]
    return _CSAT_INTERACTION_MAP.get(second_last, np.nan)

# Round 12 fix: HealthServices' "How did you hear" (QID3) is a multi-select checkbox
# question -- the real CSV export splits it into Q3_1..Q3_6 (one column per choice,
# value = choice text when selected, blank otherwise), NOT a single "Q3" column (which
# does not exist in the real export -- confirmed via live export-responses pull + the
# question's real Selector=MAVR/multi-answer type). Verified against all 18 real rows:
# the overwhelming majority select exactly one option, one row selects two, one row
# selects none (blank) -- the comma-join below handles all three cases without raising
# and without misattributing any row's answer to another question (the Q3_1..Q3_6
# tags themselves are confirmed via the CSV's ImportId metadata to all belong to QID3).
# Not independently verified against Qlik's own historical QVD output format (no access
# to that data) -- this is the most defensible reconstruction given the real column
# structure, but the exact join style (comma-separated) is a reasonable choice rather
# than a byte-for-byte confirmed match to legacy Qlik connector behaviour.
def multiselect_join(df: pd.DataFrame, cols: list) -> pd.Series:
    existing = [c for c in cols if c in df.columns]
    if not existing:
        return pd.Series(np.nan, index=df.index)
    joined = df[existing].apply(
        lambda row: ", ".join(str(v) for v in row if pd.notna(v) and str(v).strip()),
        axis=1
    )
    return joined.replace("", np.nan)
```

# CSAT — full field replication
Qlik source: `Qualtrics_Data_Extract.md` lines 355-408. Excludes `[Provider Type]`/`[Provider Address]`/`[Provider Suburb]`/`[Provider Postcode]`/`[Provider State]` (BranchLocations.xlsx-dependent).

The sibling `qualtrics_fetch.ipynb` / `progress.md` empirically verified via the `survey-definitions` API (real question text) + real response value distributions that this survey's true NPS field is `Q3`/`Q3_NPS_GROUP`, true Satisfaction field is `Q4`/`Q4_NPS_GROUP` (not `Q2`/`Q20` as the Qlik comments literally state), **and true Program Type ("Which service did you receive?") field is `Q2`, not `Q21`** — round 5 correction: this notebook previously carried over Qlik's `QID21` comment naively as `Q21`, but the sibling notebook's `standardise_wellbeing_v2` explicitly documents (twice) that `Q2` is the real "which service" export tag for this survey, and `Q21` was never independently verified for this meaning. Fixed to `Q2` with the same blank-fallback-to-`Survey` structure as before.

```python
def standardise_wellbeing_v2_full(df: pd.DataFrame) -> pd.DataFrame:
    out = pd.DataFrame(index=df.index)
    out["Survey_id"] = constant_col(SURVEYS["WellbeingV2"], df)
    # Qlik: ApplyMap('Survey_MAP', Survey_id) as [Survey Name] — always the survey's CURRENT
    # name per the live Qualtrics_Surveys fetch, not a hardcoded string. Round 11 fix.
    out["Survey Name"] = constant_col(SURVEY_NAME_MAP.get(SURVEYS["WellbeingV2"]), df)
    out["ResponseId"] = df["ResponseId"]
    out["StartDate"] = qlik_date_only(df["StartDate"])
    out["MonthYear"] = pd.to_datetime(df["StartDate"], errors="coerce").dt.strftime("%b %Y")
    out["EndDate"] = qlik_date_only(df["EndDate"])
    out["Status"] = notnull_or_blank(df, "Status")
    out["IPAddress"] = notnull_or_blank(df, "IPAddress")
    # Qlik: Interval(duration/86400) — a blank/non-numeric duration is Null()/86400 = Null() in
    # Qlik (no error), so pd.to_numeric(..., errors="coerce") is used instead of .astype(float),
    # which would raise on a blank string like "".
    out["Duration"] = pd.to_numeric(notnull_or_blank(df, "Duration (in seconds)"), errors="coerce") / 86400
    out["Recipient"] = notnull_or_blank(df, "RecipientFirstName").astype(str) + " " + notnull_or_blank(df, "RecipientLastName").astype(str)
    out["Email"] = notnull_or_blank(df, "RecipientEmail")
    out["ExternalReference"] = notnull_or_blank(df, "ExternalReference")
    out["LocationLatitude"] = notnull_or_blank(df, "LocationLatitude")
    out["LocationLongitude"] = notnull_or_blank(df, "LocationLongitude")

    # Program Type: Round 10 live API dump (WellbeingV2-only) confirms Q2(QID21)="Which service
    # did you receive?" — matches previous round-5 fix, unchanged.
    out["Program Type"] = blank_fallback(notnull_or_blank(df, "Q2"), notnull_or_blank(df, "Survey"))
    # NPS/Satisfaction: Round 10 live API dump confirms Q3(QID2)="On a scale from 0-10, how
    # likely are you to recommend Westfund..." and Q4(QID20)="Overall how satisfied were you
    # with the program?" — matches previous fix, unchanged.
    out["Net Promoter Level"] = blank_to_null(notnull_or_blank(df, "Q3_NPS_GROUP"))
    out["Net Promoter Score"] = blank_to_null(notnull_or_blank(df, "Q3"))
    out["Customer Satisfaction Level"] = blank_to_null(notnull_or_blank(df, "Q4_NPS_GROUP"))
    out["Customer Satisfaction Score"] = blank_to_null(notnull_or_blank(df, "Q4"))
    # Round 10 fix (live API question-dump, WellbeingV2-only): Q6_TEXT does not exist as a tag in
    # this survey (Q6 itself is an MC question, "achieve goals"/"professional and courteous" —
    # same duplicate-tag pattern as elsewhere in this survey family). The real "how we could
    # improve this program" free-text question is Q5(QID6, TE type — tag has no _TEXT suffix).
    out["Feedback Improvements"] = notnull_or_blank(df, "Q5")

    out["Interaction"] = constant_col("Health Services", df)
    out["ProviderNo"] = program_provider_no(notnull_or_blank(df, "Survey"))  # Qlik: ApplyMap('ProgramProviderId_Map', Survey) — raw Survey, pre-HWFL substitution
    # Qlik: ApplyMap('ProgramProvider_Map', ProviderNo) as [Provider Group] — ProgramProvider_Map
    # is a static inline table (Qualtrics_Data_Extract.md lines 68-75), NOT BRONZE-dependent, so
    # this IS in scope. Round 7 fix: PROGRAM_PROVIDER_MAP/program_provider_group() were defined
    # in the shared helpers cell but never actually wired into this function — column was silently
    # missing from output.
    out["Provider Group"] = program_provider_group(out["ProviderNo"])
    out["Provider/Program/Staff"] = hwfl_substitute(df)
    out["Age"] = notnull_or_blank(df, "Age")
    out["Product"] = product_with_extras_fallback(df, "HospitalProduct")
    out["Gender"] = notnull_or_blank(df, "Gender")
    out["Member Number"] = notnull_or_blank(df, "MemberNumber")
    out["Member Type"] = notnull_or_blank(df, "MemberType")
    out["Postcode"] = notnull_or_blank(df, "Postcode")
    out["State"] = notnull_or_blank(df, "State")
    out["TenureMonths"] = notnull_or_blank(df, "TenureMonths")
    out["Region"] = notnull_or_blank(df, "Region")
    out["PreviousFund"] = notnull_or_blank(df, "PreviousFund")
    out["Promotion"] = notnull_or_blank(df, "Promotion")
    return out
```
