```sql
/****** Object:  View [dbo].[ME_Total_Membership]    Script Date: 4/06/2026 10:39:10 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE VIEW [dbo].[ME_Total_Membership]  
AS
SELECT CONVERT(date,[Run Month]) AS [Run Month]
	,'H' AS [Product Type]
	,RIGHT([Product Cover],2) AS [Product Cover]
	,[Hosp Product Id] AS [Product]
	,[Product Cover Group] AS [Product Group]
    ,COUNT([Membership Number]) AS [Policies]
FROM [SILVER].[dbo].[Membership_Group_Key]
INNER JOIN [SILVER].[dbo].[Product] ON [Hosp Product Id] = [Product Id]
WHERE ISNULL([Hosp Product Id],0) <> 0
AND [Product Cover] <> 'OS'
AND [Membership Status] IN ('A','J')
GROUP BY [Run Month], [Product Cover], [Hosp Product Id], [Product Cover Group]

UNION ALL

SELECT CONVERT(date,[Run Month]) AS [Run Month]
	,'E' AS [Product Type]
	,RIGHT([Product Cover],2) AS [Product Cover]
	,[Extras Product Id] AS [Product]
	,[Product Cover Group] AS [Product Group]
    ,COUNT([Membership Number]) AS [Policies]
FROM [SILVER].[dbo].[Membership_Group_Key]
INNER JOIN [SILVER].[dbo].[Product] ON [Extras Product Id] = [Product Id]
WHERE ISNULL([Extras Product Id],0) <> 0
AND [Product Cover] <> 'OS'
AND [Membership Status] IN ('A','J')
GROUP BY [Run Month], [Product Cover], [Extras Product Id], [Product Cover Group]

UNION ALL

SELECT CONVERT(date,[Run Month]) AS [Run Month]
	,'A' AS [Product Type]
	,RIGHT([Product Cover],2) AS [Product Cover]
	,[Ambuln Product Id] AS [Product]
	,[Product Cover Group] AS [Product Group]
    ,COUNT([Membership Number]) AS [Policies]
FROM [SILVER].[dbo].[Membership_Group_Key] 
INNER JOIN [SILVER].[dbo].[Product] ON [Ambuln Product Id] = [Product Id]
WHERE ISNULL([Ambuln Product Id],0) <> 0
AND [Product Cover] <> 'OS'
AND [Membership Status] IN ('A','J')
GROUP BY [Run Month], [Product Cover], [Ambuln Product Id], [Product Cover Group];

GO
```

# [dbo].[Membership_Budget]

```sql
/****** Object:  View [dbo].[Membership_Budget]    Script Date: 4/06/2026 10:39:10 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE VIEW [dbo].[Membership_Budget]  
AS
SELECT CAST([Run Month] AS DATE) AS [Run Month],
       [Run Month Num],
	   CASE WHEN MONTH([Run Month]) > 6 THEN YEAR([Run Month]) + 1 ELSE YEAR([Run Month]) END AS [Fin Year],
	   --CONCAT('Q',CEILING(CASE WHEN MONTH([Run Month]) > 6 THEN MONTH([Run Month]) - 6 ELSE MONTH([Run Month]) + 6 END / 3)) AS [Fin Quarter],
	   LEFT(DATENAME(month,[Run Month]),3) AS [Fin Month],
	   CASE WHEN MONTH([Run Month]) > 6 THEN MONTH([Run Month]) - 6 ELSE MONTH([Run Month]) + 6 END AS [Fin Month Order],
	   CASE WHEN ISNULL([Sales Channel Group],'null')<>'null' THEN 'SALES'
			ELSE CASE WHEN ISNULL([Hosp Product Id],0)>0 THEN 'HOSP'
					  WHEN ISNULL([Extras Product Id],0)>0 THEN 'EXTRA'
					  WHEN ISNULL([Ambuln Product Id],0)>0 THEN 'AMBU' END END AS [Budget Type],
	   [Product Cover],
       ISNULL([Hosp Product Id],0) AS [Hosp Product Id],
	   H.[Product Cover Group] AS [Hospital Product],
	   H.[Product Cover Tier] AS [Hospital Tier],
	   H.[Product Status] AS [Hospital Status],
	   ISNULL([Extras Product Id],0) AS [Extras Product Id],
	   E.[Product Cover Group] AS [Extras Product],
	   E.[Product Cover Tier] AS [Extras Tier],
	   E.[Product Status] AS [Extras Status],
	   ISNULL([Ambuln Product Id],0) AS [Ambuln Product Id],
	   A.[Product Cover Group] AS [Ambuln Product],
       [Sales Channel],
       [Sales Channel Group],
       [Active Budget],
       [Join Budget],
       [Termination Budget],
       [Net Move Budget]
FROM [SILVER].[dbo].[Membership_Budget] B
LEFT JOIN [SILVER].[dbo].[Product] H ON B.[Hosp Product Id] = H.[Product Id]
LEFT JOIN [SILVER].[dbo].[Product] E ON B.[Extras Product Id] = E.[Product Id]
LEFT JOIN [SILVER].[dbo].[Product] A ON B.[Ambuln Product Id] = A.[Product Id]
LEFT JOIN (SELECT [Run Month] as [Month],ROW_NUMBER() OVER (ORDER BY [Run Month] asc) as [Run Month Num]
			FROM (SELECT Distinct [Run Month] FROM [SILVER].[dbo].[Membership_Budget]) AS A) AS N ON N.[Month] = B.[Run Month];

GO
```

```sql
/****** Object:  View [dbo].[Claim_Aggr]    Script Date: 4/06/2026 10:39:10 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO







CREATE               VIEW [dbo].[Claim_Aggr] AS
SELECT 
    EOMONTH(service_date,0) AS service_date, 
    EOMONTH(status_date,0) AS paid_date, 
    claim_type, 
    service_type, 
    'NA' AS stay_type, 
    cover_state, 
    'NA' AS provider_type,
    SUM(serv) AS serv, 
    SUM(fee_amt) AS fee_amt, 
    SUM(benefit_amt) AS benefit_amt,
	SUM(gross_deficit) AS gross_deficit,
	product_id
FROM SILVER.dbo.Claim_Fact
WHERE 
  benefit_amt <> 0
  AND claim_type IN ('A','B','M')
GROUP BY 
    EOMONTH(service_date,0), 
    EOMONTH(status_date,0), 
    claim_type, 
    service_type, 
    cover_state,
	product_id

UNION ALL

SELECT 
    EOMONTH(service_date,0) AS service_date, 
    EOMONTH(status_date,0) AS paid_date, 
    claim_type, 
    service_type, 
    stay_type, 
    cover_state, 
    provider_type,
    SUM(serv) AS serv, 
    SUM(fee_amt) AS fee_amt, 
    SUM(benefit_amt) AS benefit_amt,
	SUM(gross_deficit) AS gross_deficit,
	product_id
FROM SILVER.dbo.Claim_Fact
WHERE 
  benefit_amt <> 0
  AND claim_type = 'H'
GROUP BY 
    EOMONTH(service_date,0), 
    EOMONTH(status_date,0), 
    claim_type, 
    service_type, 
    stay_type, 
    cover_state, 
    provider_type,
	product_id;
GO
```

# [dbo].[ME_Membership_Terminations]

```sql
/****** Object:  View [dbo].[ME_Membership_Terminations]    Script Date: 4/06/2026 10:39:10 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE VIEW [dbo].[ME_Membership_Terminations]  
AS
SELECT [Run Month],
	[Product Cover], 
	ISNULL([Hosp Product Id],0) AS [Hosp Product Id],
	H.[Product Cover Group] AS [Hospital Product],
	H.[Product Cover Tier] AS [Hospital Tier],
	ISNULL([Extras Product Id],0) AS [Extras Product Id],
	E.[Product Cover Group] AS [Extras Product],
	E.[Product Cover Tier] AS [Extras Tier],
	ISNULL([Ambuln Product Id],0) AS [Ambuln Product Id], 
	[Branch Group Id],
	[Cover Type],
	[Cover State],
	[Sales Channel],
	ISNULL([Sales Channel Group],'Other') AS [Sales Channel Group],
	[Agent],
	[Promotion],
	CASE WHEN DATEDIFF(year, [Effective Join Date], [Run Month]) < 1 THEN 'L<1'
		WHEN DATEDIFF(year, [Effective Join Date], [Run Month]) BETWEEN 1 AND 2 THEN 'L01-02'
		WHEN DATEDIFF(year, [Effective Join Date], [Run Month]) BETWEEN 3 AND 4 THEN 'L03-04'
		WHEN DATEDIFF(year, [Effective Join Date], [Run Month]) BETWEEN 5 AND 7 THEN 'L05-07'
		WHEN DATEDIFF(year, [Effective Join Date], [Run Month]) BETWEEN 8 AND 10 THEN 'L08-10'
		WHEN DATEDIFF(year, [Effective Join Date], [Run Month]) BETWEEN 11 AND 15 THEN 'L11-15'
		WHEN DATEDIFF(year, [Effective Join Date], [Run Month]) BETWEEN 16 AND 20 THEN 'L16-20'
		WHEN DATEDIFF(year, [Effective Join Date], [Run Month]) >= 21 THEN 'L21+'
		ELSE 'NA' END AS [Tenure],
	T.[Termination Code],
	ISNULL(C.[Termination Group],'Other') AS [Termination Reason],
	T.[Transfered Fund Id],
	ISNULL([Transfered Fund Group],'Other') AS [Transfered Fund],
	COUNT(M.[Membership Number]) AS [Policies]

FROM [SILVER].[dbo].[Membership_Group_Key] M
LEFT JOIN [SILVER].[dbo].[Product] H ON M.[Hosp Product Id] = H.[Product Id]
LEFT JOIN [SILVER].[dbo].[Product] E ON M.[Extras Product Id] = E.[Product Id]
LEFT JOIN [SILVER].[dbo].[Membership_History] T ON M.[Membership Number] = T.[Membership Number] AND M.[Run Month] = T.[SnapShot Month]
LEFT JOIN [SILVER].[dbo].[Termination_Code] C ON T.[Termination Code] = C.[Termination Code]
WHERE M.[Membership Status] = 'T'
AND [Product Cover] <> 'OS'
GROUP BY [Run Month],[Product Cover],
	[Hosp Product Id],H.[Product Cover Group],H.[Product Cover Tier],
	[Extras Product Id],E.[Product Cover Group],E.[Product Cover Tier],
	[Ambuln Product Id],[Branch Group Id],[Cover Type],[Cover State],[Sales Channel],[Sales Channel Group],[Agent],
	[Promotion],[Effective Join Date],T.[Termination Code],[Termination Group],[Transfered Fund Id],[Transfered Fund Group];

GO
```

# [dbo].[Membership_Movement]

```sql
/****** Object:  View [dbo].[Membership_Movement]    Script Date: 4/06/2026 10:39:10 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE VIEW [dbo].[Membership_Movement]  
AS 
SELECT [Fin Year],
	[Membership Number],
	[From Cover Type],
	CASE [From Cover Type] 
		WHEN 'Combined' THEN 1
		WHEN 'Hospital only' THEN 2
		WHEN 'Extras only' THEN 3
		WHEN 'Ambulance' THEN 4
		ELSE 5 END AS [From Cover Type Order],
	[To Cover Type],
	CASE [To Cover Type] 
		WHEN 'Combined' THEN 1
		WHEN 'Hospital only' THEN 2
		WHEN 'Extras only' THEN 3
		WHEN 'Ambulance' THEN 4
		ELSE 5 END AS [To Cover Type Order],
	[From Hospital Tier],
	CASE [From Hospital Tier] 
		WHEN 'Gold' THEN 1
		WHEN 'Silver+' THEN 2
		WHEN 'Silver' THEN 3
		WHEN 'Bronze+' THEN 4
		WHEN 'Bronze' THEN 5
		WHEN 'Basic' THEN 6
		ELSE 7 END AS [From Hospital Tier Order],
	[To Hospital Tier],
	CASE [To Hospital Tier] 
		WHEN 'Gold' THEN 1
		WHEN 'Silver+' THEN 2
		WHEN 'Silver' THEN 3
		WHEN 'Bronze+' THEN 4
		WHEN 'Bronze' THEN 5
		WHEN 'Basic' THEN 6
		ELSE 7 END AS [To Hospital Tier Order],
	[From Extras Tier],
	CASE [From Extras Tier] 
		WHEN 'Top' THEN 1
		WHEN 'High' THEN 2
		WHEN 'Mid' THEN 3
		WHEN 'Low' THEN 4
		ELSE 5 END AS [From Extras Tier Order],
	[To Extras Tier],
	CASE [To Extras Tier] 
		WHEN 'Top' THEN 1
		WHEN 'High' THEN 2
		WHEN 'Mid' THEN 3
		WHEN 'Low' THEN 4
		ELSE 5 END AS [To Extras Tier Order]
	
FROM (SELECT Distinct [Membership Number],
	[Fin Year],
	FIRST_VALUE([Membership Status]) OVER (PARTITION BY [Membership Number],[Fin Year] ORDER BY [Run Month]) AS [From Membership Status],
    LAST_VALUE([Membership Status]) OVER (PARTITION BY [Membership Number],[Fin Year] ORDER BY [Run Month] ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING) AS [To Membership Status],
	FIRST_VALUE([Product Cover]) OVER (PARTITION BY [Membership Number],[Fin Year] ORDER BY [Run Month]) AS [From Cover Type],
    LAST_VALUE([Product Cover]) OVER (PARTITION BY [Membership Number],[Fin Year] ORDER BY [Run Month] ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING) AS [To Cover Type],

	FIRST_VALUE([Hospital Tier]) OVER (PARTITION BY [Membership Number],[Fin Year] ORDER BY [Run Month]) AS [From Hospital Tier],
    LAST_VALUE([Hospital Tier]) OVER (PARTITION BY [Membership Number],[Fin Year] ORDER BY [Run Month] ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING) AS [To Hospital Tier],	
	FIRST_VALUE([Hospital Product]) OVER (PARTITION BY [Membership Number],[Fin Year] ORDER BY [Run Month]) AS [From Hospital Product],
    LAST_VALUE([Hospital Product]) OVER (PARTITION BY [Membership Number],[Fin Year] ORDER BY [Run Month] ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING) AS [To Hospital Product],

	FIRST_VALUE([Extras Tier]) OVER (PARTITION BY [Membership Number],[Fin Year] ORDER BY [Run Month]) AS [From Extras Tier],
    LAST_VALUE([Extras Tier]) OVER (PARTITION BY [Membership Number],[Fin Year] ORDER BY [Run Month] ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING) AS [To Extras Tier],	
	FIRST_VALUE([Extras Product]) OVER (PARTITION BY [Membership Number],[Fin Year] ORDER BY [Run Month]) AS [From Extras Product],
    LAST_VALUE([Extras Product]) OVER (PARTITION BY [Membership Number],[Fin Year] ORDER BY [Run Month] ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING) AS [To Extras Product]

FROM (SELECT [Membership Number],
	[Membership Status],
	CASE WHEN MONTH([Run Month]) > 6 THEN YEAR([Run Month]) + 1 ELSE YEAR([Run Month]) END AS [Fin Year],
	[Run Month],
	CASE WHEN [Membership Status] IN ('J','T') THEN 'No cover' 
	ELSE CASE [Product Cover]
		WHEN 'PK' THEN 'Combined'
		WHEN 'HSA' THEN 'Hospital only'
		WHEN 'ESA' THEN 'Extras only'
		WHEN 'ASA' THEN 'Ambulance' 
		WHEN 'OS' THEN 'Overseas'
		ELSE [Product Cover] END END AS [Product Cover],
	CASE WHEN [Membership Status] IN ('J','T') THEN 'No cover' ELSE COALESCE(H.[Product Cover Tier],'No cover') END AS [Hospital Tier],
	CASE WHEN [Membership Status] IN ('J','T') THEN 'No cover' ELSE COALESCE(H.[Product Cover Group],'No cover') END [Hospital Product],
	CASE WHEN [Membership Status] IN ('J','T') THEN 'No cover' ELSE COALESCE(E.[Product Cover Tier],'No cover') END AS [Extras Tier],
	CASE WHEN [Membership Status] IN ('J','T') THEN 'No cover' ELSE COALESCE(E.[Product Cover Group],'No cover') END [Extras Product]
FROM [SILVER].[dbo].[Membership_Group_Key] M
LEFT JOIN [SILVER].[dbo].[Product] H ON M.[Hosp Product Id] = H.[Product Id]
LEFT JOIN [SILVER].[dbo].[Product] E ON M.[Extras Product Id] = E.[Product Id]
WHERE [Product Cover] <> 'OS') AS Move1) AS Move2
WHERE [From Membership Status] = 'A'
AND [To Membership Status] = 'A'; 

GO
```

# [dbo].[vw_Calculated_Deficit]

```sql
/****** Object:  View [dbo].[vw_Calculated_Deficit]    Script Date: 4/06/2026 10:39:10 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO


CREATE   VIEW [dbo].[vw_Calculated_Deficit]
AS
WITH SEUs_CTE AS (
    SELECT
        DATEFROMPARTS(YEAR(DATEADD(DAY, -1, rundate)), MONTH(DATEADD(DAY, -1, rundate)), 1) AS MonthYear,
        cover_state AS State,
        SUM(
            CASE
                WHEN cover_type IN ('C', 'F', 'E') THEN 2
                ELSE 1
            END
        ) AS [Hospital SEUs]
    FROM BRONZE.dbo.group_key_full_by_branch
    WHERE rundate >= DATEADD(MONTH, -13, GETDATE()) ----------------------------#######################################
      AND cover LIKE '%Hos%'
    GROUP BY
        DATEFROMPARTS(YEAR(DATEADD(DAY, -1, rundate)), MONTH(DATEADD(DAY, -1, rundate)), 1),
        cover_state
),

ABP_CTE AS (
    SELECT
        SUM(
            CASE
                WHEN (0 + CONVERT(CHAR(8), cl.service_date, 112)
                        - CONVERT(CHAR(8), p.date_of_birth, 112)) / 10000 < 55
                    THEN 0 * COALESCE(ch.benefit, cg.benefit)
                WHEN (0 + CONVERT(CHAR(8), cl.service_date, 112)
                        - CONVERT(CHAR(8), p.date_of_birth, 112)) / 10000 BETWEEN 55 AND 59
                    THEN 0.15 * COALESCE(ch.benefit, cg.benefit)
                WHEN (0 + CONVERT(CHAR(8), cl.service_date, 112)
                        - CONVERT(CHAR(8), p.date_of_birth, 112)) / 10000 BETWEEN 60 AND 64
                    THEN 0.425 * COALESCE(ch.benefit, cg.benefit)
                WHEN (0 + CONVERT(CHAR(8), cl.service_date, 112)
                        - CONVERT(CHAR(8), p.date_of_birth, 112)) / 10000 BETWEEN 65 AND 69
                    THEN 0.60 * COALESCE(ch.benefit, cg.benefit)
                WHEN (0 + CONVERT(CHAR(8), cl.service_date, 112)
                        - CONVERT(CHAR(8), p.date_of_birth, 112)) / 10000 BETWEEN 70 AND 74
                    THEN 0.70 * COALESCE(ch.benefit, cg.benefit)
                WHEN (0 + CONVERT(CHAR(8), cl.service_date, 112)
                        - CONVERT(CHAR(8), p.date_of_birth, 112)) / 10000 BETWEEN 75 AND 79
                    THEN 0.76 * COALESCE(ch.benefit, cg.benefit)
                WHEN (0 + CONVERT(CHAR(8), cl.service_date, 112)
                        - CONVERT(CHAR(8), p.date_of_birth, 112)) / 10000 BETWEEN 80 AND 84
                    THEN 0.78 * COALESCE(ch.benefit, cg.benefit)
                WHEN (0 + CONVERT(CHAR(8), cl.service_date, 112)
                        - CONVERT(CHAR(8), p.date_of_birth, 112)) / 10000 >= 85
                    THEN 0.82 * COALESCE(ch.benefit, cg.benefit)
            END
        ) AS ABP,
        DATEFROMPARTS(YEAR(cl.status_date), MONTH(cl.status_date), 1) AS MonthYear,
        cd2.cover_state AS State
    FROM BRONZE.dbo.claim_line AS cl
    LEFT JOIN BRONZE.dbo.claim_hospitalitem AS ch
        ON cl.claim_id = ch.claim_id AND cl.claim_line_id = ch.claim_line_id
    LEFT JOIN BRONZE.dbo.claim_generalitem AS cg
        ON cl.claim_id = cg.claim_id AND cl.claim_line_id = cg.claim_line_id
    INNER JOIN BRONZE.dbo.person AS p
        ON cl.person_id = p.person_id
    INNER JOIN BRONZE.dbo.memship AS m
        ON cl.membership_id = m.membership_id
    INNER JOIN [SILVER].[dbo].[ClaimDetailsAtService_optimised] AS cd2
        ON cl.claim_id = cd2.claim_id AND cl.claim_line_id = cd2.claim_line_id
    WHERE cl.claim_line_status_type = 'P'
      AND cl.claim_type IN ('H', 'M')
      AND cl.status_date >= DATEADD(MONTH, -13, GETDATE()) --------------##############################
    GROUP BY
        DATEFROMPARTS(YEAR(cl.status_date), MONTH(cl.status_date), 1),
        cd2.cover_state
)

SELECT
    s.MonthYear,
    s.State,
    s.[Hospital SEUs],
    a.ABP
FROM SEUs_CTE s
INNER JOIN ABP_CTE a
    ON s.MonthYear = a.MonthYear
    AND s.State = a.State;
GO
```

```sql
/****** Object:  View [dbo].[vw_HCS_Claims]    Script Date: 4/06/2026 10:39:10 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO











CREATE                       VIEW [dbo].[vw_HCS_Claims] AS

WITH
/*------------------------------------------------------------------------------
  CTE 1: Item Mapping - Item description lookup
------------------------------------------------------------------------------*/
ItemDescMap AS (
    SELECT
        service_type + item_number AS map_key,
        description
    FROM BRONZE.dbo.item
),

/*------------------------------------------------------------------------------
  CTE 2: Service Type Mapping - Service type description lookup
------------------------------------------------------------------------------*/
ServiceTypeMap AS (
    SELECT
        service_type,
        description
    FROM BRONZE.dbo.service_type
),

/*------------------------------------------------------------------------------
  CTE 3: Gender Mapping - Gender lookup by person
------------------------------------------------------------------------------*/
GenderMap AS (
    SELECT
        person_id,
        gender
    FROM BRONZE.dbo.person
),

/*------------------------------------------------------------------------------
  CTE 4: Cover Type Mapping - Cover type description lookup
------------------------------------------------------------------------------*/
CoverTypeMap AS (
    SELECT
        cover_type,
        description
    FROM BRONZE.dbo.cover_type
),

/*------------------------------------------------------------------------------
  CTE 5: Fund Classification Mapping - Fund classification lookup
------------------------------------------------------------------------------*/
FundClassMapping AS (
    SELECT
        id AS fund_classification_id,
        item_type
    FROM BRONZE.dbo.fund_classification
),

/*------------------------------------------------------------------------------
  CTE 6: Item Group Mapping - Item group description lookup
------------------------------------------------------------------------------*/
ItemGroupDescMap AS (
    SELECT
        item_group,
        description
    FROM BRONZE.dbo.item_group
),

/*------------------------------------------------------------------------------
  CTE 7: Age Cohorts - Age group definitions
------------------------------------------------------------------------------*/
AgeCohorts AS (
    SELECT -10 AS agemin, 24 AS agemax, '0-24' AS AgeCohort
    UNION ALL SELECT 25, 34, '25-34'
    UNION ALL SELECT 35, 44, '35-44'
    UNION ALL SELECT 45, 54, '45-54'
    UNION ALL SELECT 55, 64, '55-64'
    UNION ALL SELECT 65, 74, '65-74'
    UNION ALL SELECT 75, 84, '75-84'
    UNION ALL SELECT 85, 199, '85+'
),

/*------------------------------------------------------------------------------
  CTE 8: LOM Cohorts - Length of Membership group definitions
  LOM = Length of Membership (in years)
------------------------------------------------------------------------------*/
LOMCohorts AS (
    SELECT -10 AS lommin, 1 AS lommax, '<1' AS LOMCohort
    UNION ALL SELECT 2, 3, '2-3'
    UNION ALL SELECT 4, 5, '4-5'
    UNION ALL SELECT 6, 7, '6-7'
    UNION ALL SELECT 8, 10, '8-10'
    UNION ALL SELECT 11, 15, '11-15'
    UNION ALL SELECT 16, 25, '16-25'
    UNION ALL SELECT 26, 199, '26+'
),

/*------------------------------------------------------------------------------
  CTE 9: Base Claims - Build base claims data
  Integrate data from multiple QVD sources and apply filter conditions
------------------------------------------------------------------------------*/
--BaseClaims AS (
--    SELECT
--        -- Membership info (from Paragon_Memberships.qvd)
--        m.membership_id,
--        m.termination_code AS MemshipTermCode,
--        m.memship_status AS MemshipStatus,

--        -- Person membership relationship (from Paragon_PersonMembership.qvd)
--        pm.person_id,
--        pm.relationship AS Relationship,
--        pm.status_flag AS PersonStatus,
--        pm.termination_code AS PersonTermCode,

--        -- Claim details (from Paragon_ClaimDetailGenAndHosp.qvd)
--        cd.claim_id,
--        cd.claim_line_id,
--        cd.line_status,
--        cd.claim_type,
--        cd.service_type,

--        -- Date fields
--        CONVERT(VARCHAR(8), cd.service_date, 112) AS MonthYear, -- To be converted to MonthName format
--        CONVERT(VARCHAR(8), cd.service_date, 112) AS IncurredMonthYear,
--        cd.service_date,
--        CAST(CAST(FLOOR(CAST(cd.create_datetime AS FLOAT)) AS DATETIME) AS DATE) AS [Claim Create Date],
--        CONVERT(VARCHAR(8), cd.status_date, 112) AS PaidMonthYear,

--        -- Claim amounts and services
--        cd.item_number,
--        cd.fee,
--        cd.benefit,
--        LTRIM(RTRIM(cd.provider_number_id)) AS ProviderNumberId,
--        cd.Num_services

--    FROM BRONZE.dbo.memship m

--    LEFT JOIN BRONZE.dbo.person_membership pm
--        ON m.membership_id = pm.membership_id

--    LEFT JOIN BRONZE.dbo.ClaimDetailGenAndHosp cd
--        ON pm.person_id = cd.person_id

--    WHERE cd.line_status = 'Paid'
--      AND cd.service_date > '2021-01-01'
--),

BaseClaims AS (
    SELECT
        -- Membership info (from Paragon_Memberships.qvd)
        m.membership_id,
        m.termination_code AS MemshipTermCode,
        m.memship_status AS MemshipStatus,

        -- Person membership relationship (from Paragon_PersonMembership.qvd)
        pm.person_id,
        pm.relationship AS Relationship,
        pm.status_flag AS PersonStatus,
        pm.termination_code AS PersonTermCode,

        -- Claim details (replaced view with base tables)
        cl.claim_id,
        cl.claim_line_id,
        cls.description AS line_status,
		--cl.claim_line_status_type, ------------------------
        
        -- claim_type logic (from view definition)
        CASE 
            WHEN ch.service_type IS NULL AND cg.service_type NOT IN ('MEDICAL','AMBULANCE') 
                THEN 'General'
            WHEN ch.service_type IS NULL AND cg.service_type = 'AMBULANCE' 
                THEN 'Ambulance'
            WHEN ch.service_type IS NULL AND cg.service_type = 'MEDICAL' 
                THEN 'Medical' 
            ELSE 'Hospital' 
        END AS claim_type,
        
        COALESCE(ch.service_type, cg.service_type) AS service_type,

        -- Date fields
        CONVERT(VARCHAR(8), cl.service_date, 112) AS MonthYear,
        CONVERT(VARCHAR(8), cl.service_date, 112) AS IncurredMonthYear,
        cl.service_date,
        CAST(CAST(FLOOR(CAST(cl.create_datetime AS FLOAT)) AS DATETIME) AS DATE) AS [Claim Create Date],
        CONVERT(VARCHAR(8), cl.status_date, 112) AS PaidMonthYear,

        -- Claim amounts and services
        COALESCE(ch.item_number, cg.item_number) AS item_number,
        COALESCE(ch.fee, cg.fee) AS fee,
        COALESCE(ch.benefit, cg.benefit) AS benefit,
        LTRIM(RTRIM(cl.provider_number_id)) AS ProviderNumberId,
        COALESCE(ch.num_services, cg.num_services) AS Num_services

    FROM BRONZE.dbo.memship m

    LEFT JOIN BRONZE.dbo.person_membership pm
        ON m.membership_id = pm.membership_id

    -- Replace view with base tables
    LEFT JOIN BRONZE.dbo.claim_line cl
        ON pm.person_id = cl.person_id
		AND pm.membership_id = cl.membership_id ------------------------ 24/02/2026
    
    INNER JOIN BRONZE.dbo.claim_line_status_type cls
        ON cl.claim_line_status_type = cls.claim_line_status_type
    
    LEFT OUTER JOIN BRONZE.dbo.claim_hospitalitem ch
        ON cl.claim_id = ch.claim_id 
        AND cl.claim_line_id = ch.claim_line_id
    
    LEFT OUTER JOIN BRONZE.dbo.claim_generalitem cg
        ON cl.claim_id = cg.claim_id 
        AND cl.claim_line_id = cg.claim_line_id

    WHERE cls.description = 'Paid'
      AND cl.service_date > '2021-01-01'
),

/*------------------------------------------------------------------------------
  CTE 10: Claims with Item Info - Add item information
------------------------------------------------------------------------------*/
ClaimsWithItem AS (
    SELECT
        bc.*,

        -- Apply Mapping: Item Description
        ISNULL(idm.description, '') AS [Item Description],

        -- Get additional info from item table
        ISNULL(igm.description, 'Missing') AS [Item Group Description],
        ISNULL(fcm.item_type, 'No Classification') AS [Fund Classification],
        i.disable_hicap_flag AS [Disabled HICAPS Flag],
        i.online_claiming AS [Onlien Claiming Flag],
        i.topup_flag AS [Include in Top up Job]

    FROM BaseClaims bc

    -- Apply ItemDescMap
    LEFT JOIN ItemDescMap idm
        ON bc.service_type + bc.item_number = idm.map_key

    -- Join with item table for additional fields
    LEFT JOIN BRONZE.dbo.item i
        ON bc.service_type = i.service_type
       AND bc.item_number = i.item_number

    -- Apply ItemGroupDescMap
    LEFT JOIN ItemGroupDescMap igm
        ON i.item_group = igm.item_group

    -- Apply FundClassMapping
    LEFT JOIN FundClassMapping fcm
        ON i.fund_classification_id = fcm.fund_classification_id
),

/*------------------------------------------------------------------------------
  CTE 11: Claims with Provider - Add healthcare provider information
------------------------------------------------------------------------------*/
ClaimsWithProvider AS (
    SELECT
        cwi.*,

        -- Provider Number info
        LTRIM(RTRIM(pn.provider_id)) AS ProviderId,

        -- Provider info
        p.provider_type,
        p.provider_name,

        -- Provider Type description
        pt.description AS [Provider Type],

        -- Hospital Provider address info
        pn2.postcode AS HospitalProviderPostcode,
        pn2.state AS HospitalProviderState,
        pn2.suburb AS HospitalProviderSuburb

    FROM ClaimsWithItem cwi

    LEFT JOIN BRONZE.dbo.provider_number pn
        ON cwi.ProviderNumberId = LTRIM(RTRIM(pn.provider_number_id))

    LEFT JOIN BRONZE.dbo.provider p
        ON LTRIM(RTRIM(pn.provider_id)) = LTRIM(RTRIM(p.provider_id))

    LEFT JOIN BRONZE.dbo.provider_type pt
        ON p.provider_type = pt.provider_type

    LEFT JOIN BRONZE.dbo.provider_number pn2
        ON cwi.ProviderNumberId = pn2.provider_number_id
),

/*------------------------------------------------------------------------------
  CTE 12: Claims with Service Details - Add service details
  Includes complex product grouping and classification logic
  FIX: Corrected Product Group logic for 'Extras Only' condition
------------------------------------------------------------------------------*/
ClaimsWithServiceDetails AS (
    SELECT
        cwp.*,

        -- Fields from ClaimDetailsAtService
        ISNULL(gm.gender, '') AS Gender,
        cds.claim_line_status_type,----------------------
        cds.status_date,
        ISNULL(ctm.description, '') AS CoverType,
        cds.cover_state AS State,
        cds.age_at_claim AS Age,
        cds.mem_branch_at_claim,
        cds.mem_cover_at_claim,
        cds.Product_Description_at_claim,
        cds.Hospital_Product_Code,
        cds.Ancillary_Product_Code,
        cds.Hospital_Product,
        cds.Extras_Product,

        -- Product Group - Complex nested logic
        -- FIX: Changed IN ('%Extras Only', 'Value Extras Ancillary') to LIKE + exact match
        CASE
            WHEN cds.mem_cover_at_claim IN ('OSC','NZO','FCO') THEN 'Overseas'
            WHEN (cds.Product_Description_at_claim LIKE '%&%' OR cds.Product_Description_at_claim LIKE '%with%')
                 AND cds.mem_cover_at_claim NOT IN ('CH','ECH','VE') THEN 'Hospital & Extras'
            WHEN (cds.Product_Description_at_claim LIKE '%&%' OR cds.Product_Description_at_claim LIKE '%with%')
                 AND cds.mem_cover_at_claim IN ('CH','ECH') THEN 'Hospital Only'
            WHEN (cds.Product_Description_at_claim LIKE '%&%' OR cds.Product_Description_at_claim LIKE '%with%')
                 AND cds.mem_cover_at_claim = 'VE' THEN 'Extras Only'
            WHEN cds.Product_Description_at_claim LIKE '%with%' THEN 'Hospital & Extras'
            WHEN cds.Product_Description_at_claim LIKE '%Extras' THEN 'Extras Only'
            WHEN cds.Product_Description_at_claim LIKE '%Extras Only'
                 OR cds.Product_Description_at_claim = 'Value Extras Ancillary' THEN 'Extras Only'
            WHEN cds.Product_Description_at_claim LIKE '%Ambulance%' THEN 'Ambulance'
            ELSE 'To Be Classified'
        END AS [Product Group],

        -- Hospital Tier
        CASE
            WHEN cds.Product_Description_at_claim LIKE '%Gold%' THEN 'Gold'
            WHEN cds.Product_Description_at_claim LIKE '%Silver%' THEN 'Silver'
            WHEN cds.Product_Description_at_claim LIKE '%Bronze%' THEN 'Bronze'
            WHEN cds.Product_Description_at_claim LIKE '%Basic%' THEN 'Basic'
            WHEN cds.Product_Description_at_claim LIKE '%Overseas%' THEN 'Overseas'
            ELSE 'Old Product'
        END AS [Hospital Tier],

        -- Product Classification
        CASE
            WHEN cds.mem_cover_at_claim IN ('J52','J53','J54','J58','J62','NZ','ENZ','FC','EFC','FCE','NZE','SP','ESP')
                 OR cds.mem_cover_at_claim LIKE 'J52%'
                 OR cds.mem_cover_at_claim LIKE 'J53%'
                 OR cds.mem_cover_at_claim LIKE 'J54%'
                 OR cds.mem_cover_at_claim LIKE 'J58%'
                 OR cds.mem_cover_at_claim LIKE 'J62%' THEN 'Top Cover'
            WHEN cds.mem_cover_at_claim IN ('NI','FI','VE')
                 OR cds.mem_cover_at_claim LIKE 'I%' THEN 'Extras Only'
            WHEN cds.mem_cover_at_claim IN ('AMBU','AMBUU','PAM','BAM','AMB') THEN 'Ambulance'
            ELSE 'Exclusionary'
        END AS [Product Classification]

    FROM ClaimsWithProvider cwp

    LEFT JOIN SILVER.dbo.ClaimDetailsAtService_optimised cds
	--LEFT JOIN BRONZE.dbo.ClaimDetailsAtService cds
        ON cwp.claim_id = cds.claim_id
       AND cwp.claim_line_id = cds.claim_line_id
       AND cds.service_date > '2021-01-01'
       AND cds.claim_line_status_type = 'P'

    -- Apply GenderMap
    LEFT JOIN GenderMap gm
        ON cwp.person_id = gm.person_id

    -- Apply CoverTypeMap
    LEFT JOIN CoverTypeMap ctm
        ON cds.cover_type = ctm.cover_type
),

/*------------------------------------------------------------------------------
  CTE 13: Claims with Hospital Items - Add hospital item information
------------------------------------------------------------------------------*/
ClaimsWithHospital AS (
    SELECT
        cwsd.*,

        CAST(CAST(FLOOR(CAST(chi.admission_date AS FLOAT)) AS DATETIME) AS DATE) AS [Admission Date],
        CAST(CAST(FLOOR(CAST(chi.discharge_date AS FLOAT)) AS DATETIME) AS DATE) AS [Discharge Date],
        chi.excess AS Excess,
        chi.num_beddays

    FROM ClaimsWithServiceDetails cwsd

    LEFT JOIN BRONZE.dbo.claim_hospitalitem chi
        ON cwsd.claim_id = chi.claim_id
       AND cwsd.claim_line_id = chi.claim_line_id
),

/*------------------------------------------------------------------------------
  CTE 14: Claims with Membership Date - Add membership join date
------------------------------------------------------------------------------*/
ClaimsWithMemshipDate AS (
    SELECT
        cwh.*,
        ms.effective_join_date

    FROM ClaimsWithHospital cwh

    INNER JOIN BRONZE.dbo.memship ms
        ON cwh.membership_id = ms.membership_id
),

/*------------------------------------------------------------------------------
  CTE 15: Claims with LOM - Calculate Length of Membership
  LOM = Age in complete years (similar to Qlik's Age() function)
------------------------------------------------------------------------------*/
ClaimsWithLOM AS (
    SELECT
        cwmd.*,
        DATEDIFF(YEAR, cwmd.effective_join_date, cwmd.service_date) -
        CASE
            WHEN MONTH(cwmd.service_date) < MONTH(cwmd.effective_join_date)
                 OR (MONTH(cwmd.service_date) = MONTH(cwmd.effective_join_date)
                     AND DAY(cwmd.service_date) < DAY(cwmd.effective_join_date))
            THEN 1
            ELSE 0
        END AS LOM

    FROM ClaimsWithMemshipDate cwmd
),

/*------------------------------------------------------------------------------
  CTE 16: Claims with Episode Classification - Add clinical classification
------------------------------------------------------------------------------*/
ClaimsWithEpisode AS (
    SELECT
        cwl.*,

        ec.EpisodeID,
        ec.ChapterItemNumber,
        ec.admission_date AS EpsiodeAdmissionDate,
        ec.discharge_date AS EpisodeDischargeDate,
        ec.episode_type AS Episode_Type,
        ec.ICD_Type AS [ICD Type],
        ec.Procedure_Block,
        ec.ACHI_Code,
        ec.long_description,
        ec.short_description,
        ec.Chapter,
        ec.EpisodeProcedureItemNumber,
        ec.EpisodeDiagnosisItemNumber,
        ec.code_id_clean

    FROM ClaimsWithLOM cwl

    --LEFT JOIN SILVER.dbo.episode_classification ec
	 LEFT JOIN SILVER.dbo.Episode_Classification ec
        ON cwl.claim_id = ec.EpisodeClaimID
       AND cwl.claim_line_id = ec.EpisodeClaimLineID
),

/*------------------------------------------------------------------------------
  CTE 17: Claims with Condition Group - Add clinical condition group
------------------------------------------------------------------------------*/
ClaimsWithCondition AS (
    SELECT
        cwe.*,
        ecg.[Condition Group]

    FROM ClaimsWithEpisode cwe

    LEFT JOIN SILVER.dbo.Episode_Condition_Group ecg
        ON cwe.claim_id = ecg.claim_id
       AND cwe.claim_line_id = ecg.claim_line_id
       AND cwe.EpisodeID = ecg.EpisodeID
),

/*------------------------------------------------------------------------------
  CTE 18: Claims with Member Address - Add member address information
  Get most recent member address (relationship = 1, primary member)
------------------------------------------------------------------------------*/
ClaimsWithMemAddress AS (
    SELECT
        cwc.*,
        pah1.postcode AS MemPostcode,
        pah1.suburb AS MemSuburb,
        pah1.state AS [Current State]

    FROM ClaimsWithCondition cwc

    LEFT JOIN (
        SELECT
            membership_id,
            postcode,
            suburb,
            state,
            ROW_NUMBER() OVER (PARTITION BY membership_id ORDER BY addressid DESC) AS rn
        FROM BRONZE.dbo.PersonAddressHomePostal
        WHERE relationship = 1
    ) pah1 ON cwc.membership_id = pah1.membership_id AND pah1.rn = 1
),

/*------------------------------------------------------------------------------
  CTE 19: Claims with Person Address - Add individual person address information
  Get most recent address for each person
------------------------------------------------------------------------------*/
ClaimsWithPersonAddress AS (
    SELECT
        cwma.*,
        pah2.postcode AS PersonPostcode,
        pah2.suburb AS PersonSuburb

    FROM ClaimsWithMemAddress cwma

    LEFT JOIN (
        SELECT
            membership_id,
            person_id,
            postcode,
            suburb,
            ROW_NUMBER() OVER (PARTITION BY membership_id, person_id ORDER BY addressid DESC) AS rn
        FROM BRONZE.dbo.PersonAddressHomePostal
    ) pah2 ON cwma.membership_id = pah2.membership_id
         AND cwma.person_id = pah2.person_id
         AND pah2.rn = 1
),

/*------------------------------------------------------------------------------
  CTE 20: GeoData - Geographic data optimization
  Choose most accurate Postcode and Suburb (prioritize person address, fallback to member address)
------------------------------------------------------------------------------*/
GeoData AS (
    SELECT
        cwpa.*,

        -- Postcode: Prioritize PersonPostcode
        CASE
            WHEN cwpa.PersonPostcode IS NULL THEN cwpa.MemPostcode
            WHEN LEN(cwpa.PersonPostcode) < 2 THEN cwpa.MemPostcode
            ELSE cwpa.PersonPostcode
        END AS Postcode,

        -- Suburb: Prioritize PersonSuburb
        CASE
            WHEN cwpa.PersonSuburb IS NULL THEN cwpa.MemSuburb
            WHEN LEN(cwpa.PersonSuburb) < 2 THEN cwpa.MemSuburb
            ELSE cwpa.PersonSuburb
        END AS Suburb

    FROM ClaimsWithPersonAddress cwpa
),

/*------------------------------------------------------------------------------
  CTE 21: Final Claims with Cohorts - Apply age and LOM groupings
  Use BETWEEN to implement Qlik's IntervalMatch logic
------------------------------------------------------------------------------*/
FinalClaims AS (
    SELECT
        gd.*,
        ac.AgeCohort,
        lc.LOMCohort

    FROM GeoData gd

    LEFT JOIN AgeCohorts ac
        ON gd.Age BETWEEN ac.agemin AND ac.agemax

    LEFT JOIN LOMCohorts lc
        ON gd.LOM BETWEEN lc.lommin AND lc.lommax
)

/*------------------------------------------------------------------------------
  Main SELECT - Final output
  Note: Excludes SA Level 3/4 and Region (LGA) fields (from Excel mappings)
------------------------------------------------------------------------------*/
SELECT
    -- Membership identifiers
    membership_id,
    person_id,

    -- Membership status
    MemshipTermCode,
    MemshipStatus,
    Relationship,
    PersonStatus,
    PersonTermCode,

    -- Claim identifiers
    claim_id,
    claim_line_id,

    -- Claim status and type
    line_status,
    claim_type,
    claim_line_status_type,

    -- Service information
    service_type,
    item_number,
    [Item Description],
    [Item Group Description],
    [Fund Classification],
    [Disabled HICAPS Flag],
    [Onlien Claiming Flag],
    [Include in Top up Job],

    -- Date fields
    service_date,
    [Claim Create Date],
    status_date,
    MonthYear,
    IncurredMonthYear,
    PaidMonthYear,

    -- Amounts and quantities
    fee,
    benefit,
    Excess,
    Num_services,
    num_beddays,

    -- Healthcare providers
    ProviderNumberId,
    ProviderId,
    provider_type,
    provider_name,
    [Provider Type],
    HospitalProviderPostcode,
    HospitalProviderState,
    HospitalProviderSuburb,

    -- Member and person information
    Gender,
    Age,
    AgeCohort,
    CoverType,
    State,
    [Current State],

    -- Product information
    mem_branch_at_claim,
    mem_cover_at_claim,
    Product_Description_at_claim,
    Hospital_Product_Code,
    Ancillary_Product_Code,
    Hospital_Product,
    Extras_Product,
    [Product Group],
    [Hospital Tier],
    [Product Classification],

    -- Hospital information
    [Admission Date],
    [Discharge Date],

    -- Length of membership
    effective_join_date,
    LOM,
    LOMCohort,

    -- Clinical classification
    EpisodeID,
    ChapterItemNumber,
    EpsiodeAdmissionDate,
    EpisodeDischargeDate,
    Episode_Type,
    [ICD Type],
    Procedure_Block,
    ACHI_Code,
    long_description,
    short_description,
    Chapter,
    EpisodeProcedureItemNumber,
    EpisodeDiagnosisItemNumber,
    code_id_clean,
    [Condition Group],

    -- Geographic information
    MemPostcode,
    MemSuburb,
    PersonPostcode,
    PersonSuburb,
    Postcode,
    Suburb

    -- Note: Following fields are missing (from Excel mappings, can be added later):
    -- [SA Level 3]
    -- [SA Level 4]
    -- [Region (LGA)]

FROM FinalClaims
WHERE service_date >= '2023-01-01'

;

GO
```

# [dbo].[vw_Membership_Current]

```sql
/****** Object:  View [dbo].[vw_Membership_Current]    Script Date: 4/06/2026 10:39:10 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO




CREATE           VIEW [dbo].[vw_Membership_Current]
AS
WITH NewJoins AS (
    SELECT m.membership_id,
           m.effective_join_date
    FROM BRONZE.dbo.memship m
    LEFT JOIN BRONZE.dbo.memship_20260529 snap 
        ON m.membership_id = snap.membership_id
    WHERE snap.membership_id IS NULL
      AND m.memship_status = 'A'
),
LatestCover AS (
    SELECT
        co.membership_id,
        co.description AS cover_description,
        COALESCE(h.product_code, '') + 
        COALESCE(a.product_code, '') + 
        COALESCE(b.product_code, '') AS product_code,
        ROW_NUMBER() OVER (PARTITION BY co.membership_id ORDER BY co.cover_version DESC) AS rn
    FROM BRONZE.dbo.cover co
    LEFT JOIN BRONZE.dbo.cover_product cp_h 
        ON co.membership_id = cp_h.membership_id 
        AND co.cover_version = cp_h.cover_version
    LEFT JOIN BRONZE.dbo.product h 
        ON cp_h.product_id = h.product_id 
        AND h.product_type = 'H'
    LEFT JOIN BRONZE.dbo.cover_product cp_a 
        ON co.membership_id = cp_a.membership_id 
        AND co.cover_version = cp_a.cover_version
    LEFT JOIN BRONZE.dbo.product a 
        ON cp_a.product_id = a.product_id 
        AND a.product_type = 'A'
    LEFT JOIN BRONZE.dbo.cover_product cp_b 
        ON co.membership_id = cp_b.membership_id 
        AND co.cover_version = cp_b.cover_version
    LEFT JOIN BRONZE.dbo.product b 
        ON cp_b.product_id = b.product_id 
        AND b.product_type = 'B'
    WHERE co.status_flag = 'A'
),
LatestPromo AS (
    SELECT
        main_ref AS person_id,
        promotion_id,
        sales_channel_id,
        operator,
        agency_group_id,
        ROW_NUMBER() OVER (PARTITION BY main_ref ORDER BY status_date DESC) AS rn
    FROM BRONZE.dbo.promotion_reference
    WHERE status_date <= GETDATE()
),
LatestMembership AS (
    SELECT
        pm.person_id,
        pm.membership_id,
        pm.relationship,
        ROW_NUMBER() OVER (PARTITION BY pm.person_id ORDER BY pm.membership_id DESC) AS rn
    FROM BRONZE.dbo.person_membership pm
    LEFT JOIN BRONZE.dbo.memship m ON pm.membership_id = m.membership_id
)
SELECT
    nj.membership_id,
    nj.effective_join_date,
    lc.cover_description,
    lc.product_code,
    lm.person_id,
    lp.promotion_id,
    p.description AS promotion_description,
    lp.sales_channel_id,
    ps.description AS sales_channel_description,
    o.first_name + ' ' + o.surname AS Operator,
    COALESCE(g.description, 'No Agency') AS Agent
FROM NewJoins nj
LEFT JOIN LatestCover lc 
    ON nj.membership_id = lc.membership_id 
    AND lc.rn = 1
LEFT JOIN LatestMembership lm 
    ON nj.membership_id = lm.membership_id 
    AND lm.rn = 1 
    AND lm.relationship = 1
LEFT JOIN LatestPromo lp 
    ON lm.person_id = lp.person_id 
    AND lp.rn = 1
LEFT JOIN BRONZE.dbo.promotion p 
    ON lp.promotion_id = p.promotion_id
LEFT JOIN BRONZE.dbo.promotion_sales_channel ps 
    ON lp.sales_channel_id = ps.sales_channel_id
LEFT JOIN BRONZE.dbo.operator o 
    ON lp.operator = o.oper_name
LEFT JOIN BRONZE.dbo.grouping g 
    ON lp.agency_group_id = g.group_id;
GO
```

# [dbo].[vw_RebateLineCheck]

```sql
/****** Object:  View [dbo].[vw_RebateLineCheck]    Script Date: 4/06/2026 10:39:10 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE   VIEW [dbo].[vw_RebateLineCheck]
AS
-- 1. Active members
WITH ActiveMembers AS (
    SELECT 
        membership_id AS MemberID,
        memship_status AS Status
    FROM BRONZE.dbo.memship
    WHERE memship_status = 'A'
),

-- 2. Main member rebate info
MemberRebateInfo AS (
    SELECT 
        am.MemberID,
        mr.person_id AS PersonID,
        mr.in_rebate AS InRebate2,
        mr.rebate_form_flag AS RebateFormFlag
    FROM ActiveMembers am
    INNER JOIN BRONZE.dbo.MemberRebate mr
        ON mr.membership_id = am.MemberID
),

-- 3. Non-primary persons (relationship > 1)
NonPrimaryPersons AS (
    SELECT 
        mpi.membership_id AS MemberID,
        mpi.person_id AS PersonID,
        mpi.relationship AS PersonRelationship,
        mpi.status_flag AS PersonStatus
    FROM BRONZE.dbo.person_membership mpi
    WHERE mpi.relationship > 1
),

-- 4. Person names
PersonNames AS (
    SELECT
        p.person_id AS PersonID,
        p.first_name + ' ' + p.surname AS RebateLineName
    FROM BRONZE.dbo.person p
),

-- 5. Main member info
MainMemberInfo AS (
    SELECT
        pc.membership_id AS MemberID,
        pc.person_id AS MainMemberPersonID,
        pc.first_name + ' ' + pc.surname AS MainMemberName
    FROM BRONZE.dbo.PersonContact pc
    WHERE pc.relationship = 1
),

-- 6. In Rebate description mapping
InRebateDescription AS (
    SELECT DISTINCT
        in_rebate AS InRebate2,
        description AS InRebateDescription
    FROM BRONZE.dbo.in_rebate
)

-- Final view
SELECT 
    am.MemberID,
    am.Status,
    np.PersonID,
    np.PersonRelationship,
    np.PersonStatus,
    mm.MainMemberPersonID,
    mm.MainMemberName,
    mr.InRebate2,
    ir.InRebateDescription,
    mr.RebateFormFlag,
    pn.RebateLineName
FROM ActiveMembers am
INNER JOIN MemberRebateInfo mr
    ON mr.MemberID = am.MemberID
INNER JOIN NonPrimaryPersons np
    ON np.MemberID = mr.MemberID
    AND np.PersonID = mr.PersonID
INNER JOIN PersonNames pn
    ON pn.PersonID = np.PersonID
INNER JOIN MainMemberInfo mm
    ON mm.MemberID = am.MemberID
LEFT JOIN InRebateDescription ir
    ON ir.InRebate2 = mr.InRebate2;
GO
```

# [dbo].[vw_RebateRegistrations]

```sql
/****** Object:  View [dbo].[vw_RebateRegistrations]    Script Date: 4/06/2026 10:39:10 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE   VIEW [dbo].[vw_RebateRegistrations]
AS
WITH RebateBase AS (
    SELECT 
        mr.membership_id        AS [Membership ID],
        m.memship_status        AS [Membership Status],        -- ② 新增 Membership Status
        mr.rebate_version       AS [Rebate Version],
        mr.in_rebate            AS [In Rebate],
        CASE 
            WHEN mr.in_rebate = 'Y' THEN 'Yes'
            WHEN mr.in_rebate = 'F' THEN 'Final'
            ELSE 'Unknown'
        END                     AS [In Rebate Description],
        mr.receive_date         AS [Receive Date]
    FROM BRONZE.dbo.MemberRebate mr
    INNER JOIN BRONZE.dbo.memship m                            -- ① 新增 Active Member 过滤
        ON mr.membership_id = m.membership_id
        AND m.memship_status = 'A'
    WHERE mr.in_rebate IN ('Y','F')
),
RebateStatus AS (
    SELECT 
        rr.membership_id        AS [Membership ID],
        rr.rebate_version       AS [Rebate Version],
        CONVERT(date, rr.status_date) AS [Status Date],
        CASE 
            WHEN CONVERT(date, rr.status_date) = CONVERT(date, DATEADD(day, -1, GETDATE())) 
                THEN 'NPrint' 
            ELSE 'Not NPrint' 
        END                     AS [NPrint Flag],
        rr.hic_rebate_error     AS [HIC Rebate Error],
        ISNULL(he.description, 'Unknown') AS [HIC Rebate Error Description],
        rr.data_element_id      AS [Data Element ID 1],
        rr.hic_rebate_segment   AS [HIC Rebate Segment],
        ISNULL(hs.description, 'Unknown') AS [HIC Rebate Segment Description],
        NULL                    AS [Data Element ID 2],
        rr.hic_rebate_message_id
    FROM BRONZE.dbo.rebate_reg_errors rr
    LEFT JOIN BRONZE.dbo.hic_rebate_error he 
        ON rr.hic_rebate_error = he.hic_rebate_error
    LEFT JOIN BRONZE.dbo.hic_rebate_segment hs 
        ON rr.hic_rebate_segment = hs.hic_rebate_segment
    WHERE rr.hic_rebate_error <> 0
),
SegmentErrors AS (
    SELECT 
        hse.hic_rebate_segment  AS [HIC Rebate Segment],
        hse.data_element_id     AS [Data Element ID 2]
    FROM BRONZE.dbo.hic_rebate_segment_error hse
)
SELECT 
    rb.[Membership ID],
    rb.[Membership Status],                                    -- ② 新增字段
    rb.[Rebate Version],
    rb.[In Rebate],
    rb.[In Rebate Description],
    rb.[Receive Date],
    rs.[Status Date],
    rs.[NPrint Flag],
    rs.[HIC Rebate Error],
    rs.[HIC Rebate Error Description],
    rs.[Data Element ID 1],
    rs.[HIC Rebate Segment],
    rs.[HIC Rebate Segment Description],
    se.[Data Element ID 2],
    rs.[hic_rebate_message_id]
FROM RebateBase rb
INNER JOIN RebateStatus rs
    ON rb.[Membership ID] = rs.[Membership ID]
    AND rb.[Rebate Version] = rs.[Rebate Version]
LEFT JOIN SegmentErrors se
    ON rs.[HIC Rebate Segment] = se.[HIC Rebate Segment];    -- ③ 保持原 join key 不变
GO
```

# [dbo].[vw_RebateReminders]

```sql
/****** Object:  View [dbo].[vw_RebateReminders]    Script Date: 4/06/2026 10:39:10 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE   VIEW [dbo].[vw_RebateReminders]
AS
SELECT
    -- Step 1: BRONZE.dbo.memship
    m.membership_id,
    m.memship_status,
    m.effective_join_date,
    m.effective_termination_date,
    m.date_paidto,
    m.contribution_due,
    m.rejoin_date,
    m.create_operator                                               AS [Membership Create Operator],
    m.create_datetime,

    -- Step 1: Branch and Cover via mapping tables
    mb.description                                                  AS Branch,
    mc.Product_Description                                          AS Cover,

    -- Step 2: BRONZE.dbo.MemberRebate
    mr.rebate_version,
    mr.rebate_product_type,
    mr.rebate_category,
    mr.rebate_applicant,
    mr.rebate_year,
    mr.person_id,
    mr.in_rebate,
    mr.receive_date,
    mr.entry_date,
    CAST(mr.effective_date AS DATE)                                 AS [Last Effective Date],
    mr.suspend_flag,
    mr.declaration_flag,
    mr.create_operator                                              AS [Rebate Create Operator],
    CAST(mr.create_datetime AS DATE)                                AS [Rebate Create Date],
    DATEDIFF(DAY, CAST(mr.create_datetime AS DATE), CAST(GETDATE() AS DATE))
                                                                    AS DaysDiff,
    mr.rebate_form_flag,
    mr.rebate_tier_id,
    rff.description                                                 AS [Rebate Form Flag Description],

    -- Step 2: Max Form Flag from BRONZE.dbo.rebate subquery
    reb.max_form_flag,
    rff_max.description                                             AS [Max Form Flag],

    -- Step 3: BRONZE.dbo.person_membership (relationship = 1)
    pm.relationship,
    pm.status_flag,
    pm.eligibility_status,
    pm.newborn_source,
    pm.newborn_rebate_selected,
    pm.rebate_form_printed,

    -- Step 4: BRONZE.dbo.person
    p.first_name + ' ' + p.surname                                 AS [Person Name],
    p.medicare_number                                               AS [Medicare Number],

    -- Step 5: BRONZE.dbo.PersonContact (relationship = 1)
    pc.title_text,
    pc.first_name,
    pc.surname,
    pc.detailH,
    pc.detailW,
    pc.detailF,
    pc.status_flag                                                  AS [Contact Status Flag],
    pc.date_of_birth,
    CASE
        WHEN LEN(pc.detailE) < 3        THEN 'No Email'
        WHEN pc.detailE LIKE '%@%'      THEN 'Email Supplied'
        ELSE 'No Email'
    END                                                             AS detailE,
    CASE
        WHEN pc.detailM IS NULL         THEN pc.detailH
        ELSE pc.detailM
    END                                                             AS [Contact Number],

    -- Step 6: BRONZE.dbo.PersonAddressHomePostal
    pa.address_line1 + ' ' + ISNULL(pa.address_line2, '') + ', '
        + pa.suburb + ', ' + CAST(pa.postcode AS VARCHAR(10)) + ', ' + pa.state
                                                                    AS Address,

    -- Step 7: BRONZE.dbo.web_security
    ws.postal_preference                                            AS [Postal Preference],

    -- Step 8: Postal Preference Description
    CASE
        WHEN LEN(pc.detailE) < 3 OR pc.detailE NOT LIKE '%@%'  THEN 'Post'
        WHEN ws.postal_preference = 'E'                         THEN 'Email'
        WHEN ws.postal_preference = 'P'                         THEN 'Post'
        WHEN ws.postal_preference IS NULL                       THEN 'Not Registered to MO'
        ELSE 'Not Registered to MO'
    END                                                             AS [Postal Preference Description],

    -- Step 9: Days Bracket
    CASE
        WHEN DATEDIFF(DAY, CAST(mr.create_datetime AS DATE), CAST(GETDATE() AS DATE)) BETWEEN 0 AND 6   THEN '1 to 6 days'
        WHEN DATEDIFF(DAY, CAST(mr.create_datetime AS DATE), CAST(GETDATE() AS DATE)) = 7               THEN '7 Day Reminder'
        WHEN DATEDIFF(DAY, CAST(mr.create_datetime AS DATE), CAST(GETDATE() AS DATE)) BETWEEN 8 AND 13  THEN '8 to 13 days'
        WHEN DATEDIFF(DAY, CAST(mr.create_datetime AS DATE), CAST(GETDATE() AS DATE)) = 14              THEN '14 Day Reminder'
        WHEN DATEDIFF(DAY, CAST(mr.create_datetime AS DATE), CAST(GETDATE() AS DATE)) BETWEEN 15 AND 29 THEN '15 to 29 days'
        WHEN DATEDIFF(DAY, CAST(mr.create_datetime AS DATE), CAST(GETDATE() AS DATE)) = 30              THEN '30 Day Reminder'
        WHEN DATEDIFF(DAY, CAST(mr.create_datetime AS DATE), CAST(GETDATE() AS DATE)) BETWEEN 31 AND 44 THEN '31 to 44 days'
        WHEN DATEDIFF(DAY, CAST(mr.create_datetime AS DATE), CAST(GETDATE() AS DATE)) >= 45             THEN 'Terminate - 45 Days'
        ELSE 'Fix Me'
    END                                                             AS [Days Bracket]

FROM BRONZE.dbo.memship m

-- Step 1: Branch mapping
LEFT JOIN BRONZE.dbo.MemberBranch mb
    ON mb.membership_id = m.membership_id

-- Step 1: Cover mapping
LEFT JOIN BRONZE.dbo.MemberCover mc
    ON mc.membership_id = m.membership_id

-- Step 2: Inner Join MemberRebate
INNER JOIN BRONZE.dbo.MemberRebate mr
    ON mr.membership_id = m.membership_id

-- Step 2: Rebate Form Flag description
LEFT JOIN BRONZE.dbo.rebate_form_flag rff
    ON rff.rebate_form_flag = mr.rebate_form_flag

-- Step 2: Max Form Flag from latest rebate_version per membership
LEFT JOIN (
    SELECT re.membership_id, re.rebate_form_flag AS max_form_flag
    FROM BRONZE.dbo.rebate re
    WHERE re.rebate_version = (
        SELECT MAX(rebate_version)
        FROM BRONZE.dbo.rebate
        WHERE membership_id = re.membership_id
    )
) reb ON reb.membership_id = mr.membership_id

-- Step 2: Max Form Flag description
LEFT JOIN BRONZE.dbo.rebate_form_flag rff_max
    ON rff_max.rebate_form_flag = reb.max_form_flag

-- Step 3: Inner Join person_membership (primary member only, relationship = '1')
INNER JOIN BRONZE.dbo.person_membership pm
    ON  pm.membership_id = m.membership_id
    AND pm.relationship  = '1'

-- Step 4: Inner Join person
INNER JOIN BRONZE.dbo.person p
    ON p.person_id = pm.person_id

-- Step 5: Inner Join PersonContact (primary member only, relationship = '1')
INNER JOIN BRONZE.dbo.PersonContact pc
    ON  pc.membership_id = m.membership_id
    AND pc.relationship  = '1'

-- Step 6: Inner Join PersonAddressHomePostal
INNER JOIN BRONZE.dbo.PersonAddressHomePostal pa
    ON  pa.membership_id = m.membership_id
    AND pa.relationship  = '1'

-- Step 7: Left Join web_security (main_ref_type = 'M')
LEFT JOIN BRONZE.dbo.web_security ws
    ON  ws.main_ref_id   = m.membership_id
    AND ws.main_ref_type = 'M'

-- Step 1 filter: Active members only
WHERE m.memship_status = 'A';
GO
```

# [GOV].[vw_glossary_hierarchy]

```sql
/****** Object:  View [GOV].[vw_glossary_hierarchy]    Script Date: 4/06/2026 10:39:10 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

;CREATE VIEW [GOV].[vw_glossary_hierarchy] AS

WITH

-- -----------------------------------------------------------------------
-- Level 1: Business Domain
-- -----------------------------------------------------------------------
domain AS (
    SELECT
        Nick_Name   AS Domain_Nick_Name,
        [Name]      AS Domain_Name
    FROM SILVER.GOV.glossary_full_cleaned
    WHERE REPLACE(Term_Template_Names, ';', '') = 'Business Domain'
),

-- -----------------------------------------------------------------------
-- Level 2: Business Sub Domain
-- -----------------------------------------------------------------------
sub_domain AS (
    SELECT
        Nick_Name           AS Sub_Domain_Nick_Name,
        [Name]              AS Sub_Domain_Name,
        Parent_Term_Name    AS Sub_Domain_Parent  -- joins to Domain.Domain_Nick_Name
    FROM SILVER.GOV.glossary_full_cleaned
    WHERE REPLACE(Term_Template_Names, ';', '') = 'Business Sub Domain'
)

-- -----------------------------------------------------------------------
-- Main query: all 62 rows, with hierarchy columns and unified Attributes
-- -----------------------------------------------------------------------
SELECT

    -- Core identity
    g.Item_ID,
    g.Nick_Name,
    g.[Name],
    g.[Definition],
    g.[Status],
    g.Acronym,
    g.Synonyms,
    g.Related_Terms,
    g.Experts,
    g.Stewards,
    g.Resources,
    g.IsDefinitionRichText,

    -- Hierarchy columns
    CASE
        WHEN REPLACE(g.Term_Template_Names, ';', '') = 'Business Domain'
            THEN g.Nick_Name
        WHEN REPLACE(g.Term_Template_Names, ';', '') = 'Business Sub Domain'
            THEN sd_direct.Sub_Domain_Parent
        ELSE
            d.Domain_Nick_Name
    END                                                     AS Business_Domain,

    CASE
        WHEN REPLACE(g.Term_Template_Names, ';', '') = 'Business Sub Domain'
            THEN g.Nick_Name
        WHEN REPLACE(g.Term_Template_Names, ';', '') NOT IN ('Business Domain', 'Business Sub Domain')
            THEN sd.Sub_Domain_Nick_Name
        ELSE NULL
    END                                                     AS Business_Sub_Domain,

    CASE
        WHEN REPLACE(g.Term_Template_Names, ';', '')
             IN ('Business Metric', 'Data Element', 'Business Terms')
            THEN g.Nick_Name
        ELSE NULL
    END                                                     AS Term_Name,

    REPLACE(g.Term_Template_Names, ';', '')                 AS Term_Type,

    -- Unified Attribute columns
    COALESCE(
        g.Attribute_Business_Metric_Business_Data_Product,
        g.Attribute_Data_Element_Business_Data_Product,
        g.Attribute_Business_Terms_Business_Data_Product
    )                                                       AS Business_Data_Product,

    COALESCE(
        g.Attribute_Business_Metric_Representation_Terms,
        g.Attribute_Data_Element_Representation_Terms
    )                                                       AS Representation_Terms,

    COALESCE(
        g.Attribute_Business_Metric_Sample_Values,
        g.Attribute_Data_Element_Sample_Values
    )                                                       AS Sample_Values,

    g.Attribute_Data_Element_Comments                       AS Comments,
    g.Attribute_Data_Element_Criticality_Flag               AS Criticality_Flag,
    g.Attribute_Data_Element_Criticality_Reason             AS Criticality_Reason,
    g.Attribute_Data_Element_Sensitivity_Classification     AS Sensitivity_Classification

FROM SILVER.GOV.glossary_full_cleaned AS g

-- Join 1: Term -> Sub Domain (via Term.Parent_Term_Name = Sub Domain.Name)
LEFT JOIN sub_domain AS sd
    ON  g.Parent_Term_Name = sd.Sub_Domain_Name
    AND REPLACE(g.Term_Template_Names, ';', '')
        IN ('Business Metric', 'Data Element', 'Business Terms')

-- Join 2: Sub Domain (resolved from Term) -> Domain
LEFT JOIN domain AS d
    ON sd.Sub_Domain_Parent = d.Domain_Nick_Name

-- Join 3: Sub Domain row direct -> its own Parent (= Domain Nick_Name)
LEFT JOIN sub_domain AS sd_direct
    ON  g.[Name] = sd_direct.Sub_Domain_Name
    AND REPLACE(g.Term_Template_Names, ';', '') = 'Business Sub Domain';
GO
```
