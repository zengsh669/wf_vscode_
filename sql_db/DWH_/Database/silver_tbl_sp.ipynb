# [dbo].[Load_ArrearsReportPayroll]

```sql
/****** Object:  StoredProcedure [dbo].[Load_ArrearsReportPayroll]    Script Date: 29/06/2026 1:38:34 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE   PROCEDURE [dbo].[Load_ArrearsReportPayroll]
AS
BEGIN
    SET NOCOUNT ON;

    TRUNCATE TABLE dbo.Arrears_Report;

    WITH ArrearsBase AS (
        SELECT
            m.membership_id                             AS [Membership ID],
            m.memship_status,
            CAST(m.date_paidto AS DATE)                 AS [Date Paid To],
            ISNULL(s.description, 'Unknown')            AS [Security Level]
        FROM BRONZE.dbo.memship AS m
        LEFT JOIN BRONZE.dbo.security_level AS s
            ON m.security_level = s.security_level
        WHERE m.memship_status = 'A'
    ),
    JoinedData AS (
        SELECT
            a.[Membership ID],
            a.memship_status,
            a.[Date Paid To],
            a.[Security Level],
            pm.person_id,
            pm.relationship,
            mb.description                              AS Branch,
            mc.FixCode,
            mc.Product_Description,
            mg.group_id,
            p.surname,
            p.first_name,
            p.first_name + ' ' + p.surname             AS name,
            g.description                              AS GroupingDesc,
            bg.billing_type,
            bg.billing_freq,
            bg.tpt_period                              AS [Tpt Period],
            ma.description                             AS AgentDesc,
            ws.postal_preference                       AS [Post Preference],
            ws.main_ref_type                           AS [Main Ref Type]
        FROM ArrearsBase AS a
        LEFT JOIN BRONZE.dbo.person_membership AS pm
            ON a.[Membership ID] = pm.membership_id
            AND pm.relationship = 1
        LEFT JOIN BRONZE.dbo.MemberBranch AS mb
            ON a.[Membership ID] = mb.membership_id
        LEFT JOIN BRONZE.dbo.MemberCover AS mc
            ON a.[Membership ID] = mc.membership_id
        LEFT JOIN BRONZE.dbo.MemberGroup AS mg
            ON a.[Membership ID] = mg.membership_id
        LEFT JOIN BRONZE.dbo.person AS p
            ON pm.person_id = p.person_id
        LEFT JOIN BRONZE.dbo.grouping AS g
            ON mg.group_id = g.group_id
        LEFT JOIN BRONZE.dbo.billing_group AS bg
            ON mg.group_id = bg.group_id
        LEFT JOIN BRONZE.dbo.MemberAgent AS ma
            ON a.[Membership ID] = ma.membership_id
        LEFT JOIN BRONZE.dbo.web_security AS ws
            ON a.[Membership ID] = ws.main_ref_id
            AND ws.main_ref_type = 'M'
    ),
    RegisteredMO AS (
        SELECT
            j.*,
            CASE
                WHEN j.[Post Preference] = 'E' THEN 'Email'
                WHEN j.[Post Preference] = 'P' THEN 'Postal'
                ELSE 'Not Registered to MO'
            END AS [Postal Preference]
        FROM JoinedData AS j
    ),
    BillingTypeJoin AS (
        SELECT
            r.*,
            bt.description AS BillingDes,
            CASE
                WHEN bt.description = 'Payroll'          THEN 'Payroll'
                WHEN bt.description LIKE '%Direct%'      THEN 'Direct Debit/Payer'
                ELSE 'Other'
            END AS [Payment Type]
        FROM RegisteredMO AS r
        INNER JOIN BRONZE.dbo.billing_type AS bt
            ON r.billing_type = bt.billing_type
    ),
    BillingFreqMap AS (
        SELECT *
        FROM (VALUES
            (0, 'd', 1),
            (1, 'd', 7),
            (2, 'd', 14),
            (3, 'd', 28),
            (4, 'm', 1),
            (5, 'm', 2),
            (6, 'm', 3),
            (7, 'm', 6),
            (8, 'm', 12),
            (9, 'd', 1)
        ) AS f(billing_freq, typebill, amount)
    ),
    Bucket1 AS (
        SELECT
            b.*,
            f.typebill,
            f.amount,
            CASE
                WHEN b.[Security Level] = 'No Access Restrictions' THEN b.name
                ELSE b.[Security Level]
            END AS [Name],
            CASE
                WHEN f.typebill = 'd' THEN DATEADD(DAY,   f.amount * b.[Tpt Period], b.[Date Paid To])
                WHEN f.typebill = 'm' THEN DATEADD(MONTH, f.amount * b.[Tpt Period], b.[Date Paid To])
            END AS [tpt_day],
            CASE
                WHEN b.Product_Description = 'Ambulance' THEN 'Ambulance Product'
                ELSE 'Non Ambulance Products'
            END AS [Directdebitfilter]
        FROM BillingTypeJoin AS b
        INNER JOIN BillingFreqMap AS f
            ON b.billing_freq = f.billing_freq
    ),
    Bucket2 AS (
        SELECT
            *,
            DATEDIFF(DAY, [Date Paid To], CAST(GETDATE() AS DATE))                         AS [DaysInArrears],
            CAST(GETDATE() AS DATE)                                                         AS [Run Date],
            FORMAT(DATEFROMPARTS(YEAR(GETDATE()), MONTH(GETDATE()), 1), 'MMM yyyy')        AS [Run Month]
        FROM Bucket1
    ),
    Bucket3 AS (
        SELECT
            *,
            CASE
                WHEN [DaysInArrears] > 57                        THEN '58 Days plus in arrears'
                WHEN [DaysInArrears] BETWEEN 45 AND 57           THEN '45-57 Days in arrears'
                WHEN [DaysInArrears] BETWEEN 30 AND 44           THEN '30-44 Days in arrears'
                WHEN [DaysInArrears] BETWEEN 15 AND 29           THEN '15-29 Days in arrears'
                WHEN [DaysInArrears] BETWEEN 1  AND 14           THEN 'Less than 15 Days in arrears'
                ELSE 'No Arrears'
            END AS [Arrears Category]
        FROM Bucket2
    )

    INSERT INTO dbo.Arrears_Report (
        [Membership ID],
        memship_status,
        [Date Paid To],
        [Security Level],
        person_id,
        relationship,
        Branch,
        FixCode,
        Product_Description,
        group_id,
        surname,
        first_name,
        name,
        GroupingDesc,
        billing_type,
        billing_freq,
        [Tpt Period],
        AgentDesc,
        [Post Preference],
        [Main Ref Type],
        BillingDes,
        [Payment Type],
        typebill,
        amount,
        tpt_day,
        Directdebitfilter,
        [DaysInArrears],
        [Run Date],
        [Run Month],
        [Arrears Category]
    )
    SELECT
        [Membership ID],
        memship_status,
        [Date Paid To],
        [Security Level],
        CAST(person_id   AS NVARCHAR(50)),
        CAST(relationship AS NVARCHAR(10)),
        Branch,
        FixCode,
        Product_Description,
        CAST(group_id    AS NVARCHAR(50)),
        surname,
        first_name,
        name,
        GroupingDesc,
        billing_type,
        CAST(billing_freq  AS NVARCHAR(10)),
        CAST([Tpt Period]  AS NVARCHAR(10)),
        AgentDesc,
        [Postal Preference],
        [Main Ref Type],
        BillingDes,
        [Payment Type],
        typebill,
        CAST(amount        AS NVARCHAR(10)),
        tpt_day,
        Directdebitfilter,
        CAST([DaysInArrears] AS NVARCHAR(10)),
        [Run Date],
        [Run Month],
        [Arrears Category]
    FROM Bucket3;

END;
GO
```

# [dbo].[Load_Claim_Fact]

```sql
/****** Object:  StoredProcedure [dbo].[Load_Claim_Fact]    Script Date: 29/06/2026 1:38:34 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[Load_Claim_Fact]
AS
BEGIN
    SET NOCOUNT ON;

    /* -----------------------
       DATE PARAMETERS
       ----------------------- */
    DECLARE @StartDate DATE = '2023-07-01';
    DECLARE @EndDate   DATE = '2026-05-31';

    /* -----------------------
       CLEAR TARGET TABLE
       ----------------------- */
    TRUNCATE TABLE dbo.Claim_Fact;

    /* -----------------------
       INSERT DATA (LOGIC IDENTICAL TO ORIGINAL QUERY)
       ----------------------- */
    ;WITH All_Claims AS (
        -- Ambulatory and Medical claims
        SELECT
            a.claim_id,
            a.claim_line_id,
            a.provider_number_id,
            a.episode_id,
            a.claim_type,
            a.membership_id,
            a.person_id,
            a.service_date,
            a.claim_line_status_type,
            a.status_date,
            a.product_id AS amb_id,
            b.service_type,
            b.num_services,
            b.fee,
            b.benefit,
            b.excess,
            0 AS bed_days,
            'NA' AS stay_type,
            CASE WHEN (DATEDIFF(year, p.date_of_birth, a.service_date) - CASE WHEN a.service_date < DATEADD(year, DATEDIFF(year, p.date_of_birth, a.service_date), p.date_of_birth) THEN 1 ELSE 0 END) BETWEEN 0 AND 54 THEN '00-54'
                 WHEN (DATEDIFF(year, p.date_of_birth, a.service_date) - CASE WHEN a.service_date < DATEADD(year, DATEDIFF(year, p.date_of_birth, a.service_date), p.date_of_birth) THEN 1 ELSE 0 END) BETWEEN 55 AND 59 THEN '55-59'
                 WHEN (DATEDIFF(year, p.date_of_birth, a.service_date) - CASE WHEN a.service_date < DATEADD(year, DATEDIFF(year, p.date_of_birth, a.service_date), p.date_of_birth) THEN 1 ELSE 0 END) BETWEEN 60 AND 64 THEN '60-64'
                 WHEN (DATEDIFF(year, p.date_of_birth, a.service_date) - CASE WHEN a.service_date < DATEADD(year, DATEDIFF(year, p.date_of_birth, a.service_date), p.date_of_birth) THEN 1 ELSE 0 END) BETWEEN 65 AND 69 THEN '65-69'
                 WHEN (DATEDIFF(year, p.date_of_birth, a.service_date) - CASE WHEN a.service_date < DATEADD(year, DATEDIFF(year, p.date_of_birth, a.service_date), p.date_of_birth) THEN 1 ELSE 0 END) BETWEEN 70 AND 74 THEN '70-74'
                 WHEN (DATEDIFF(year, p.date_of_birth, a.service_date) - CASE WHEN a.service_date < DATEADD(year, DATEDIFF(year, p.date_of_birth, a.service_date), p.date_of_birth) THEN 1 ELSE 0 END) BETWEEN 75 AND 79 THEN '75-79'
                 WHEN (DATEDIFF(year, p.date_of_birth, a.service_date) - CASE WHEN a.service_date < DATEADD(year, DATEDIFF(year, p.date_of_birth, a.service_date), p.date_of_birth) THEN 1 ELSE 0 END) BETWEEN 80 AND 84 THEN '80-84'
                 WHEN (DATEDIFF(year, p.date_of_birth, a.service_date) - CASE WHEN a.service_date < DATEADD(year, DATEDIFF(year, p.date_of_birth, a.service_date), p.date_of_birth) THEN 1 ELSE 0 END) >= 85 THEN '85+'
            ELSE 'NA' END AS age_group
        FROM BRONZE.dbo.claim_line a
        LEFT JOIN BRONZE.dbo.claim_generalitem b
            ON a.claim_id = b.claim_id
           AND a.claim_line_id = b.claim_line_id
        LEFT JOIN BRONZE.dbo.person p
            ON a.person_id = p.person_id
        WHERE a.status_date BETWEEN @StartDate AND @EndDate
          AND a.claim_type IN ('A','B','M')
          AND a.claim_line_status_type = 'P'

        UNION ALL

        -- Hospital claims
        SELECT
            a.claim_id,
            a.claim_line_id,
            a.provider_number_id,
            a.episode_id,
            a.claim_type,
            a.membership_id,
            a.person_id,
            a.service_date,
            a.claim_line_status_type,
            a.status_date,
            a.product_id AS amb_id,
            b.service_type,
            b.num_services,
            b.fee,
            b.benefit,
            b.excess,
            b.num_beddays,
            CASE
                WHEN b.admission_date = b.discharge_date THEN 'SD'
                ELSE 'OV'
            END AS stay_type,
            CASE WHEN (DATEDIFF(year, p.date_of_birth, a.service_date) - CASE WHEN a.service_date < DATEADD(year, DATEDIFF(year, p.date_of_birth, a.service_date), p.date_of_birth) THEN 1 ELSE 0 END) BETWEEN 0 AND 54 THEN '00-54'
                 WHEN (DATEDIFF(year, p.date_of_birth, a.service_date) - CASE WHEN a.service_date < DATEADD(year, DATEDIFF(year, p.date_of_birth, a.service_date), p.date_of_birth) THEN 1 ELSE 0 END) BETWEEN 55 AND 59 THEN '55-59'
                 WHEN (DATEDIFF(year, p.date_of_birth, a.service_date) - CASE WHEN a.service_date < DATEADD(year, DATEDIFF(year, p.date_of_birth, a.service_date), p.date_of_birth) THEN 1 ELSE 0 END) BETWEEN 60 AND 64 THEN '60-64'
                 WHEN (DATEDIFF(year, p.date_of_birth, a.service_date) - CASE WHEN a.service_date < DATEADD(year, DATEDIFF(year, p.date_of_birth, a.service_date), p.date_of_birth) THEN 1 ELSE 0 END) BETWEEN 65 AND 69 THEN '65-69'
                 WHEN (DATEDIFF(year, p.date_of_birth, a.service_date) - CASE WHEN a.service_date < DATEADD(year, DATEDIFF(year, p.date_of_birth, a.service_date), p.date_of_birth) THEN 1 ELSE 0 END) BETWEEN 70 AND 74 THEN '70-74'
                 WHEN (DATEDIFF(year, p.date_of_birth, a.service_date) - CASE WHEN a.service_date < DATEADD(year, DATEDIFF(year, p.date_of_birth, a.service_date), p.date_of_birth) THEN 1 ELSE 0 END) BETWEEN 75 AND 79 THEN '75-79'
                 WHEN (DATEDIFF(year, p.date_of_birth, a.service_date) - CASE WHEN a.service_date < DATEADD(year, DATEDIFF(year, p.date_of_birth, a.service_date), p.date_of_birth) THEN 1 ELSE 0 END) BETWEEN 80 AND 84 THEN '80-84'
                 WHEN (DATEDIFF(year, p.date_of_birth, a.service_date) - CASE WHEN a.service_date < DATEADD(year, DATEDIFF(year, p.date_of_birth, a.service_date), p.date_of_birth) THEN 1 ELSE 0 END) >= 85 THEN '85+'
            ELSE 'NA' END AS age_group
        FROM BRONZE.dbo.claim_line a
        LEFT JOIN BRONZE.dbo.claim_hospitalitem b
            ON a.claim_id = b.claim_id
           AND a.claim_line_id = b.claim_line_id
        LEFT JOIN BRONZE.dbo.person p
            ON a.person_id = p.person_id
        WHERE a.status_date BETWEEN @StartDate AND @EndDate
          AND a.claim_type = 'H'
          AND a.claim_line_status_type = 'P'
    ),
    Member_details AS (
        SELECT
            cp.membership_id,
            cp.cover_version,
            cp.product_id,
            p.product_type,
            p.product_code,
            c.cover_type,
            c.cover_state,
            c.cover_from_date,
            c.termination_date,
            c.description
        FROM BRONZE.dbo.cover_product cp
        LEFT JOIN BRONZE.dbo.product p
            ON cp.product_id = p.product_id
        LEFT JOIN BRONZE.dbo.cover c
            ON c.membership_id = cp.membership_id
           AND c.cover_version = cp.cover_version
    ),
    Provider_details AS (
        SELECT
            pn.provider_number_id,
            pn.provider_id,
            pt.provider_type
        FROM BRONZE.dbo.provider_number pn
        LEFT JOIN BRONZE.dbo.provider pt
            ON pn.provider_id = pt.provider_id
    ),
    Combined_data AS (
        -- Hospital & Medical claims
        SELECT
            ac.claim_id,
            ac.claim_type,
            ac.membership_id,
            ac.person_id,
            ac.service_date,
            ac.status_date,
            ac.service_type,
            ac.stay_type,
            SUM(ac.num_services) AS serv,
            SUM(ac.fee) AS fee_amt,
            SUM(ac.benefit) AS benefit_amt,
            SUM(ac.excess) AS excess_amt,
            SUM(ac.bed_days) AS bed_days,
            md.cover_type,
            md.cover_state,
            ISNULL(md.product_id,0) AS product_id,
            ac.provider_number_id,
            pt.provider_type,
            ac.age_group,
            SUM(CASE WHEN ISNULL(md.product_id,0) NOT IN (10,34,65,191,192,193,194,195,196) AND ac.age_group = '00-54' THEN 0
                     WHEN ISNULL(md.product_id,0) NOT IN (10,34,65,191,192,193,194,195,196) AND ac.age_group = '55-59' THEN 0.15 * ac.benefit
                     WHEN ISNULL(md.product_id,0) NOT IN (10,34,65,191,192,193,194,195,196) AND ac.age_group = '60-64' THEN 0.425 * ac.benefit
                     WHEN ISNULL(md.product_id,0) NOT IN (10,34,65,191,192,193,194,195,196) AND ac.age_group = '65-69' THEN 0.60 * ac.benefit
                     WHEN ISNULL(md.product_id,0) NOT IN (10,34,65,191,192,193,194,195,196) AND ac.age_group = '70-74' THEN 0.70 * ac.benefit
                     WHEN ISNULL(md.product_id,0) NOT IN (10,34,65,191,192,193,194,195,196) AND ac.age_group = '75-79' THEN 0.76 * ac.benefit
                     WHEN ISNULL(md.product_id,0) NOT IN (10,34,65,191,192,193,194,195,196) AND ac.age_group = '80-84' THEN 0.78 * ac.benefit
                     WHEN ISNULL(md.product_id,0) NOT IN (10,34,65,191,192,193,194,195,196) AND ac.age_group = '85+' THEN 0.82 * ac.benefit
                ELSE 0 END) AS gross_deficit
        FROM All_Claims ac
        LEFT JOIN Provider_details pt
            ON ac.provider_number_id = pt.provider_number_id
        LEFT JOIN Member_details md
            ON ac.membership_id = md.membership_id
        WHERE ac.claim_type IN ('H', 'M')
          AND md.product_type = 'H'
          AND md.cover_version = (
              SELECT MAX(cover_version)
              FROM Member_details
              WHERE cover_from_date <= ac.service_date
                AND membership_id = ac.membership_id
          )
        GROUP BY
            ac.claim_id, ac.claim_type, ac.membership_id, ac.person_id,
            ac.service_date, ac.status_date, ac.service_type, ac.stay_type,
            md.cover_type, md.cover_state, md.product_id,
            ac.provider_number_id, pt.provider_type, ac.age_group

        UNION ALL

        -- Ambulatory claims
        SELECT
            ac.claim_id,
            ac.claim_type,
            ac.membership_id,
            ac.person_id,
            ac.service_date,
            ac.status_date,
            ac.service_type,
            ac.stay_type,
            SUM(ac.num_services) AS serv,
            SUM(ac.fee) AS fee_amt,
            SUM(ac.benefit) AS benefit_amt,
            SUM(ac.excess) AS excess_amt,
            SUM(ac.bed_days) AS bed_days,
            md.cover_type,
            md.cover_state,
            ISNULL(md.product_id,0) AS product_id,
            ac.provider_number_id,
            pt.provider_type,
            ac.age_group,
            0 AS gross_deficit
        FROM All_Claims ac
        LEFT JOIN Provider_details pt
            ON ac.provider_number_id = pt.provider_number_id
        LEFT JOIN Member_details md
            ON ac.membership_id = md.membership_id
        WHERE ac.claim_type = 'A'
          AND md.product_type = 'A'
          AND md.cover_version = (
              SELECT MAX(cover_version)
              FROM Member_details
              WHERE cover_from_date <= ac.service_date
                AND membership_id = ac.membership_id
          )
        GROUP BY
            ac.claim_id, ac.claim_type, ac.membership_id, ac.person_id,
            ac.service_date, ac.status_date, ac.service_type, ac.stay_type,
            md.cover_type, md.cover_state, md.product_id,
            ac.provider_number_id, pt.provider_type, ac.age_group

        UNION ALL

        -- B claims
        SELECT
            ac.claim_id,
            ac.claim_type,
            ac.membership_id,
            ac.person_id,
            ac.service_date,
            ac.status_date,
            ac.service_type,
            ac.stay_type,
            SUM(ac.num_services) AS serv,
            SUM(ac.fee) AS fee_amt,
            SUM(ac.benefit) AS benefit_amt,
            SUM(ac.excess) AS excess_amt,
            SUM(ac.bed_days) AS bed_days,
            md.cover_type,
            md.cover_state,
            ISNULL(md.product_id,0) AS product_id,
            ac.provider_number_id,
            pt.provider_type,
            ac.age_group,
            0 AS gross_deficit
        FROM All_Claims ac
        LEFT JOIN Provider_details pt
            ON ac.provider_number_id = pt.provider_number_id
        LEFT JOIN Member_details md
            ON ac.membership_id = md.membership_id
           AND ac.amb_id = md.product_id
        WHERE ac.claim_type = 'B'
          AND md.cover_version = (
              SELECT MAX(cover_version)
              FROM Member_details
              WHERE cover_from_date <= ac.service_date
                AND membership_id = ac.membership_id
          )
        GROUP BY
            ac.claim_id, ac.claim_type, ac.membership_id, ac.person_id,
            ac.service_date, ac.status_date, ac.service_type, ac.stay_type,
            md.cover_type, md.cover_state, md.product_id,
            ac.provider_number_id, pt.provider_type, ac.age_group
    )
    INSERT INTO dbo.Claim_Fact
    SELECT *
    FROM Combined_data
    WHERE service_date >= @StartDate
      AND benefit_amt <> 0
      AND claim_type IN ('A','B','M','H');

END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[Load_RebateReminders]    Script Date: 29/06/2026 1:38:34 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[Load_RebateReminders]
AS
BEGIN
    SET NOCOUNT ON;

    IF OBJECT_ID('SILVER.dbo.RebateReminders', 'U') IS NOT NULL
        DROP TABLE SILVER.dbo.RebateReminders;

    SELECT
        -- Step 1: BRONZE.dbo.memship
        m.membership_id,
        m.memship_status,
        m.effective_join_date,
        m.effective_termination_date,
        m.date_paidto,
        m.contribution_due,
        m.rejoin_date,
        m.create_operator                                               AS [Membership Create Operator],
        m.create_datetime,

        -- Step 1: Branch and Cover via mapping tables
        mb.description                                                  AS Branch,
        mc.Product_Description                                          AS Cover,

        -- Step 2: BRONZE.dbo.MemberRebate
        mr.rebate_version,
        mr.rebate_product_type,
        mr.rebate_category,
        mr.rebate_applicant,
        mr.rebate_year,
        mr.person_id,
        mr.in_rebate,
        mr.receive_date,
        mr.entry_date,
        CAST(mr.effective_date AS DATE)                                 AS [Last Effective Date],
        mr.suspend_flag,
        mr.declaration_flag,
        mr.create_operator                                              AS [Rebate Create Operator],
        CAST(mr.create_datetime AS DATE)                                AS [Rebate Create Date],
        DATEDIFF(DAY, CAST(mr.create_datetime AS DATE), CAST(GETDATE() AS DATE))
                                                                        AS DaysDiff,
        mr.rebate_form_flag,
        mr.rebate_tier_id,
        rff.description                                                 AS [Rebate Form Flag Description],

        -- Step 2: Max Form Flag from BRONZE.dbo.rebate subquery
        reb.max_form_flag,
        rff_max.description                                             AS [Max Form Flag],

        -- Step 3: BRONZE.dbo.person_membership (relationship = '1')
        pm.relationship,
        pm.status_flag,
        pm.eligibility_status,
        pm.newborn_source,
        pm.newborn_rebate_selected,
        pm.rebate_form_printed,

        -- Step 4: BRONZE.dbo.person
        p.first_name + ' ' + p.surname                                 AS [Person Name],
        p.medicare_number                                               AS [Medicare Number],

        -- Step 5: BRONZE.dbo.PersonContact (relationship = '1')
        pc.title_text,
        pc.first_name,
        pc.surname,
        pc.detailH,
        pc.detailW,
        pc.detailF,
        pc.status_flag                                                  AS [Contact Status Flag],
        pc.date_of_birth,
        CASE
            WHEN LEN(pc.detailE) < 3        THEN 'No Email'
            WHEN pc.detailE LIKE '%@%'      THEN 'Email Supplied'
            ELSE 'No Email'
        END                                                             AS detailE,
        CASE
            WHEN pc.detailM IS NULL         THEN pc.detailH
            ELSE pc.detailM
        END                                                             AS [Contact Number],

        -- Step 6: BRONZE.dbo.PersonAddressHomePostal
        pa.address_line1 + ' ' + ISNULL(pa.address_line2, '') + ', '
            + pa.suburb + ', ' + CAST(pa.postcode AS VARCHAR(10)) + ', ' + pa.state
                                                                        AS Address,

        -- Step 7: BRONZE.dbo.web_security
        ws.postal_preference                                            AS [Postal Preference],

        -- Step 8: Postal Preference Description
        CASE
            WHEN LEN(pc.detailE) < 3 OR pc.detailE NOT LIKE '%@%'  THEN 'Post'
            WHEN ws.postal_preference = 'E'                         THEN 'Email'
            WHEN ws.postal_preference = 'P'                         THEN 'Post'
            WHEN ws.postal_preference IS NULL                       THEN 'Not Registered to MO'
            ELSE 'Not Registered to MO'
        END                                                             AS [Postal Preference Description],

        -- Step 9: Days Bracket
        CASE
            WHEN DATEDIFF(DAY, CAST(mr.create_datetime AS DATE), CAST(GETDATE() AS DATE)) BETWEEN 0 AND 6   THEN '1 to 6 days'
            WHEN DATEDIFF(DAY, CAST(mr.create_datetime AS DATE), CAST(GETDATE() AS DATE)) = 7               THEN '7 Day Reminder'
            WHEN DATEDIFF(DAY, CAST(mr.create_datetime AS DATE), CAST(GETDATE() AS DATE)) BETWEEN 8 AND 13  THEN '8 to 13 days'
            WHEN DATEDIFF(DAY, CAST(mr.create_datetime AS DATE), CAST(GETDATE() AS DATE)) = 14              THEN '14 Day Reminder'
            WHEN DATEDIFF(DAY, CAST(mr.create_datetime AS DATE), CAST(GETDATE() AS DATE)) BETWEEN 15 AND 29 THEN '15 to 29 days'
            WHEN DATEDIFF(DAY, CAST(mr.create_datetime AS DATE), CAST(GETDATE() AS DATE)) = 30              THEN '30 Day Reminder'
            WHEN DATEDIFF(DAY, CAST(mr.create_datetime AS DATE), CAST(GETDATE() AS DATE)) BETWEEN 31 AND 44 THEN '31 to 44 days'
            WHEN DATEDIFF(DAY, CAST(mr.create_datetime AS DATE), CAST(GETDATE() AS DATE)) >= 45             THEN 'Terminate - 45 Days'
            ELSE 'Fix Me'
        END                                                             AS [Days Bracket]

    INTO SILVER.dbo.RebateReminders

    FROM BRONZE.dbo.memship m

    -- Step 1: Branch mapping
    LEFT JOIN BRONZE.dbo.MemberBranch mb
        ON mb.membership_id = m.membership_id

    -- Step 1: Cover mapping
    LEFT JOIN BRONZE.dbo.MemberCover mc
        ON mc.membership_id = m.membership_id

    -- Step 2: Inner Join MemberRebate
    INNER JOIN BRONZE.dbo.MemberRebate mr
        ON mr.membership_id = m.membership_id

    -- Step 2: Rebate Form Flag description
    LEFT JOIN BRONZE.dbo.rebate_form_flag rff
        ON rff.rebate_form_flag = mr.rebate_form_flag

    -- Step 2: Max Form Flag from latest rebate_version per membership
    LEFT JOIN (
        SELECT re.membership_id, re.rebate_form_flag AS max_form_flag
        FROM BRONZE.dbo.rebate re
        WHERE re.rebate_version = (
            SELECT MAX(rebate_version)
            FROM BRONZE.dbo.rebate
            WHERE membership_id = re.membership_id
        )
    ) reb ON reb.membership_id = mr.membership_id

    -- Step 2: Max Form Flag description
    LEFT JOIN BRONZE.dbo.rebate_form_flag rff_max
        ON rff_max.rebate_form_flag = reb.max_form_flag

    -- Step 3: Inner Join person_membership (primary member only, relationship = '1')
    INNER JOIN BRONZE.dbo.person_membership pm
        ON  pm.membership_id = m.membership_id
        AND pm.relationship  = '1'

    -- Step 4: Inner Join person
    INNER JOIN BRONZE.dbo.person p
        ON p.person_id = pm.person_id

    -- Step 5: Inner Join PersonContact (primary member only, relationship = '1')
    INNER JOIN BRONZE.dbo.PersonContact pc
        ON  pc.membership_id = m.membership_id
        AND pc.relationship  = '1'

    -- Step 6: Inner Join PersonAddressHomePostal
    INNER JOIN BRONZE.dbo.PersonAddressHomePostal pa
        ON  pa.membership_id = m.membership_id
        AND pa.relationship  = '1'

    -- Step 7: Left Join web_security (main_ref_type = 'M')
    LEFT JOIN BRONZE.dbo.web_security ws
        ON  ws.main_ref_id   = m.membership_id
        AND ws.main_ref_type = 'M'

    -- Step 1 filter: Active members only
    WHERE m.memship_status = 'A';

END;
GO
```

# [dbo].[LoadAgentAgreementStatus]

```sql
/****** Object:  StoredProcedure [dbo].[LoadAgentAgreementStatus]    Script Date: 29/06/2026 1:38:34 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE   PROCEDURE [dbo].[LoadAgentAgreementStatus]
AS
BEGIN
    SET NOCOUNT ON;

    /*
        This procedure extracts agent grouping data from BRONZE database,
        applies cleansing and business logic, and loads the final results
        into SILVER.dbo.AgentAgreementStatus table.
    */

    -- Optional: clear old data before inserting new
    TRUNCATE TABLE dbo.AgentAgreementStatus;

    WITH SafeGrouping AS (
        SELECT
            group_id,
            group_type,
            description,
            TRY_CONVERT(DATE, NULLIF(LTRIM(RTRIM(commencement_date)), '')) AS CommencementDate,
            TRY_CONVERT(DATE, NULLIF(LTRIM(RTRIM(termination_date)), '')) AS ExpiryDate,
            TRY_CONVERT(DATE, NULLIF(LTRIM(RTRIM(create_datetime)), '')) AS CreateDate,
            TRY_CONVERT(DATE, NULLIF(LTRIM(RTRIM(update_datetime)), '')) AS UpdateDate,
            create_operator,
            update_operator,
            grp_discount_amount
        FROM BRONZE.dbo.grouping
    ),

    BaseAgent AS (
        SELECT
            group_id AS AgentID,
            description AS Agency,
            CommencementDate,
            ExpiryDate,
            CAST(grp_discount_amount / 100.0 AS DECIMAL(5,2)) AS DiscountAmount,
            group_type
        FROM SafeGrouping
        WHERE group_type = 'A'
    ),

    AgreementStatus AS (
        SELECT *,
				CASE 
					WHEN ExpiryDate IS NULL THEN 'Check Expiry Date'
					WHEN DATEDIFF(DAY, ExpiryDate, GETDATE()) < -182 THEN 'Current'
					WHEN DATEDIFF(DAY, ExpiryDate, GETDATE()) > 0 THEN 'Expired'
					WHEN DATEDIFF(DAY, GETDATE(), ExpiryDate) <= 182 AND DATEDIFF(DAY, GETDATE(), ExpiryDate) > 152 THEN 'Expiring in 6 Months'
					WHEN DATEDIFF(DAY, GETDATE(), ExpiryDate) <= 152 AND DATEDIFF(DAY, GETDATE(), ExpiryDate) > 122 THEN 'Expiring in 5 Months'
					WHEN DATEDIFF(DAY, GETDATE(), ExpiryDate) <= 122 AND DATEDIFF(DAY, GETDATE(), ExpiryDate) > 91 THEN 'Expiring in 4 Months'
					WHEN DATEDIFF(DAY, GETDATE(), ExpiryDate) <= 91 AND DATEDIFF(DAY, GETDATE(), ExpiryDate) > 61 THEN 'Expiring in 3 Months'
					WHEN DATEDIFF(DAY, GETDATE(), ExpiryDate) <= 61 AND DATEDIFF(DAY, GETDATE(), ExpiryDate) > 30 THEN 'Expiring in 2 Months'
					WHEN DATEDIFF(DAY, GETDATE(), ExpiryDate) <= 30 AND DATEDIFF(DAY, GETDATE(), ExpiryDate) > 0 THEN 'Expiring in 1 Month'
					ELSE 'Ok'
				END AS AgreementStatus,
            CASE 
                WHEN ExpiryDate IS NULL THEN 'NPrint'
                WHEN DATEDIFF(DAY, GETDATE(), ExpiryDate) <= 182 AND DATEDIFF(DAY, GETDATE(), ExpiryDate) >= 0 THEN 'NPrint'
                ELSE NULL
            END AS NPrintFlag
        FROM BaseAgent
    ),

    CreateUpdate AS (
        SELECT
            group_id AS AgentID,
            description AS AgentName,
            CommencementDate,
            CASE 
                WHEN ExpiryDate IS NULL THEN 'Active' 
                ELSE CONVERT(VARCHAR(10), ExpiryDate, 23) 
            END AS TerminationDate,
            create_operator,
            CreateDate,
            update_operator,
            UpdateDate,
            FORMAT(CreateDate, 'yyyy-MM') AS CreateMonthyear
        FROM SafeGrouping
        WHERE group_type = 'A'
          AND description <> 'No Agency'
          AND (
              (DATEFROMPARTS(YEAR(CreateDate), MONTH(CreateDate), 1) <= EOMONTH(GETDATE())
               AND EOMONTH(CreateDate) >= DATEFROMPARTS(YEAR(GETDATE()), MONTH(GETDATE()), 1))
              OR
              (DATEFROMPARTS(YEAR(UpdateDate), MONTH(UpdateDate), 1) <= EOMONTH(GETDATE())
               AND EOMONTH(UpdateDate) >= DATEFROMPARTS(YEAR(GETDATE()), MONTH(GETDATE()), 1))
          )
    )

    INSERT INTO dbo.AgentAgreementStatus (
        AgentID, Agency, group_type, CommencementDate, ExpiryDate,
        DiscountAmount, AgreementStatus, NPrintFlag,
        AgentName, TerminationDate, create_operator, CreateDate,
        update_operator, UpdateDate, CreateMonthyear
    )
    SELECT 
        a.AgentID,
        a.Agency,
        a.group_type,
        a.CommencementDate,
        a.ExpiryDate,
        a.DiscountAmount,
        a.AgreementStatus,
        a.NPrintFlag,
        c.AgentName,
        c.TerminationDate,
        c.create_operator,
        c.CreateDate,
        c.update_operator,
        c.UpdateDate,
        c.CreateMonthyear
    FROM AgreementStatus a
    LEFT JOIN CreateUpdate c
        ON a.AgentID = c.AgentID;
END;
GO
```

# [dbo].[sp_LoadFactClaimData]

```sql
/****** Object:  StoredProcedure [dbo].[sp_LoadFactClaimData]    Script Date: 29/06/2026 1:38:34 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE   PROCEDURE [dbo].[sp_LoadFactClaimData]
AS
BEGIN
    SET NOCOUNT ON;

    --TRUNCATE TABLE fact_claim_data;
	TRUNCATE TABLE Claim_Episode_Staging;

    WITH BaseClaimData AS (
        SELECT
            cl.claim_id,
            cl.claim_line_id,
            cl.person_id,
            cl.service_date,
            cl.status_date,
            cl.episode_id,
            CASE
                WHEN ch.service_type IS NULL AND cg.service_type NOT IN ('MEDICAL','AMBULANCE') THEN 'General'
                WHEN ch.service_type IS NULL AND cg.service_type = 'AMBULANCE' THEN 'Ambulance'
                WHEN ch.service_type IS NULL AND cg.service_type = 'MEDICAL' THEN 'Medical'
                ELSE 'Hospital'
            END AS claim_type,
            cls.description AS line_status
        FROM BRONZE.dbo.claim_line cl
        INNER JOIN BRONZE.dbo.claim_line_status_type cls
            ON cl.claim_line_status_type = cls.claim_line_status_type
        INNER JOIN BRONZE.dbo.payee p
            ON cl.payee = p.payee
        LEFT JOIN BRONZE.dbo.claim_hospitalitem ch
            ON cl.claim_id = ch.claim_id AND cl.claim_line_id = ch.claim_line_id
        LEFT JOIN BRONZE.dbo.claim_generalitem cg
            ON cl.claim_id = cg.claim_id AND cl.claim_line_id = cg.claim_line_id
    ),
    MissingClaims AS (
        SELECT
            CAST(person_id AS VARCHAR(20)) + '-' + CONVERT(VARCHAR(10), service_date, 112) AS [Key],
            claim_id,
            claim_line_id,
            claim_type,
            person_id,
            service_date
        FROM BaseClaimData
        WHERE claim_type IN ('Hospital','Medical')
          AND status_date > '2021-01-01'
          AND episode_id IS NULL
          AND line_status = 'Paid'
    ),
    LatestEpisodes AS (
        SELECT
            b.episode_id,
            CAST(b.person_id AS VARCHAR(20)) + '-' + CONVERT(VARCHAR(10), b.service_date, 112) AS [Key],
            b.service_date
        FROM BaseClaimData b
        WHERE b.claim_type IN ('Hospital','Medical')
          AND b.status_date > '2021-01-01'
          AND b.episode_id IS NOT NULL
          AND b.line_status = 'Paid'
          AND b.episode_id = (
              SELECT MAX(e1.episode_id)
              FROM BRONZE.dbo.claim_line e1
              WHERE e1.claim_line_status_type = 'P'
                AND e1.claim_type IN ('H','M')
                AND e1.person_id = b.person_id
                AND e1.service_date = b.service_date
          )
    )
    --INSERT INTO fact_claim_data (
	INSERT INTO Claim_Episode_Staging (
        claim_id,
        claim_line_id,
        claim_type,
        person_id,
        episode_id,
        service_date,
        [Key]
    )
    SELECT DISTINCT
        m.claim_id,
        m.claim_line_id,
        m.claim_type,
        m.person_id,
        e.episode_id,
        m.service_date,
        m.[Key]
    FROM MissingClaims m
    INNER JOIN LatestEpisodes e
        ON m.[Key] = e.[Key];
END
GO
```

# [dbo].[usp_Create_ClaimDetailsAtService_Optimised]

```sql
/****** Object:  StoredProcedure [dbo].[usp_Create_ClaimDetailsAtService_Optimised]    Script Date: 29/06/2026 1:38:34 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE   PROCEDURE [dbo].[usp_Create_ClaimDetailsAtService_Optimised]
    @StartDate DATE = NULL,
    @EndDate DATE = NULL,
    @ClaimLineStatus VARCHAR(10) = NULL
AS
BEGIN
    SET NOCOUNT ON;
    
    -- Drop table if exists
    IF OBJECT_ID('SILVER.dbo.ClaimDetailsAtService_optimised', 'U') IS NOT NULL
        DROP TABLE SILVER.dbo.ClaimDetailsAtService_optimised;
    
    -- Create and populate table using SELECT INTO
    ;WITH 
    -- Step 1: 为每个 claim 找到有效的 cover (替代相关子查询)
    RankedCovers AS (
        SELECT 
            co.*,
            cl.claim_id,
            cl.claim_line_id,
            cl.person_id,
            cl.claim_type,
            cl.service_date,
            cl.claim_line_status_type,
            cl.status_date,
            cl.membership_id AS cl_membership_id,
            ROW_NUMBER() OVER (
                PARTITION BY cl.claim_line_id, cl.claim_id
                ORDER BY co.cover_version DESC
            ) AS cover_rn
        FROM BRONZE.dbo.claim_line cl
        INNER JOIN BRONZE.dbo.cover co
            ON cl.membership_id = co.membership_id
            AND co.cover_from_date <= cl.service_date
        WHERE 1=1
            AND (@StartDate IS NULL OR cl.service_date >= @StartDate)
            AND (@EndDate IS NULL OR cl.service_date < @EndDate)
            AND (@ClaimLineStatus IS NULL OR cl.claim_line_status_type = @ClaimLineStatus)
    ),
    
    -- Step 2: 为每个 claim 找到有效的 membership_group (替代相关子查询)
    RankedMembershipGroups AS (
        SELECT 
            mg.*,
            cl.claim_line_id,
            cl.claim_id,
            ROW_NUMBER() OVER (
                PARTITION BY cl.claim_line_id, cl.claim_id
                ORDER BY mg.membership_site_version DESC
            ) AS mg_rn
        FROM BRONZE.dbo.claim_line cl
        INNER JOIN BRONZE.dbo.membership_group mg
            ON cl.membership_id = mg.membership_id
            AND mg.commencement_date <= cl.service_date
            AND mg.group_id < 20000
        INNER JOIN BRONZE.dbo.grouping gr
            ON mg.group_id = gr.group_id
            AND gr.group_type = 'C'
        WHERE 1=1
            AND (@StartDate IS NULL OR cl.service_date >= @StartDate)
            AND (@EndDate IS NULL OR cl.service_date < @EndDate)
    ),
    
    -- Step 3: 预计算 Hospital 产品 (减少重复扫描)
    HospitalProducts AS (
        SELECT 
            cp.membership_id,
            cp.cover_version,
            p.product_code,
            p.description
        FROM BRONZE.dbo.cover_product cp
        JOIN BRONZE.dbo.product p 
            ON cp.product_id = p.product_id
        WHERE p.product_type = 'H'
    ),
    
    -- Step 4: 预计算 Ancillary 产品
    AncillaryProducts AS (
        SELECT 
            cp.membership_id,
            cp.cover_version,
            p.product_code,
            p.description
        FROM BRONZE.dbo.cover_product cp
        JOIN BRONZE.dbo.product p 
            ON cp.product_id = p.product_id
        WHERE p.product_type = 'A'
    ),
    
    -- Step 5: 预计算 Ambulance 产品
    AmbulanceProducts AS (
        SELECT 
            cp.membership_id,
            cp.cover_version,
            p.product_code
        FROM BRONZE.dbo.cover_product cp
        JOIN BRONZE.dbo.product p 
            ON cp.product_id = p.product_id
        WHERE p.product_type = 'B'
    ),
    
    -- Step 6: 预计算 Plan Detail (用于 Product_Description)
    PlanDetails AS (
        SELECT 
            cp.membership_id,
            cp.cover_version,
            pd.description,
            ROW_NUMBER() OVER (
                PARTITION BY cp.membership_id, cp.cover_version
                ORDER BY cp.product_id
            ) AS pd_rn
        FROM BRONZE.dbo.cover_product cp
        INNER JOIN BRONZE.dbo.plan_detail pd 
            ON cp.plan_id = pd.plan_id
    )
    
    -- Main Query with SELECT INTO
    SELECT  
        rc.claim_id, 
        rc.claim_line_id, 
        rc.cl_membership_id AS membership_id, 
        rc.person_id, 
        rc.claim_type, 
        rc.service_date, 
        rc.claim_line_status_type, 
        rc.status_date, 
        rc.cover_type, 
        rc.cover_state, 
        
        -- 内联年龄计算
        CASE 
            WHEN MONTH(rc.service_date) > MONTH(p.date_of_birth) 
                THEN DATEDIFF(YEAR, p.date_of_birth, rc.service_date)
            WHEN MONTH(rc.service_date) = MONTH(p.date_of_birth) 
                AND DAY(rc.service_date) >= DAY(p.date_of_birth)
                THEN DATEDIFF(YEAR, p.date_of_birth, rc.service_date)
            ELSE DATEDIFF(YEAR, p.date_of_birth, rc.service_date) - 1
        END AS age_at_claim,
        
        rmg.group_id AS mem_branch_at_claim,
        
        -- 产品代码组合
        CASE 
            WHEN COALESCE(hp.product_code, '') = COALESCE(ap.product_code, '') 
                THEN COALESCE(hp.product_code, '') + COALESCE(amb.product_code, '')
            ELSE COALESCE(hp.product_code, '') + COALESCE(ap.product_code, '') + COALESCE(amb.product_code, '')
        END AS mem_cover_at_claim,
        
        -- 产品描述
        COALESCE(
            (CASE  
                WHEN rc.description LIKE '%Platinum Extras%' THEN 'Platinum Extras'
                WHEN rc.description LIKE '%Gold Extras%' THEN 'Gold Extras'
                WHEN rc.description LIKE '%Value Extras%' THEN 'Value Extras'
                WHEN rc.description LIKE '%Premium Ambulance%' THEN 'Premium Ambulance' 
                WHEN rc.description LIKE '%Basic Ambulance%' THEN 'Basic Ambulance'
                WHEN rc.description LIKE '%Ambulance%' 
                    AND (rc.description NOT LIKE '%Basic Ambulance%' OR rc.description NOT LIKE '%Premium Ambulance%') 
                    THEN 'Ambulance'
                ELSE pd.description
            END),
            REPLACE(
                REPLACE(
                    REPLACE(
                        REPLACE(
                            REPLACE(
                                REPLACE(rc.description, 'SP Family ', ''),
                            'Couple ', ''),
                        'Single ', ''),
                    'Sole Parent Family ', ''),
                'Sole Parent ', ''),
            'Family ', '')
        ) AS Product_Description_at_claim,
        
        hp.product_code AS Hospital_Product_Code,
        ap.product_code AS Ancillary_Product_Code,
        hp.description AS Hospital_Product,
        ap.description AS Extras_Product
    
    INTO SILVER.dbo.ClaimDetailsAtService_optimised  -- SELECT INTO to create table
    
    FROM RankedCovers rc
    
    -- Join person for age calculation
    INNER JOIN BRONZE.dbo.person p 
        ON rc.person_id = p.person_id
    
    -- Join ranked membership groups
    INNER JOIN RankedMembershipGroups rmg
        ON rc.claim_id = rmg.claim_id
        AND rc.claim_line_id = rmg.claim_line_id
        AND rmg.mg_rn = 1
    
    -- Join grouping for final filter
    INNER JOIN BRONZE.dbo.grouping gr
        ON rmg.group_id = gr.group_id
    
    -- Join precomputed products
    LEFT JOIN HospitalProducts hp
        ON rc.membership_id = hp.membership_id
        AND rc.cover_version = hp.cover_version
    
    LEFT JOIN AncillaryProducts ap
        ON rc.membership_id = ap.membership_id
        AND rc.cover_version = ap.cover_version
    
    LEFT JOIN AmbulanceProducts amb
        ON rc.membership_id = amb.membership_id
        AND rc.cover_version = amb.cover_version
    
    -- Join plan details
    LEFT JOIN PlanDetails pd
        ON rc.membership_id = pd.membership_id
        AND rc.cover_version = pd.cover_version
        AND pd.pd_rn = 1
    
    WHERE 
        rc.cover_rn = 1  -- 只取最新的 cover version
        AND rmg.group_id < 20000 
        AND gr.group_type = 'C';
    
    -- Return statistics
    SELECT 
        'SILVER.dbo.ClaimDetailsAtService_optimised' AS TableName,
        COUNT(*) AS RowsInserted,
        MIN(service_date) AS MinServiceDate,
        MAX(service_date) AS MaxServiceDate
    FROM SILVER.dbo.ClaimDetailsAtService_optimised;
    
END
GO
```

# [dbo].[usp_Deceased_Active_Membership]

```sql
/****** Object:  StoredProcedure [dbo].[usp_Deceased_Active_Membership]    Script Date: 29/06/2026 1:38:34 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE   PROCEDURE [dbo].[usp_Deceased_Active_Membership]
AS
BEGIN
    SET NOCOUNT ON;

    -- Drop table if exists
    IF OBJECT_ID('dbo.Deceased_Active_Membership', 'U') IS NOT NULL
        DROP TABLE dbo.Deceased_Active_Membership;

    -- Create and populate table using SELECT INTO
    WITH Membership_Base AS (
        SELECT
            m.membership_id AS [Membership ID],
            m.memship_status AS [Membership Status],
            ISNULL(s.description, 'Other') AS [Security Level]
        FROM BRONZE.dbo.memship AS m
        LEFT JOIN BRONZE.dbo.security_level AS s
            ON m.security_level = s.security_level
        WHERE m.memship_status = 'A'
    ),

    Membership_PersonMembership AS (
        SELECT
            pm.person_id AS [Person ID],
            pm.membership_id AS [Membership ID],
            pm.relationship AS [Relationship],
            pm.status_flag AS [Status Flag],
            pm.termination_code AS [Termination Reason],
            CAST(pm.termination_date AS date) AS [Termination Date],
            CASE WHEN pm.termination_code IS NULL THEN 'Incomplete' ELSE 'Complete' END AS [Eligibility]
        FROM BRONZE.dbo.person_membership AS pm
    ),

    Membership_Person AS (
        SELECT
            p.person_id AS [Person ID],
            p.surname AS [Surname],
            p.first_name AS [First Name],
            p.deceased_flag AS [Deceased Flag],
            CAST(p.update_datetime AS date) AS [Date of Status Update],
            DATEADD(DAY, 7 - DATEPART(WEEKDAY, p.update_datetime), CAST(p.update_datetime AS date)) AS [Weekend]
        FROM BRONZE.dbo.person AS p
    ),

    Membership_Group AS (
        SELECT
            g.description AS [Group Description],
            g.membership_id AS [Membership ID],
            CAST(g.commencement_date AS date) AS [Group Commencement Date]
        FROM BRONZE.dbo.MemberGroup AS g
    )

    SELECT
        mb.[Membership ID],
        mb.[Membership Status],
        mb.[Security Level],
        pm.[Person ID],
        pm.[Relationship],
        pm.[Status Flag],
        pm.[Termination Reason],
        pm.[Termination Date],
        pm.[Eligibility],
        p.[Surname],
        p.[First Name],
        p.[Deceased Flag],
        p.[Date of Status Update],
        p.[Weekend],
        g.[Group Description],
        g.[Group Commencement Date],
        CASE
            WHEN mb.[Security Level] = 'No Access Restrictions' THEN p.[First Name]
            ELSE mb.[Security Level]
        END AS [First Names],
        CASE
            WHEN mb.[Security Level] = 'No Access Restrictions' THEN p.[Surname]
            ELSE mb.[Security Level]
        END AS [Surnames]
    INTO dbo.Deceased_Active_Membership
    FROM Membership_Base AS mb
    INNER JOIN Membership_PersonMembership AS pm
        ON mb.[Membership ID] = pm.[Membership ID]
    INNER JOIN Membership_Person AS p
        ON pm.[Person ID] = p.[Person ID]
    INNER JOIN Membership_Group AS g
        ON mb.[Membership ID] = g.[Membership ID];

END
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[usp_generate_episode_condition_group]    Script Date: 29/06/2026 1:38:34 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
-- ============================================
-- Stored Procedure: usp_generate_episode_condition_group
-- Description: 从 episode_classification 生成 episode_condition_group
-- 从 Qlik Sense 转换为 T-SQL
-- ============================================

CREATE   PROCEDURE [dbo].[usp_generate_episode_condition_group]
AS
BEGIN
    SET NOCOUNT ON;
    
    -- 删除旧表（如果存在）
    IF OBJECT_ID('[SILVER].[dbo].[Episode_Condition_Group]', 'U') IS NOT NULL
        DROP TABLE [SILVER].[dbo].[Episode_Condition_Group];
    
    -- 使用 CTE 生成所有映射和条件分组逻辑
    WITH BaseTable AS (
        SELECT 
            EpisodeClaimID as claim_id,
            EpisodeClaimLineID as claim_line_id,
            EpisodeID,
            ChapterItemNumber,
            membership_id,
            admission_date as EpsiodeAdmissionDate,
            discharge_date as EpisodeDischargeDate,
            drg_number,
            drg_version,
            episode_type as Episode_Type,
            person_id,
            ICD_Type as [ICD Type],
            block,
            ACHI_Code,
            long_description,
            short_description,
            Chapter,
            EpisodeProcedureItemNumber,
            EpisodeDiagnosisItemNumber,
            Procedure_Block,
            code_id_clean,
            -- 派生字段
            UPPER(SUBSTRING(EpisodeDiagnosisItemNumber, 1, CHARINDEX('.', EpisodeDiagnosisItemNumber + '.') - 1)) AS ICD10Prefix_Level1,
            UPPER(LEFT(EpisodeDiagnosisItemNumber, 1)) AS ICD10Prefix,
            TRY_CAST(SUBSTRING(EpisodeDiagnosisItemNumber, 2, 2) AS INT) AS ICD10_CodeNum
        FROM [SILVER].[dbo].[Episode_Classification]
    ),
    
    -- A 组映射：肠道感染等
    A_ConditionGroup_Map AS (
        SELECT * FROM (VALUES
            (0, 9, 'Intestinal Infections'),
            (15, 19, 'Infectious services'),
            (20, 28, 'Bacterial Infection'),
            (30, 49, 'Bacterial Infection'),
            (50, 64, 'Viral Infections'),
            (65, 69, 'Bacterial Infection'),
            (70, 74, 'Bacterial Infection'),
            (75, 79, 'Viral Infections'),
            (80, 89, 'Viral Infections'),
            (90, 99, 'Viral Infections')
        ) AS t(RangeStart, RangeEnd, ConditionGroup)
    ),
    
    -- B 组映射：病毒感染等
    B_ConditionGroup_Map AS (
        SELECT * FROM (VALUES
            (0, 9, 'Viral Infections'),
            (15, 19, 'Viral Infections'),
            (20, 24, 'Immunological Malignancy'),
            (25, 34, 'Viral Infections'),
            (35, 49, 'Fungal Infections'),
            (50, 64, 'Infectious services'),
            (65, 83, 'Infectious services'),
            (85, 89, 'Infectious services'),
            (90, 94, 'Infectious services'),
            (95, 97, 'Bacterial Infection'),
            (99, 99, 'Infectious services')
        ) AS t(RangeStart, RangeEnd, ConditionGroup)
    ),
    
    -- C 组映射：恶性肿瘤
    C_ConditionGroup_Map AS (
        SELECT * FROM (VALUES
            (0, 14, 'Primary malignant neoplasm'),
            (15, 26, 'Primary malignant neoplasm'),
            (30, 39, 'Primary malignant neoplasm'),
            (40, 41, 'Primary malignant neoplasm'),
            (43, 44, 'Primary malignant neoplasm'),
            (45, 49, 'Primary malignant neoplasm'),
            (50, 50, 'Primary malignant neoplasm'),
            (51, 58, 'Primary malignant neoplasm'),
            (60, 63, 'Primary malignant neoplasm'),
            (64, 68, 'Services for Neoplasms'),
            (69, 72, 'Primary malignant neoplasm'),
            (73, 75, 'Primary malignant neoplasm'),
            (76, 80, 'Secondary malignant neoplasm'),
            (81, 90, 'Haematological Malignancy'),
            (91, 91, 'Immunological Malignancy'),
            (92, 95, 'Haematological Malignancy'),
            (96, 96, 'Immunological Malignancy')
        ) AS t(RangeStart, RangeEnd, ConditionGroup)
    ),
    
    -- D 组映射：原位肿瘤、血液疾病
    D_ConditionGroup_Map AS (
        SELECT * FROM (VALUES
            (0, 9, 'In situ neoplasms'),
            (10, 36, 'Benign neoplasms'),
            (37, 48, 'Unspecified neoplasm'),
            (50, 53, 'Nutritional Disorders (Endocrine)'),
            (55, 59, 'Blood cell disorders'),
            (60, 64, 'Blood cell disorders'),
            (65, 69, 'Bleeding disorders'),
            (70, 77, 'Blood cell disorders'),
            (80, 89, 'Immune disorders')
        ) AS t(RangeStart, RangeEnd, ConditionGroup)
    ),
    
    -- E 组映射：内分泌疾病
    E_ConditionGroup_Map AS (
        SELECT * FROM (VALUES
            (0, 7, 'Thyroid Disease'),
            (10, 14, 'Other metabolic disorders'),
            (15, 16, 'Fluid balance disorders'),
            (20, 35, 'Other endocrine disorders'),
            (40, 46, 'Nutritional Disorders (Endocrine)'),
            (50, 64, 'Nutritional Disorders (Endocrine)'),
            (65, 68, 'Other metabolic disorders'),
            (70, 90, 'Other metabolic disorders')
        ) AS t(RangeStart, RangeEnd, ConditionGroup)
    ),
    
    -- F 组映射：精神疾病
    F_ConditionGroup_Map AS (
        SELECT * FROM (VALUES
            (0, 9, 'Organic brain syndromes'),
            (10, 19, 'Behavioural disorders due to substance use'),
            (20, 29, 'Psychotic Disorders'),
            (30, 39, 'Mood Disorders'),
            (40, 48, 'Anxiety Disorders'),
            (50, 59, 'Behavioural syndromes with physiological'),
            (60, 69, 'Personality and Behavioural Disorders'),
            (70, 79, 'Developmental Disorders'),
            (80, 89, 'Childhood behavioural disorders'),
            (90, 98, 'Signs and Symptom of behavioural disorders'),
            (99, 99, 'Behavioural disorders due to substance use')
        ) AS t(RangeStart, RangeEnd, ConditionGroup)
    ),
    
    -- G 组映射：神经系统疾病
    G_ConditionGroup_Map AS (
        SELECT * FROM (VALUES
            (0, 9, 'Encephalal disease'),
            (10, 14, 'Other Neurological disease'),
            (20, 26, 'Other Neurological disease'),
            (30, 32, 'Central nerve Disease'),
            (35, 37, 'Central nerve Disease'),
            (40, 47, 'Central nerve Disease (paraplegias)'),
            (50, 59, 'Peripheral nerve disease'),
            (60, 64, 'Peripheral nerve disease'),
            (70, 73, 'Muscle Disorders'),
            (80, 83, 'Central nerve Disease (paraplegias)'),
            (90, 99, 'Other Neurological disease')
        ) AS t(RangeStart, RangeEnd, ConditionGroup)
    ),
    
    -- H 组映射：眼耳疾病
    H_ConditionGroup_Map AS (
        SELECT * FROM (VALUES
            (0, 6, 'Eyelid Disorders'),
            (10, 13, 'Conjunctiva Disease'),
            (15, 22, 'Scleral disorders'),
            (25, 28, 'Lens Disorders'),
            (30, 36, 'Retinal Disorders'),
            (43, 45, 'Vitreous Disorders'),
            (60, 62, 'External Ear Disorders'),
            (65, 75, 'Middle Ear Disorders'),
            (80, 83, 'Other Neurological disease'),
            (90, 95, 'Tympanic Disorders')
        ) AS t(RangeStart, RangeEnd, ConditionGroup)
    ),
    
    -- I 组映射：心血管疾病
    I_ConditionGroup_Map AS (
        SELECT * FROM (VALUES
            (0, 2, 'Infectious services'),
            (5, 9, 'Endocardial Disease'),
            (10, 15, 'Arterial Vessel Disease'),
            (20, 24, 'Myocardial Disease'),
            (25, 25, 'Arterial Vessel Disease'),
            (26, 28, 'Venous Vessel Disease (thromboembolism)'),
            (30, 32, 'Other Cardiovascular Conditions'),
            (33, 33, 'Endocardial Disease'),
            (34, 34, 'Endocardial Disease'),
            (35, 35, 'Endocardial Disease'),
            (36, 41, 'Other Cardiovascular Conditions'),
            (42, 42, 'Myocardial Disease'),
            (43, 43, 'Signs and Symptoms (Cardiovascular)'),
            (44, 44, 'Conduction Disorders'),
            (45, 45, 'Conduction Disorders'),
            (46, 47, 'Arterial Vessel Disease'),
            (48, 48, 'Conduction Disorders'),
            (49, 49, 'Signs and Symptoms (Cardiovascular)'),
            (50, 50, 'Myocardial Disease'),
            (51, 52, 'Other Cardiovascular Conditions'),
            (60, 64, 'Other Cardiovascular Conditions'),
            (65, 65, 'Other Cardiovascular Conditions'),
            (66, 69, 'Other Cardiovascular Conditions'),
            (70, 79, 'Arterial Vessel Disease'),
            (80, 89, 'Venous Vessel Disease (thromboembolism)'),
            (95, 99, 'Other Cardiovascular Conditions')
        ) AS t(RangeStart, RangeEnd, ConditionGroup)
    ),
    
    -- J 组映射：呼吸系统疾病
    J_ConditionGroup_Map AS (
        SELECT * FROM (VALUES
            (0, 6, 'Upper Airway Disease'),
            (9, 18, 'Lower Airway Disease'),
            (20, 22, 'Lower Airway Disease'),
            (30, 39, 'Upper Airway Disease'),
            (40, 47, 'Lower Airway Disease'),
            (60, 70, 'Other Respiratory Disorders'),
            (80, 84, 'Interstitial disease'),
            (85, 86, 'Infectious services'),
            (90, 94, 'Pleural Disease'),
            (95, 99, 'Other Respiratory Disorders')
        ) AS t(RangeStart, RangeEnd, ConditionGroup)
    ),
    
    -- K 组映射：消化系统疾病
    K_ConditionGroup_Map AS (
        SELECT * FROM (VALUES
            (0, 3, 'Tooth Disorders'),
            (4, 4, 'Mouth Disorders'),
            (5, 5, 'Tooth Disorders'),
            (6, 6, 'Tooth Disorders'),
            (7, 14, 'Mouth Disorders'),
            (20, 31, 'Oesophageal disorders'),
            (35, 38, 'Digestive disorders'),
            (40, 46, 'Digestive disorders'),
            (50, 59, 'Intestinal Disorders'),
            (60, 68, 'Anal Disorders'),
            (70, 77, 'Liver Disease'),
            (80, 87, 'Biliary disorders'),
            (90, 93, 'Other Digestive Conditions')
        ) AS t(RangeStart, RangeEnd, ConditionGroup)
    ),
    
    -- L 组映射：皮肤疾病
    L_ConditionGroup_Map AS (
        SELECT * FROM (VALUES
            (0, 8, 'Skin disorders (Integumentary)'),
            (10, 14, 'Skin disorders (Integumentary)'),
            (20, 30, 'Skin disorders (Integumentary)'),
            (40, 45, 'Skin disorders (Integumentary)'),
            (50, 54, 'Skin disorders (Integumentary)'),
            (55, 59, 'Skin Services'),
            (60, 75, 'Skin disorders (Integumentary)'),
            (80, 99, 'Skin disorders (Integumentary)')
        ) AS t(RangeStart, RangeEnd, ConditionGroup)
    ),
    
    -- M 组映射：肌肉骨骼疾病
    M_ConditionGroup_Map AS (
        SELECT * FROM (VALUES
            (0, 13, 'Signs and Symptoms (Musculoskeletal)'),
            (14, 15, 'Joint Disorders'),
            (16, 19, 'Joint Disorders'),
            (20, 21, 'Signs and Symptoms (Musculoskeletal)'),
            (22, 25, 'Joint Disorders'),
            (30, 36, 'Systemic Connective Tissue Disease'),
            (40, 43, 'Musculoskeletal disorders'),
            (45, 49, 'Musculoskeletal disorders'),
            (50, 50, 'Signs and Symptoms (Musculoskeletal)'),
            (51, 51, 'Joint Disorders'),
            (52, 53, 'Musculoskeletal disorders'),
            (54, 54, 'Signs and Symptoms (Musculoskeletal)'),
            (60, 63, 'Muscle Disorders'),
            (65, 68, 'Tendon Disorders'),
            (70, 74, 'Soft Tissue Disorders'),
            (75, 75, 'Tendon Disorders'),
            (76, 79, 'Soft Tissue Disorders'),
            (80, 94, 'Bone Disorders'),
            (95, 95, 'Signs and Symptoms (Musculoskeletal)'),
            (96, 99, 'Congenital Disorders (Musculoskeletal)')
        ) AS t(RangeStart, RangeEnd, ConditionGroup)
    ),
    
    -- N 组映射：泌尿生殖系统疾病
    N_ConditionGroup_Map AS (
        SELECT * FROM (VALUES
            (0, 8, 'Glomerular disorders'),
            (10, 16, 'Kidney disorders'),
            (17, 19, 'Kidney disorders'),
            (20, 23, 'Ureteric disorders'),
            (25, 29, 'Kidney disorders'),
            (30, 39, 'Bladder disorders'),
            (40, 51, 'Prostate disorders'),
            (60, 64, 'Breast Disorders'),
            (70, 77, 'Female genital Disorders'),
            (80, 98, 'Female genital Disorders'),
            (99, 99, 'Genitourinary services')
        ) AS t(RangeStart, RangeEnd, ConditionGroup)
    ),
    
    -- O 组映射：妊娠相关
    O_ConditionGroup_Map AS (
        SELECT * FROM (VALUES
            (0, 8, 'Conception'),
            (9, 99, 'Labour and delivery')
        ) AS t(RangeStart, RangeEnd, ConditionGroup)
    ),
    
    -- P 组映射：新生儿
    P_ConditionGroup_Map AS (
        SELECT * FROM (VALUES
            (0, 96, 'Trauma to neonate')
        ) AS t(RangeStart, RangeEnd, ConditionGroup)
    ),
    
    -- R 组映射：症状和体征
    R_ConditionGroup_Map AS (
        SELECT * FROM (VALUES
            (0, 9, 'Signs and Symptoms (Cardiovascular)'),
            (10, 19, 'Signs and Symptoms of GIT'),
            (20, 23, 'Signs and Symptoms (Skin)'),
            (25, 29, 'Signs and symptoms Neurological'),
            (30, 39, 'SIgns and Symptom of Genitourinary'),
            (40, 46, 'Signs and Symptoms (Neurological / behaviour)'),
            (47, 49, 'Signs and Symptoms (Respiratory)'),
            (50, 69, 'Signs and Symptoms (Infectious Diseases)'),
            (70, 79, 'Signs and Symptoms (Endocrine)'),
            (80, 89, 'Signs and Symptoms (Other)'),
            (90, 94, 'Signs and Symptoms (Neurological)'),
            (95, 99, 'Signs and Symptoms (Other)')
        ) AS t(RangeStart, RangeEnd, ConditionGroup)
    ),
    
    -- S 组映射：创伤
    S_ConditionGroup_Map AS (
        SELECT * FROM (VALUES
            (0, 9, 'Traumatic (Neurological)'),
            (10, 19, 'Traumatic musculoskeletal'),
            (20, 29, 'Traumatic (Cardiovascular)'),
            (30, 39, 'Trauma of GIT'),
            (40, 49, 'Traumatic musculoskeletal'),
            (50, 50, 'Traumatic musculoskeletal'),
            (51, 51, 'Trauma to Skin'),
            (52, 59, 'Traumatic musculoskeletal'),
            (60, 61, 'Trauma to Skin'),
            (62, 69, 'Traumatic musculoskeletal'),
            (70, 79, 'Traumatic musculoskeletal'),
            (80, 80, 'Traumatic musculoskeletal'),
            (81, 81, 'Trauma to Skin'),
            (82, 89, 'Traumatic musculoskeletal'),
            (90, 90, 'Traumatic musculoskeletal'),
            (91, 91, 'Trauma to Skin'),
            (92, 99, 'Traumatic musculoskeletal')
        ) AS t(RangeStart, RangeEnd, ConditionGroup)
    ),
    
    -- Z 组映射：其他伤害和服务
    Z_ConditionGroup_Map AS (
        SELECT * FROM (VALUES
            (0, 7, 'Other Injuries'),
            (8, 8, 'Services for Neoplasms'),
            (9, 9, 'Services for Neoplasms'),
            (10, 11, 'Infectious services'),
            (12, 12, 'Services for Neoplasms'),
            (13, 13, 'Infectious services'),
            (14, 15, 'Other Injuries'),
            (17, 17, 'Other endocrine disorders'),
            (19, 19, 'Other endocrine disorders'),
            (20, 29, 'Signs and Symptom of behavioural disorders'),
            (30, 31, 'Genitourinary services'),
            (32, 37, 'Pregnancy complications'),
            (38, 38, 'Trauma to neonate'),
            (39, 39, 'Pregnancy complications'),
            (40, 40, 'Services for Neoplasms'),
            (41, 41, 'Intestinal Disorders'),
            (42, 42, 'Skin disorders (Integumentary)'),
            (43, 44, 'Intestinal Disorders'),
            (45, 45, 'Conduction Disorders'),
            (46, 48, 'Other'),
            (49, 49, 'Genitourinary services'),
            (50, 50, 'Other Injuries'),
            (51, 51, 'Services for Neoplasms'),
            (52, 54, 'Other Injuries'),
            (55, 65, 'Signs and Symptom of behavioural disorders'),
            (67, 67, 'Haematological Malignancy'),
            (70, 76, 'Other Injuries'),
            (80, 80, 'Services for Neoplasms'),
            (81, 81, 'Signs and Symptom of behavioural disorders'),
            (82, 82, 'Signs and Symptoms (Cardiovascular)'),
            (83, 83, 'Other Digestive Conditions'),
            (84, 84, 'Musculoskeletal disorders'),
            (85, 85, 'Services for Neoplasms'),
            (86, 86, 'Other Injuries'),
            (87, 87, 'Congenital Malformation'),
            (88, 88, 'Other Injuries'),
            (89, 89, 'Other Injuries'),
            (90, 94, 'Other Injuries'),
            (95, 95, 'Signs and Symptoms (Cardiovascular)'),
            (96, 99, 'Other Injuries')
        ) AS t(RangeStart, RangeEnd, ConditionGroup)
    ),
    
    -- Q 组映射：先天性畸形
    Q_ConditionGroup_Map AS (
        SELECT * FROM (VALUES
            (0, 37, 'Congenital Malformation'),
            (38, 38, 'Other Digestive Conditions'),
            (39, 52, 'Congenital Malformation'),
            (53, 53, 'Genitourinary services'),
            (78, 78, 'Musculoskeletal disorders'),
            (79, 99, 'Congenital Malformation')
        ) AS t(RangeStart, RangeEnd, ConditionGroup)
    ),
    
    -- T 组映射：其他伤害和中毒
    T_ConditionGroup_Map AS (
        SELECT * FROM (VALUES
            (0, 8, 'Other Injuries'),
            (9, 9, 'Traumatic musculoskeletal'),
            (10, 12, 'Other Injuries'),
            (13, 13, 'Skin disorders (Integumentary)'),
            (14, 14, 'Traumatic musculoskeletal'),
            (15, 15, 'Signs and Symptoms (Opthalmological)'),
            (16, 16, 'Tympanic Disorders'),
            (17, 17, 'Traumatic (Respiratory)'),
            (18, 18, 'Trauma of GIT'),
            (19, 19, 'Other genitourinary disorders'),
            (20, 34, 'Skin disorders (Integumentary)'),
            (36, 65, 'Poisoning'),
            (66, 77, 'Other Injuries'),
            (78, 78, 'Trauma of GIT'),
            (79, 79, 'Muscle Disorders'),
            (80, 80, 'Other Injuries'),
            (81, 81, 'Infectious services'),
            (82, 82, 'Traumatic (Cardiovascular)'),
            (83, 83, 'Other genitourinary disorders'),
            (84, 84, 'Signs and Symptoms (Musculoskeletal)'),
            (85, 99, 'Other Injuries')
        ) AS t(RangeStart, RangeEnd, ConditionGroup)
    ),
    
    -- 剩余 ICD10 分类映射 (U, V, W, X, Y)
    ICD10_Remaining_CategoryMap AS (
        SELECT * FROM (VALUES
            ('U', 'Other'),
            ('V', 'Other'),
            ('W', 'Other'),
            ('X', 'Other'),
            ('Y', 'Other')
        ) AS t(ICD10Prefix, ConditionGroup)
    ),
    
    -- MBS 条件组映射
    MBS_ConditionGroup_Map AS (
        SELECT * FROM (VALUES
            (1, 86, 'Other Neurological disease'),
            (110, 129, 'Other endocrine disorders'),
            (160, 256, 'Signs and Symptoms (Opthalmological}'),
            (300, 334, 'Tympanic Disorders'),
            (370, 422, 'Other Digestive Conditions'),
            (450, 490, 'Other Injuries'),
            (520, 572, 'Respiratory disorders'),
            (600, 777, 'Other Cardiovascular Conditions'),
            (800, 817, 'Haematological Malignancy'),
            (850, 1011, 'Other Digestive Conditions'),
            (1040, 1129, 'Genitourinary services'),
            (1160, 1203, 'Prostate disorders'),
            (1240, 1299, 'Female genital Disorders'),
            (1330, 1347, 'Pregnancy complications'),
            (1360, 1580, 'Musculoskeletal disorders'),
            (1600, 1718, 'Skin disorders (Integumentary)'),
            (1740, 1759, 'Genitourinary services'),
            (1786, 1799, 'Services for Neoplasms'),
            (1820, 1923, 'Other Injuries'),
            (1940, 2016, 'Other Injuries')
        ) AS t(RangeStart, RangeEnd, ConditionGroup)
    ),
    
    -- 将所有映射 JOIN 到基础表
    BaseWithMapping AS (
        SELECT 
            b.*,
            a.ConditionGroup AS A_ConditionGroup,
            b2.ConditionGroup AS B_ConditionGroup,
            c.ConditionGroup AS C_ConditionGroup,
            d.ConditionGroup AS D_ConditionGroup,
            e.ConditionGroup AS E_ConditionGroup,
            f.ConditionGroup AS F_ConditionGroup,
            g.ConditionGroup AS G_ConditionGroup,
            h.ConditionGroup AS H_ConditionGroup,
            i.ConditionGroup AS I_ConditionGroup,
            j.ConditionGroup AS J_ConditionGroup,
            k.ConditionGroup AS K_ConditionGroup,
            l.ConditionGroup AS L_ConditionGroup,
            m.ConditionGroup AS M_ConditionGroup,
            n.ConditionGroup AS N_ConditionGroup,
            o.ConditionGroup AS O_ConditionGroup,
            p.ConditionGroup AS P_ConditionGroup,
            r.ConditionGroup AS R_ConditionGroup,
            s.ConditionGroup AS S_ConditionGroup,
            z.ConditionGroup AS Z_ConditionGroup,
            q.ConditionGroup AS Q_ConditionGroup,
            t.ConditionGroup AS T_ConditionGroup,
            rem.ConditionGroup AS ICD10_Remaining_Condition_Group,
            mbs.ConditionGroup AS MBS_ConditionGroup
        FROM BaseTable b
        LEFT JOIN A_ConditionGroup_Map a 
            ON b.ICD10_CodeNum BETWEEN a.RangeStart AND a.RangeEnd
        LEFT JOIN B_ConditionGroup_Map b2 
            ON b.ICD10_CodeNum BETWEEN b2.RangeStart AND b2.RangeEnd
        LEFT JOIN C_ConditionGroup_Map c 
            ON b.ICD10_CodeNum BETWEEN c.RangeStart AND c.RangeEnd
        LEFT JOIN D_ConditionGroup_Map d 
            ON b.ICD10_CodeNum BETWEEN d.RangeStart AND d.RangeEnd
        LEFT JOIN E_ConditionGroup_Map e 
            ON b.ICD10_CodeNum BETWEEN e.RangeStart AND e.RangeEnd
        LEFT JOIN F_ConditionGroup_Map f 
            ON b.ICD10_CodeNum BETWEEN f.RangeStart AND f.RangeEnd
        LEFT JOIN G_ConditionGroup_Map g 
            ON b.ICD10_CodeNum BETWEEN g.RangeStart AND g.RangeEnd
        LEFT JOIN H_ConditionGroup_Map h 
            ON b.ICD10_CodeNum BETWEEN h.RangeStart AND h.RangeEnd
        LEFT JOIN I_ConditionGroup_Map i 
            ON b.ICD10_CodeNum BETWEEN i.RangeStart AND i.RangeEnd
        LEFT JOIN J_ConditionGroup_Map j 
            ON b.ICD10_CodeNum BETWEEN j.RangeStart AND j.RangeEnd
        LEFT JOIN K_ConditionGroup_Map k 
            ON b.ICD10_CodeNum BETWEEN k.RangeStart AND k.RangeEnd
        LEFT JOIN L_ConditionGroup_Map l 
            ON b.ICD10_CodeNum BETWEEN l.RangeStart AND l.RangeEnd
        LEFT JOIN M_ConditionGroup_Map m 
            ON b.ICD10_CodeNum BETWEEN m.RangeStart AND m.RangeEnd
        LEFT JOIN N_ConditionGroup_Map n 
            ON b.ICD10_CodeNum BETWEEN n.RangeStart AND n.RangeEnd
        LEFT JOIN O_ConditionGroup_Map o 
            ON b.ICD10_CodeNum BETWEEN o.RangeStart AND o.RangeEnd
        LEFT JOIN P_ConditionGroup_Map p 
            ON b.ICD10_CodeNum BETWEEN p.RangeStart AND p.RangeEnd
        LEFT JOIN R_ConditionGroup_Map r 
            ON b.ICD10_CodeNum BETWEEN r.RangeStart AND r.RangeEnd
        LEFT JOIN S_ConditionGroup_Map s 
            ON b.ICD10_CodeNum BETWEEN s.RangeStart AND s.RangeEnd
        LEFT JOIN Z_ConditionGroup_Map z 
            ON b.ICD10_CodeNum BETWEEN z.RangeStart AND z.RangeEnd
        LEFT JOIN Q_ConditionGroup_Map q 
            ON b.ICD10_CodeNum BETWEEN q.RangeStart AND q.RangeEnd
        LEFT JOIN T_ConditionGroup_Map t 
            ON b.ICD10_CodeNum BETWEEN t.RangeStart AND t.RangeEnd
        LEFT JOIN ICD10_Remaining_CategoryMap rem 
            ON b.ICD10Prefix = rem.ICD10Prefix
        LEFT JOIN MBS_ConditionGroup_Map mbs 
            ON b.Procedure_Block BETWEEN mbs.RangeStart AND mbs.RangeEnd
    ),
    
    -- 第一层：根据 ICD10 前缀选择对应的条件组
    ICD_Condition_Group_CTE AS (
        SELECT 
            *,
            CASE 
                WHEN ICD10Prefix = 'A' THEN A_ConditionGroup
                WHEN ICD10Prefix = 'B' THEN B_ConditionGroup
                WHEN ICD10Prefix = 'C' THEN C_ConditionGroup
                WHEN ICD10Prefix = 'D' THEN D_ConditionGroup
                WHEN ICD10Prefix = 'E' THEN E_ConditionGroup
                WHEN ICD10Prefix = 'F' THEN F_ConditionGroup
                WHEN ICD10Prefix = 'G' THEN G_ConditionGroup
                WHEN ICD10Prefix = 'H' THEN H_ConditionGroup
                WHEN ICD10Prefix = 'I' THEN I_ConditionGroup
                WHEN ICD10Prefix = 'J' THEN J_ConditionGroup
                WHEN ICD10Prefix = 'K' THEN K_ConditionGroup
                WHEN ICD10Prefix = 'L' THEN L_ConditionGroup
                WHEN ICD10Prefix = 'M' THEN M_ConditionGroup
                WHEN ICD10Prefix = 'N' THEN N_ConditionGroup
                WHEN ICD10Prefix = 'O' THEN O_ConditionGroup
                WHEN ICD10Prefix = 'P' THEN P_ConditionGroup
                WHEN ICD10Prefix = 'R' THEN R_ConditionGroup
                WHEN ICD10Prefix = 'S' THEN S_ConditionGroup
                WHEN ICD10Prefix = 'Z' THEN Z_ConditionGroup
                WHEN ICD10Prefix = 'T' THEN T_ConditionGroup
                WHEN ICD10Prefix = 'Q' THEN Q_ConditionGroup
                ELSE ICD10_Remaining_Condition_Group
            END AS ICD_Condition_Group
        FROM BaseWithMapping
    ),
    
    -- 第二层：优先使用 ICD 分类，否则使用 MBS 分类
    ConditionGroup_Level_1 AS (
        SELECT 
            *,
            CASE 
                WHEN LEN(LTRIM(RTRIM(ISNULL(ICD_Condition_Group, '')))) > 0 
                    THEN ICD_Condition_Group
                WHEN LEN(LTRIM(RTRIM(ISNULL(MBS_ConditionGroup, '')))) > 0 
                    THEN MBS_ConditionGroup
                ELSE 'Other'
            END AS ConditionGroup_Level_1_Value
        FROM ICD_Condition_Group_CTE
    ),
    
    -- 第三层：如果是 'Other'，则尝试使用 MBS 分类
    ConditionGroup_Level_2 AS (
        SELECT 
            *,
            CASE 
                WHEN ConditionGroup_Level_1_Value = 'Other' 
                    THEN ISNULL(MBS_ConditionGroup, ConditionGroup_Level_1_Value)
                ELSE ConditionGroup_Level_1_Value
            END AS ConditionGroup_Level_2_Value
        FROM ConditionGroup_Level_1
    ),
    
    -- 第四层：处理特殊案例（硬编码的 ICD10 代码）
    ConditionGroup_Final AS (
        SELECT 
            *,
            CASE 
                WHEN EpisodeDiagnosisItemNumber = 'I25.5' 
                    THEN 'Myocardial Disease'
                WHEN EpisodeDiagnosisItemNumber = 'Z45.89' 
                    THEN 'Venous Vessel Disease (thromboembolism)'
                WHEN EpisodeDiagnosisItemNumber = 'U07.12' 
                    THEN 'Infectious services'
                ELSE ConditionGroup_Level_2_Value
            END AS [Condition Group]
        FROM ConditionGroup_Level_2
    )
    
    -- 创建新表并插入去重后的数据
    SELECT DISTINCT
        claim_id,
        claim_line_id,
        EpisodeID,
        ChapterItemNumber,
        membership_id,
        [Condition Group],
        person_id,
        block,
        ACHI_Code,
        long_description,
        short_description,
        Chapter,
        EpisodeProcedureItemNumber,
        EpisodeDiagnosisItemNumber,
        Procedure_Block
    INTO [SILVER].[dbo].[Episode_Condition_Group]
    FROM ConditionGroup_Final;
    
END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[usp_Load_Ancillary_Lookup]    Script Date: 29/06/2026 1:38:34 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[usp_Load_Ancillary_Lookup]
AS
BEGIN
    SET NOCOUNT ON;

    -- 清空目标表
    TRUNCATE TABLE dbo.Ancillary_Lookup;

    ;WITH ProductCounts AS (
        SELECT 
            cp.membership_id,
            cp.cover_version,
            COUNT(*) AS product_count
        FROM BRONZE.dbo.cover_product cp
        JOIN SILVER.dbo.[Product] p
            ON cp.[product_id] = p.[Product Id]
        WHERE p.[Product Type] = 'A'
        GROUP BY cp.membership_id, cp.cover_version
    ),
    AncillaryProducts AS (
        SELECT 
            cp.membership_id,
            cp.cover_version,
            p.[Product Code]
        FROM BRONZE.dbo.cover_product cp
        JOIN SILVER.dbo.[Product] p
            ON cp.[product_id] = p.[Product Id]
        JOIN ProductCounts pc
            ON cp.membership_id = pc.membership_id
           AND cp.cover_version = pc.cover_version
        WHERE p.[Product Type] = 'A'
          AND (
                (pc.product_count > 1 AND p.[Product Description] LIKE '%Ancillary%')
                OR (pc.product_count = 1)
              )
    )

    INSERT INTO dbo.Ancillary_Lookup (
        claim_id,
        membership_id,
        service_date,
        cover_type,
        cover_state,
        Ancillary_Product_Code
    )
    SELECT DISTINCT
        cl.claim_id,
        cl.membership_id,
        cl.service_date,
        co.cover_type,
        co.cover_state,
        ancillary.[Product Code] AS Ancillary_Product_Code
    FROM BRONZE.dbo.claim_line AS cl
    INNER JOIN BRONZE.dbo.cover AS co
        ON cl.membership_id = co.membership_id
    LEFT JOIN AncillaryProducts AS ancillary
        ON ancillary.membership_id = co.membership_id
       AND ancillary.cover_version = co.cover_version
    WHERE cl.service_date >= '2023-07-01'
      AND co.cover_version = (
            SELECT MAX(cover_version)
            FROM BRONZE.dbo.cover
            WHERE cover_from_date <= cl.service_date
              AND membership_id = co.membership_id
        );
END
GO
```

# [dbo].[usp_Load_Claim_Detail_Gen_And_Hosp]

```sql
/****** Object:  StoredProcedure [dbo].[usp_Load_Claim_Detail_Gen_And_Hosp]    Script Date: 29/06/2026 1:38:34 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE   PROCEDURE [dbo].[usp_Load_Claim_Detail_Gen_And_Hosp]
AS
BEGIN
    SET NOCOUNT ON;

    TRUNCATE TABLE SILVER.dbo.Claim_Detail_Gen_And_Hosp;

    INSERT INTO SILVER.dbo.Claim_Detail_Gen_And_Hosp (
        claim_id,
        claim_line_id,
        provider_number_id,
        membership_id,
        person_id,
        service_date,
        status_date,
        create_operator,
        create_datetime,
        payee_method,
        line_status,
        claim_type,
        service_type,
        item_number,
        fee,
        benefit,
        adjustment_flag,
        Num_services,
        update_operator
    )
    SELECT
        cl.claim_id,
        cl.claim_line_id,
        cl.provider_number_id,
        cl.membership_id,
        cl.person_id,
        cl.service_date,
        cl.status_date,
        cl.create_operator,
        cl.create_datetime,
        p.description                                               AS payee_method,
        cls.description                                             AS line_status,
        CASE
            WHEN ch.service_type IS NULL AND cg.service_type NOT IN ('MEDICAL', 'AMBULANCE') THEN 'General'
            WHEN ch.service_type IS NULL AND cg.service_type = 'AMBULANCE'                   THEN 'Ambulance'
            WHEN ch.service_type IS NULL AND cg.service_type = 'MEDICAL'                     THEN 'Medical'
            ELSE 'Hospital'
        END                                                         AS claim_type,
        COALESCE(ch.service_type,  cg.service_type)                AS service_type,
        COALESCE(ch.item_number,   cg.item_number)                 AS item_number,
        COALESCE(ch.fee,           cg.fee)                         AS fee,
        COALESCE(ch.benefit,       cg.benefit)                     AS benefit,
        CASE COALESCE(cg.adjustment_flag, ch.adjustment_flag)
            WHEN 'A' THEN 'New Line'
            WHEN 'O' THEN 'Original Line'
            WHEN 'R' THEN 'Reversal Line'
        END                                                         AS adjustment_flag,
        COALESCE(ch.num_services,  cg.num_services)                AS Num_services,
        cl.update_operator
    FROM        BRONZE.dbo.claim_line              AS cl
    INNER JOIN  BRONZE.dbo.claim_line_status_type  AS cls ON cls.claim_line_status_type = cl.claim_line_status_type
    INNER JOIN  BRONZE.dbo.payee                   AS p   ON p.payee                   = cl.payee
    LEFT JOIN   BRONZE.dbo.claim_hospitalitem      AS ch  ON ch.claim_id               = cl.claim_id
                                                         AND ch.claim_line_id           = cl.claim_line_id
    LEFT JOIN   BRONZE.dbo.claim_generalitem       AS cg  ON cg.claim_id               = cl.claim_id
                                                         AND cg.claim_line_id           = cl.claim_line_id;

END;
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[usp_Load_Claim_Line_Detail]    Script Date: 29/06/2026 1:38:34 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE   PROCEDURE [dbo].[usp_Load_Claim_Line_Detail]
AS
BEGIN
    SET NOCOUNT ON;
	SET DATEFIRST 7; --- added 20/04/2026
    TRUNCATE TABLE SILVER.dbo.Claim_Line_Detail;

    -- Rolling 5-year window starting 1 July
    DECLARE @StartDate DATE = DATEFROMPARTS(YEAR(DATEADD(YEAR, -6, GETDATE())), 7, 1);
	--DECLARE @StartDate DATE = '2026-04-27';
    WITH

    -- --------------------------------------------------------
    -- Latest contract (DENTAL/OPTICAL) per item
    -- --------------------------------------------------------
    ContractItemCTE AS (
        SELECT
            c.item_number,
            c.service_type,
            c.contract_item_version,
            c.schedule_fee
        FROM BRONZE.dbo.contract_item c
        WHERE c.service_type IN ('DENTAL', 'OPTICAL')
          AND c.status_flag = 'A'
          AND c.contract_item_version = (
              SELECT MAX(ci.contract_item_version)
              FROM BRONZE.dbo.contract_item ci
              WHERE ci.item_number   = c.item_number
                AND ci.service_type  IN ('DENTAL', 'OPTICAL')
                AND ci.status_flag   = 'A'
          )
    ),

    -- --------------------------------------------------------
    -- Agreement type per provider_number_id
    -- --------------------------------------------------------
    AgreementCTE AS (
        SELECT
            co.provider_number_id,
            CASE
                WHEN pg.provider_group_id   = '757' THEN 'Westfund Eyecare'
                WHEN pg.provider_group_id   = '1'   THEN 'Westfund Dental'
                WHEN pg.provider_group_type = 'C'   THEN 'POC'
                ELSE 'External'
            END AS agreement_id
        FROM (
            SELECT pn.provider_number_id, pn.agreement_id
            FROM BRONZE.dbo.contract_details        cd
            LEFT JOIN BRONZE.dbo.providernumber_contract pn
                ON cd.contract_info_id = pn.contract_info_id
            WHERE cd.contract_info_id = (
                SELECT MAX(c.contract_info_id)
                FROM BRONZE.dbo.contract_details        c
                LEFT JOIN BRONZE.dbo.providernumber_contract p
                    ON c.contract_info_id = p.contract_info_id
                WHERE pn.provider_number_id = p.provider_number_id
            )
        ) co
        LEFT JOIN BRONZE.dbo.provider_number ppn ON ppn.provider_number_id = co.provider_number_id
        LEFT JOIN BRONZE.dbo.provider_group  pg  ON pg.provider_group_id   = ppn.provider_group_id
    ),

    -- --------------------------------------------------------
    -- Presenting illness per item (Medical claims only)
    -- --------------------------------------------------------
    PresentingIllnessCTE AS (
        SELECT
            ei.item_number,
            ISNULL(eic.description, 'NA') AS presenting_illness
        FROM BRONZE.dbo.eligibility_illness_code_item ei
        LEFT JOIN BRONZE.dbo.eligibility_illness_code eic
            ON eic.eligibility_illness_code = ei.eligibility_illness_code
    ),

    -- --------------------------------------------------------
    -- Hospital admission/discharge dates (DISTINCT per claim line)
    -- --------------------------------------------------------
    HospitalItemCTE AS (
        SELECT DISTINCT
            claim_id,
            claim_line_id,
            CAST(admission_date AS DATE) AS admission_date,
            CAST(discharge_date AS DATE) AS discharge_date
        FROM BRONZE.dbo.claim_hospitalitem
    )

    INSERT INTO SILVER.dbo.Claim_Line_Detail (
        -- Identifiers
        Claim_ID, Claim_Line_ID, Unique_Claim_Key, Episode_ID,
        -- Dates
        Claim_Create_Date, Service_Date, Status_Date, Week_End_Date, Month_Year,
        Effective_Join_Date,
        -- Claim Attributes
        Claim_Received_Method, Claim_Type, Claim_Line_Status_Type,
        Adjustment_Flag, Update_Operator,
        -- Financials
        Fee, Benefit, Medicare_Benefit, Westfund_Benefit, Num_Services,
        Red_Lower_Limit, Red_Upper_Limit, Blue_Lower_Limit, Blue_Upper_Limit,
        Schedule_Fee, Contract_Version,
        -- Member
        Membership_ID, Person_ID, Member_Name, Previous_Fund, Gender,
        Current_Cover, Age, Age_Cohort, LOM, LOM_Cohort, Agent, Web_Login_Count,
        -- Cover at Service
        Cover_Type, State, Branch, Cover_Code, Cover_Description,
        Product_Group, Product_Classification,
        Hospital_Product, Extras_Product, Hospital_Code, Extras_Code,
        Product_Type, Hospital_Product_Group, Extras_Product_Group, Audience_Segment,
        -- Service / Item
        Provider_Number_ID, Service_Type, Service_Description,
        Item_Number, Service_Item_Key, Item_Description, Item_Group_Description,
        Fund_Classification, Disabled_HICAPS_Flag, Online_Claiming_Flag,
        Include_In_Topup_Job, Item_Number_Create_Date, Ceased_Date,
        Dental_Classification, Optical_Classification, POC_Item_Classification,
        Presenting_Illness,
        -- Provider
        Provider_ID, Provider_Type, Provider_Name, Provider_Group,
        Provider_Surname, Provider_First_Name, Other_Names, Title,
        Provider_Postcode, Agreement_ID,
        -- Sales / Channel
        Promotion, Sales_Channel,
        -- Hospital / Clinical
        Admission_Date, Discharge_Date, Drug_Store,
        Clinical_Category,
        -- Manual Claim
        Manual_Claim, Manual_Benefit_Reason
    )
    SELECT
        -- ------------------------------------------------
        -- Identifiers
        -- ------------------------------------------------
        cl.claim_id,
        cl.claim_line_id,
        CAST(cl.claim_id AS VARCHAR(9)) + '.' + CAST(cl.claim_line_id AS VARCHAR(9)),
        cl.episode_id,

        -- ------------------------------------------------
        -- Dates
        -- ------------------------------------------------
        CAST(cl.create_datetime  AS DATE),
        CAST(cdgh.service_date   AS DATE),
        CAST(cdgh.status_date    AS DATE),
        -- Week end = Sunday (week starts Monday, FirstWeekDay=0)
        DATEADD(DAY, 6 - (DATEPART(WEEKDAY, cdgh.status_date) + 5) % 7,
                CAST(cdgh.status_date AS DATE)),
        FORMAT(cdgh.status_date, 'MMM yyyy'),
        CAST(ms.effective_join_date AS DATE),

        -- ------------------------------------------------
        -- Claim Attributes
        -- ------------------------------------------------
        ISNULL(cbc.category, 'No Channel'),
        cdgh.claim_type,
        cl.claim_line_status_type,
        ISNULL(cdgh.adjustment_flag, 'No adjustment'),
        cdgh.update_operator,

        -- ------------------------------------------------
        -- Financials
        -- ------------------------------------------------
        cdgh.fee,
        cdgh.benefit,
        -- Medicare benefit: only where drug_store_name LIKE 'Medical%'
        CASE WHEN cgi.benefit > 0 AND cgi.fee <> 0
                  AND cgi.drug_store_name LIKE 'Medical%'
             THEN cgi.medicare_benefit
             ELSE NULL
        END,
        -- Westfund benefit
        --CASE WHEN cdgh.claim_type = 'Medical'
        --     THEN cdgh.benefit - ISNULL(
        --             CASE WHEN cgi.benefit > 0 AND cgi.fee <> 0
        --                       AND cgi.drug_store_name LIKE 'Medical%'
        --                  THEN cgi.medicare_benefit ELSE NULL END, 0)
        --     ELSE cdgh.benefit
        --END,
		cdgh.benefit,
        cdgh.Num_services,
        iaf.critical_red_lower_limit,
        iaf.critical_red_upper_limit,
        iaf.warning_blue_lower_limit,
        iaf.warning_blue_upper_limit,
        ci_cte.schedule_fee,
        ci_cte.contract_item_version,

        -- ------------------------------------------------
        -- Member
        -- ------------------------------------------------
        cdgh.membership_id,
        cdgh.person_id,
        RTRIM(p.first_name) + ' ' + RTRIM(p.surname),
        ISNULL(pf.description, 'No Previous Fund'),
        p.gender,
        ISNULL(mc.Product_Description, 'Unknown'),
        cas.age_at_claim,
        -- Age cohort
        CASE
            WHEN cas.age_at_claim BETWEEN -10 AND 24 THEN '0-24'
            WHEN cas.age_at_claim BETWEEN  25 AND 34 THEN '25-34'
            WHEN cas.age_at_claim BETWEEN  35 AND 44 THEN '35-44'
            WHEN cas.age_at_claim BETWEEN  45 AND 54 THEN '45-54'
            WHEN cas.age_at_claim BETWEEN  55 AND 64 THEN '55-64'
            WHEN cas.age_at_claim BETWEEN  65 AND 74 THEN '65-74'
            WHEN cas.age_at_claim BETWEEN  75 AND 84 THEN '75-84'
            WHEN cas.age_at_claim >= 85              THEN '85+'
        END,
        -- LOM (Length of Membership in years, Qlik Age() function equivalent)
        lom_calc.lom,
        -- LOM cohort
        CASE
            WHEN lom_calc.lom BETWEEN -10 AND  1 THEN '<1'
            WHEN lom_calc.lom BETWEEN   2 AND  3 THEN '2-3'
            WHEN lom_calc.lom BETWEEN   4 AND  5 THEN '4-5'
            WHEN lom_calc.lom BETWEEN   6 AND  7 THEN '6-7'
            WHEN lom_calc.lom BETWEEN   8 AND 10 THEN '8-10'
            WHEN lom_calc.lom BETWEEN  11 AND 15 THEN '11-15'
            WHEN lom_calc.lom BETWEEN  16 AND 25 THEN '16-25'
            WHEN lom_calc.lom >= 26              THEN '26+'
        END,
        ISNULL(ma.description, 'No Agent'),
        CASE WHEN ws.oper_name IS NOT NULL THEN CAST(1 AS TINYINT) ELSE NULL END,

        -- ------------------------------------------------
        -- Cover at Service Time
        -- ------------------------------------------------
        ISNULL(ct.description, 'Missing'),
        cas.cover_state,
        ISNULL(grp.description, 'MISSING'),
        cas.mem_cover_at_claim,
        cas.Product_Description_at_claim,
        -- Product Group
        CASE
            WHEN cas.mem_cover_at_claim IN ('OSC','NZO','FCO') THEN 'Overseas'
            WHEN (cas.Product_Description_at_claim LIKE '%&%'
                  OR cas.Product_Description_at_claim LIKE '%with%')
                 AND cas.mem_cover_at_claim NOT IN ('CH','ECH','VE') THEN 'Hospital & Extras'
            WHEN (cas.Product_Description_at_claim LIKE '%&%'
                  OR cas.Product_Description_at_claim LIKE '%with%')
                 AND cas.mem_cover_at_claim IN ('CH','ECH')          THEN 'Hospital Only'
            WHEN (cas.Product_Description_at_claim LIKE '%&%'
                  OR cas.Product_Description_at_claim LIKE '%with%')
                 AND cas.mem_cover_at_claim = 'VE'                   THEN 'Extras Only'
            WHEN cas.Product_Description_at_claim LIKE '%with%'      THEN 'Hospital & Extras'
            WHEN cas.Product_Description_at_claim LIKE '%Extras'     THEN 'Extras Only'
            WHEN cas.Product_Description_at_claim LIKE '%Extras Only'
                  OR cas.Product_Description_at_claim = 'Value Extras Ancillary' THEN 'Extras Only'
            WHEN cas.Product_Description_at_claim LIKE '%Ambulance%' THEN 'Ambulance'
            ELSE 'To Be Classified'
        END,
        -- Product Classification
        CASE
            WHEN cas.mem_cover_at_claim LIKE 'J52%' OR cas.mem_cover_at_claim LIKE 'J53%'
              OR cas.mem_cover_at_claim LIKE 'J54%' OR cas.mem_cover_at_claim LIKE 'J58%'
              OR cas.mem_cover_at_claim LIKE 'J62%'
              OR cas.mem_cover_at_claim IN ('NZ','ENZ','FC','EFC','FCE','NZE','SP','ESP')
                                                                        THEN 'Top Cover'
            WHEN cas.mem_cover_at_claim IN ('NI','FI','VE')
              OR cas.mem_cover_at_claim LIKE 'I%'                       THEN 'Extras Only'
            WHEN cas.mem_cover_at_claim IN ('AMBU','AMBUU','PAM','BAM','AMB') THEN 'Ambulance'
            ELSE 'Exclusionary'
        END,
        -- Hospital_Product / Extras_Product already computed in SILVER
        cas.Hospital_Product,
        cas.Extras_Product,
        -- Hospital_Code from SILVER; Extras_Code derived from mem_cover_at_claim
        cas.Hospital_Product_Code,
        RTRIM(CASE
            WHEN RIGHT(cas.mem_cover_at_claim, 3) LIKE '5%' THEN ''
            WHEN RIGHT(cas.mem_cover_at_claim, 3) = 'MBU'   THEN 'AMBU'
            ELSE RIGHT(cas.mem_cover_at_claim, 3)
        END),
        -- Product Type: hospital code first, fall back to extras code
        ISNULL(hp_pt.product_type, ep_pt.product_type),
        -- Hospital Product Group
        CASE
            WHEN cas.Product_Description_at_claim LIKE 'Athlete%'   THEN 'Athlete'
            WHEN cas.Product_Description_at_claim LIKE '%Overseas%'  THEN 'Overseas'
            WHEN cas.Product_Description_at_claim LIKE 'Bronze%'     THEN 'Bronze'
            WHEN cas.Product_Description_at_claim LIKE '%Basic%'     THEN 'Basic'
            WHEN cas.Product_Description_at_claim LIKE '%Assure%'    THEN 'Silver Assure'
            WHEN cas.Product_Description_at_claim LIKE '%Nurture%'   THEN 'Silver Nurture'
            WHEN cas.Product_Description_at_claim LIKE 'Silver%'     THEN 'Silver'
            WHEN cas.mem_cover_at_claim LIKE '%FC%'
              OR cas.mem_cover_at_claim LIKE '%NZ%'                  THEN 'Gold Package'
            WHEN cas.Product_Description_at_claim LIKE '%Gold%'      THEN 'Gold'
            ELSE 'Ungrouped'
        END,
        -- Extras Product Group
        CASE
            WHEN cas.Product_Description_at_claim LIKE 'Athlete%'             THEN 'Athlete'
            WHEN cas.Product_Description_at_claim LIKE '%Overseas%'           THEN 'Overseas'
            WHEN cas.Product_Description_at_claim LIKE '%50s'                 THEN 'High Extras Over 50s'
            WHEN cas.Product_Description_at_claim LIKE '%High Extras%'        THEN 'High Extras'
            WHEN cas.Product_Description_at_claim LIKE '%Ultimate Pro Extras%' THEN 'Ultimate Pro Extras'
            WHEN cas.Product_Description_at_claim LIKE '%Ultimate Extras%'    THEN 'Ultimate Extras'
            WHEN cas.Product_Description_at_claim LIKE '%Mid Extras%'
              OR cas.Product_Description_at_claim LIKE '%Silver%'             THEN 'Mid Extras'
            WHEN cas.Product_Description_at_claim LIKE '%Essential Pro Extras%' THEN 'Essential Pro Extras'
            WHEN cas.Product_Description_at_claim LIKE '%Starter Extras%'
              OR cas.Product_Description_at_claim LIKE '%Value Extras%'
              OR cas.Product_Description_at_claim LIKE '%Value First%'        THEN 'Starter Extras'
            WHEN cas.mem_cover_at_claim LIKE '%FC%'
              OR cas.mem_cover_at_claim LIKE '%NZ%'                          THEN 'Gold'
            WHEN cas.Product_Description_at_claim LIKE '%Gold%'
              OR cas.Product_Description_at_claim LIKE '%Platinum%'          THEN 'Gold'
            WHEN cas.Product_Description_at_claim LIKE '%Ambulance%'         THEN 'Ambulance'
            ELSE 'Ungrouped'
        END,
        -- Audience Segment
        -- TODO: depends on Branch_Group (Excel BranchMapping, excluded from SQL).
        --       Cannot be computed without Branch_Group values ('Metro'/'Regional').
        --       Compute in Power BI once Branch Mapping Excel is loaded.
        NULL,

        -- ------------------------------------------------
        -- Service / Item
        -- ------------------------------------------------
        RTRIM(cdgh.provider_number_id),
        cdgh.service_type,
        ISNULL(svc.description, 'Other'),
        cdgh.item_number,
        RTRIM(cdgh.service_type + cdgh.item_number),
        ISNULL(itm.description, 'Missing'),
        ISNULL(ig.description,  'Missing'),
        ISNULL(fc.item_type,    'No Classification'),
        itm.disable_hicap_flag,
        itm.online_claiming,
        itm.topup_flag,
        CAST(itm.create_datetime AS DATE),
        CAST(itm.ceased_date     AS DATE),
        -- Dental Classification
        CASE
            WHEN cdgh.service_type = 'DENTAL' AND cdgh.item_number = 'Q100'                          THEN 'Dental Top Up'
            WHEN cdgh.service_type = 'DENTAL' AND TRY_CAST(cdgh.item_number AS INT) BETWEEN   0 AND  99 THEN 'Diagnostic Services'
            WHEN cdgh.service_type = 'DENTAL' AND TRY_CAST(cdgh.item_number AS INT) BETWEEN 100 AND 199 THEN 'Preventative Services'
            WHEN cdgh.service_type = 'DENTAL' AND TRY_CAST(cdgh.item_number AS INT) BETWEEN 200 AND 299 THEN 'Periodontics'
            WHEN cdgh.service_type = 'DENTAL' AND TRY_CAST(cdgh.item_number AS INT) BETWEEN 300 AND 399 THEN 'Oral Surgery'
            WHEN cdgh.service_type = 'DENTAL' AND TRY_CAST(cdgh.item_number AS INT) BETWEEN 400 AND 499 THEN 'Endodontics'
            WHEN cdgh.service_type = 'DENTAL' AND TRY_CAST(cdgh.item_number AS INT) BETWEEN 500 AND 599 THEN 'Restorative Services'
            WHEN cdgh.service_type = 'DENTAL' AND TRY_CAST(cdgh.item_number AS INT) BETWEEN 600 AND 799 THEN 'Orthodontics'
            WHEN cdgh.service_type = 'DENTAL' AND TRY_CAST(cdgh.item_number AS INT) BETWEEN 800 AND 899 THEN 'General Services'
            WHEN cdgh.service_type = 'DENTAL' AND TRY_CAST(cdgh.item_number AS INT) BETWEEN 900 AND 979 THEN 'Miscellaneous'
            ELSE 'Other'
        END,
        -- Optical Classification
        CASE
            WHEN cdgh.service_type = 'OPTICAL' AND TRY_CAST(cdgh.item_number AS INT) BETWEEN 110 AND 210 THEN 'Frames'
            WHEN cdgh.service_type = 'OPTICAL' AND TRY_CAST(cdgh.item_number AS INT) BETWEEN 211 AND 310 THEN 'Single Vision'
            WHEN cdgh.service_type = 'OPTICAL' AND TRY_CAST(cdgh.item_number AS INT) BETWEEN 311 AND 410 THEN 'Bifocal'
            WHEN cdgh.service_type = 'OPTICAL' AND TRY_CAST(cdgh.item_number AS INT) BETWEEN 411 AND 510 THEN 'Trifocal'
            WHEN cdgh.service_type = 'OPTICAL' AND TRY_CAST(cdgh.item_number AS INT) BETWEEN 511 AND 810 THEN 'Progressive'
            WHEN cdgh.service_type = 'OPTICAL' AND TRY_CAST(cdgh.item_number AS INT) >= 811              THEN 'Contacts'
            ELSE 'Other'
        END,
        -- POC Item Classification
        CASE
            WHEN cdgh.service_type = 'OPTICAL'
                 AND cdgh.item_number IN ('110','211','212','221','222','311','312','411','412',
                                          '511','512','811','812','821','822','831','832','842',
                                          '851','852','853','854','861','862','863','864',
                                          '871','872','873','874','700')          THEN 'Optical POC'
            WHEN cdgh.service_type = 'PHYSIO'
                 AND cdgh.item_number IN ('500','505','506','520','525','526','530','535','536',
                                          '560','561','570','582','591','592','593','595','596',
                                          '811','812')                            THEN 'Physio POC'
            WHEN cdgh.service_type = 'DENTAL'
                 AND (cdgh.item_number IN ('11','12','22','111','114','121','161')
                      OR cdgh.item_number LIKE '151%')                            THEN 'Dental POC'
            ELSE 'Exclusionary'
        END,
        pil.presenting_illness,

        -- ------------------------------------------------
        -- Provider
        -- ------------------------------------------------
        pn.provider_id,
        ISNULL(pt.description,  'Missing'),
        ISNULL(prv.provider_name, 'Missing'),
        ISNULL(pg2.description,  'No Group'),
        pn.surname,
        pn.first_name,
        pn.other_names,
        pn.title,
        pn.postcode,
        agr.agreement_id,

        -- ------------------------------------------------
        -- Sales / Channel
        -- ------------------------------------------------
        lpsc.promotion_description,
        lpsc.sales_channel_description,

        -- ------------------------------------------------
        -- Hospital / Clinical
        -- ------------------------------------------------
        chi.admission_date,
        chi.discharge_date,
        cgi.drug_store_name,
        NULL,  -- Clinical_Category: TODO pending source table identification

        -- Manual Claim
        CASE WHEN COALESCE(cgi.manual_flag, chi_m.manual_flag) = 1
             THEN 'Manual Claim' ELSE 'Not Manual Claim' END,
        ISNULL(mbr.description, 'No Reason')

    FROM BRONZE.dbo.claim_line cl

    -- Core claim detail (SILVER)
    INNER JOIN SILVER.dbo.Claim_Detail_Gen_And_Hosp cdgh
        ON cl.claim_id      = cdgh.claim_id
       AND cl.claim_line_id = cdgh.claim_line_id

    -- Claim received channel
    LEFT JOIN SILVER.dbo.Claims_By_Channel cbc
        ON cl.claim_id = cbc.claim_id

    -- Membership (for effective join date)
    LEFT JOIN BRONZE.dbo.memship ms
        ON cdgh.membership_id = ms.membership_id

    -- Member cover (current product description)
    LEFT JOIN BRONZE.dbo.MemberCover mc
        ON cdgh.membership_id = mc.membership_id

    -- Person (name, gender)
    LEFT JOIN BRONZE.dbo.person p
        ON cdgh.person_id = p.person_id

    -- Previous fund: join on (membership_id, person_id) — table key is composite
    LEFT JOIN SILVER.dbo.Previous_Fund pf
        ON cdgh.membership_id = pf.membership_id
       AND cdgh.person_id     = pf.person_id

    -- Cover details at service time (SILVER)
    LEFT JOIN SILVER.dbo.ClaimDetailsAtService_optimised cas
        ON cdgh.claim_id      = cas.claim_id
       AND cdgh.claim_line_id = cas.claim_line_id

    -- Cover type description
    LEFT JOIN BRONZE.dbo.cover_type ct
        ON cas.cover_type = ct.cover_type

    -- Branch (grouping)
    LEFT JOIN BRONZE.dbo.grouping grp
        ON cas.mem_branch_at_claim = grp.group_id

    -- Product type — hospital code (H only; DISTINCT handles duplicate rows per code)
    LEFT JOIN (SELECT DISTINCT product_code, product_type
               FROM BRONZE.dbo.product
               WHERE product_type = 'H') hp_pt
        ON RTRIM(cas.Hospital_Product_Code) = hp_pt.product_code

    -- Product type — extras code fallback (A only; DISTINCT handles duplicate rows per code)
    LEFT JOIN (SELECT DISTINCT product_code, product_type
               FROM BRONZE.dbo.product
               WHERE product_type = 'A') ep_pt
        ON RTRIM(CASE
                    WHEN RIGHT(cas.mem_cover_at_claim, 3) LIKE '5%' THEN ''
                    WHEN RIGHT(cas.mem_cover_at_claim, 3) = 'MBU'   THEN 'AMBU'
                    ELSE RIGHT(cas.mem_cover_at_claim, 3)
                 END) = ep_pt.product_code

    -- Service type description
    LEFT JOIN BRONZE.dbo.service_type svc
        ON cdgh.service_type = svc.service_type

    -- Item details
    LEFT JOIN BRONZE.dbo.item itm
        ON cdgh.service_type        = itm.service_type
       AND RTRIM(cdgh.item_number)  = RTRIM(itm.item_number)

    -- Item group description
    LEFT JOIN BRONZE.dbo.item_group ig
        ON itm.item_group = ig.item_group

    -- Fund classification
    LEFT JOIN BRONZE.dbo.fund_classification fc
        ON itm.fund_classification_id = fc.id

    -- Item average fee (Red/Blue limits)
    LEFT JOIN BRONZE.dbo.item_averagefee iaf
        ON cdgh.service_type       = iaf.service_type
       AND RTRIM(cdgh.item_number) = RTRIM(iaf.item_number)

    -- Provider number details
    LEFT JOIN BRONZE.dbo.provider_number pn
        ON RTRIM(cdgh.provider_number_id) = RTRIM(pn.provider_number_id)

    -- Provider (type code + name)
    LEFT JOIN BRONZE.dbo.provider prv
        ON RTRIM(pn.provider_id) = RTRIM(prv.provider_id)

    -- Provider type description
    LEFT JOIN BRONZE.dbo.provider_type pt
        ON prv.provider_type = pt.provider_type

    -- Provider group description
    LEFT JOIN BRONZE.dbo.provider_group pg2
        ON pn.provider_group_id = pg2.provider_group_id

    -- Agreement ID
    LEFT JOIN AgreementCTE agr
        ON RTRIM(cdgh.provider_number_id) = agr.provider_number_id

    -- Schedule fee (DENTAL/OPTICAL latest contract)
    LEFT JOIN ContractItemCTE ci_cte
        ON cdgh.service_type       = ci_cte.service_type
       AND RTRIM(cdgh.item_number) = RTRIM(ci_cte.item_number)

    -- Presenting illness (Medical items)
    LEFT JOIN PresentingIllnessCTE pil
        ON RTRIM(cdgh.item_number) = RTRIM(pil.item_number)
       AND cdgh.claim_type = 'Medical'

    -- Hospital admission/discharge dates
    LEFT JOIN HospitalItemCTE chi
        ON cdgh.claim_id      = chi.claim_id
       AND cdgh.claim_line_id = chi.claim_line_id

    -- General item (drug store + medicare benefit)
    LEFT JOIN BRONZE.dbo.claim_generalitem cgi
        ON cdgh.claim_id      = cgi.claim_id
       AND cdgh.claim_line_id = cgi.claim_line_id

    -- Web login flag (DISTINCT deduplicates the 3 duplicate oper_names in source)
    LEFT JOIN (SELECT DISTINCT oper_name FROM BRONZE.dbo.web_security) ws
        ON cdgh.update_operator = ws.oper_name

    -- Promotion / Sales channel (SILVER)
    LEFT JOIN SILVER.dbo.Latest_Promo_Sales_Channel_Operator lpsc
        ON cdgh.membership_id = lpsc.membership_id
       AND cdgh.person_id     = lpsc.person_id

    -- Agent
    LEFT JOIN BRONZE.dbo.MemberAgent ma
        ON cdgh.membership_id = ma.membership_id

    -- Manual flag from hospital item (general item already joined as cgi)
    LEFT JOIN BRONZE.dbo.claim_hospitalitem chi_m
        ON cdgh.claim_id      = chi_m.claim_id
       AND cdgh.claim_line_id = chi_m.claim_line_id

    -- Manual benefit reason description
    LEFT JOIN BRONZE.dbo.manual_benefit_reason mbr
        ON mbr.manual_benefit_reason_id = COALESCE(cgi.manual_benefit_reason_id, chi_m.manual_benefit_reason_id)

    -- LOM (Length of Membership): computed once, reused for cohort
    CROSS APPLY (
        SELECT
            DATEDIFF(YEAR,
                CAST(ms.effective_join_date AS DATE),
                CAST(cdgh.status_date AS DATE))
            - CASE WHEN DATEADD(YEAR,
                            DATEDIFF(YEAR,
                                CAST(ms.effective_join_date AS DATE),
                                CAST(cdgh.status_date AS DATE)),
                            CAST(ms.effective_join_date AS DATE))
                        > CAST(cdgh.status_date AS DATE)
                   THEN 1 ELSE 0 END
            AS lom
    ) lom_calc

    WHERE cl.claim_line_status_type = 'P'
      AND cl.status_date            > @StartDate;

END;
GO
```

# [dbo].[usp_Load_Claims_By_Channel]

```sql
/****** Object:  StoredProcedure [dbo].[usp_Load_Claims_By_Channel]    Script Date: 29/06/2026 1:38:34 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE   PROCEDURE [dbo].[usp_Load_Claims_By_Channel]
AS
BEGIN
    SET NOCOUNT ON;

    TRUNCATE TABLE SILVER.dbo.Claims_By_Channel;

    INSERT INTO SILVER.dbo.Claims_By_Channel (
        claim_id,
        category,
        create_operator
    )
    SELECT
        cl.claim_id,
        CASE
            WHEN (cl.create_operator LIKE '[0-9]%[A-Z]' OR cl.create_operator = 'ECLIPSE') AND cl.claim_type_flag = 'H' THEN 'HOSPECLIPSE'
            WHEN cl.create_operator LIKE '[0-9]%[A-Z]' AND ISNULL(cl.claim_type_flag, '') = ''                           THEN 'ECLIPSE'
            WHEN cl.create_operator = 'HICAPS'                                                                            THEN 'HICAPS'
            WHEN cl.create_operator = 'IBA'                                                                               THEN 'IBA'
            WHEN cl.create_operator = 'SCANNING'                                                                          THEN 'SCANNING'
            WHEN cl.create_operator = CONVERT(VARCHAR(10), cl.membership_id)                                              THEN 'ECLAIMS'
            WHEN cl.till_id IS NOT NULL                                                                                   THEN 'TILLCLAIMS'
            WHEN ab.claim_source_id = 7                                                                                   THEN 'PRACTICE'
            WHEN cl.create_operator = 'MobWeb'                                                                            THEN 'MOBILE'
            WHEN cl.create_operator LIKE '%-%'                                                                            THEN 'ECLAIMS'
            ELSE CASE WHEN cl.claim_type_flag = 'H' THEN 'HPAPER' ELSE 'PAPER' END
        END                                                         AS category,
        cl.create_operator
    FROM BRONZE.dbo.claim AS cl
    LEFT JOIN (
        SELECT   claim_id, MAX(claim_source_id) AS claim_source_id
        FROM     BRONZE.dbo.claim_line
        WHERE    claim_source_id IS NOT NULL
        AND      claim_source_id NOT IN (2, 6, 8)
        GROUP BY claim_id
    ) AS ab ON ab.claim_id = cl.claim_id
    WHERE cl.create_operator NOT IN ('system')
    AND ISNULL(cl.claim_type_flag, '') <> 'Q'
    AND (
        SELECT TOP 1 cs.claim_status_type
        FROM     BRONZE.dbo.claim_status AS cs
        WHERE    cs.claim_id = cl.claim_id
        AND      cs.status_date <= GETDATE()
        ORDER BY cs.claim_status_version DESC
    ) <> 'C';

END;
GO
```

# [dbo].[usp_Load_Current_Product_Fee]

```sql
/****** Object:  StoredProcedure [dbo].[usp_Load_Current_Product_Fee]    Script Date: 29/06/2026 1:38:34 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE   PROCEDURE [dbo].[usp_Load_Current_Product_Fee]
AS
BEGIN
    SET NOCOUNT ON;

    TRUNCATE TABLE SILVER.dbo.Current_Product_Fee;

    INSERT INTO SILVER.dbo.Current_Product_Fee (
        Cover_State,
        Cover_Type,
        Product_Id,
        Product_Fee_Amount
    )
    SELECT
        pf.cover_state,
        pf.cover_type,
        pf.product_id,
        pf.product_fee_amount
    FROM BRONZE.dbo.product_fee pf
    WHERE pf.product_fee_version = (
        SELECT MAX(pf2.product_fee_version)
        FROM BRONZE.dbo.product_fee pf2
        WHERE pf2.fee_from_date  <= GETDATE()
          AND pf2.cover_state     = pf.cover_state
          AND pf2.cover_type      = pf.cover_type
          AND pf2.product_id      = pf.product_id
    );

END;
GO
```

# [dbo].[usp_Load_Declined_Hicaps_Claim]

```sql
/****** Object:  StoredProcedure [dbo].[usp_Load_Declined_Hicaps_Claim]    Script Date: 29/06/2026 1:38:34 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE   PROCEDURE [dbo].[usp_Load_Declined_Hicaps_Claim]
AS
BEGIN
    SET NOCOUNT ON;

    -- Rolling window: first day of month 24 months ago → last day of month 24 months ahead
    DECLARE @DateStart DATE = DATEADD(MONTH, DATEDIFF(MONTH, 0, GETDATE()) - 24, 0);
    DECLARE @DateEnd   DATE = DATEADD(MONTH, DATEDIFF(MONTH, 0, GETDATE()) + 25, 0); -- exclusive upper bound

    TRUNCATE TABLE SILVER.dbo.Declined_Hicaps_Claim;

    -- Latest cover version per membership (replaces Paragon_MemberCover.qvd)
    WITH MemberCoverLatest AS (
        SELECT
            membership_id,
            cover_version,
            cover_type,
            status_flag,
            cover_state,
            cover_from_date,
            termination_date,
            description         AS Cover,
            FixCode,
            Product_Description,
            ROW_NUMBER() OVER (
                PARTITION BY membership_id
                ORDER BY cover_version DESC
            ) AS rn
        FROM BRONZE.dbo.MemberCover
    )

    -- Latest note per membership (avoids row duplication where membership has multiple notes)
    , NoteLatest AS (
        SELECT *,
            ROW_NUMBER() OVER (
                PARTITION BY main_ref_id
                ORDER BY create_datetime DESC
            ) AS rn
        FROM BRONZE.dbo.note
        WHERE sub_ref_type = 'L' OR sub_ref_type_id = 76
    )

    INSERT INTO SILVER.dbo.Declined_Hicaps_Claim (
        Claim_ID,
        Claim_Line_ID,
        Membership_ID,
        Status_Date,
        Status_Date_Month_Year,
        Create_Operator,
        Assessing_Code_Type,
        Item_Number,
        Service_Type,
        Branch,
        Main_Reference_Type,
        Reference_Type,
        Reference_Create_Operator,
        Reference_Create_Date,
        Reference_Description,
        Sub_Sub_Ref_Type_ID,
        Cover_Version,
        Cover_Type,
        Status_Flag,
        Cover_State,
        Cover_From_Date,
        Termination_Date,
        Cover,
        Fix_Code,
        Product_Description
    )
    SELECT
        cl.claim_id,
        cl.claim_line_id,
        cl.membership_id,
        CAST(cl.status_date AS DATE),
        FORMAT(cl.status_date, 'MMM yyyy'),
        cl.create_operator,
        cg.assessing_code_type,
        cg.item_number,
        cg.service_type,
        ISNULL(g.description, 'MISSING')            AS Branch,
        n.main_ref_type,
        n.sub_ref_type,
        n.create_operator,
        CAST(n.create_datetime AS DATE),
        ISNULL(srt.description, 'CheckNotes')        AS Reference_Description,
        n.sub_sub_ref_type_id,
        mc.cover_version,
        mc.cover_type,
        mc.status_flag,
        mc.cover_state,
        CAST(mc.cover_from_date AS DATE),
        CAST(mc.termination_date AS DATE),
        mc.Cover,
        mc.FixCode,
        mc.Product_Description
    FROM BRONZE.dbo.claim_line AS cl

    INNER JOIN BRONZE.dbo.claim_generalitem AS cg
        ON  cg.claim_id      = cl.claim_id
        AND cg.claim_line_id = cl.claim_line_id

    -- Branch: use pre-computed SILVER table, then look up description from grouping
    LEFT JOIN SILVER.dbo.ClaimDetailsAtService_optimised AS cds
        ON  cds.claim_id      = cl.claim_id
        AND cds.claim_line_id = cl.claim_line_id
    LEFT JOIN BRONZE.dbo.grouping AS g
        ON  g.group_id   = cds.mem_branch_at_claim
        AND g.group_type = 'C'

    -- Latest note per membership only (sub_ref_type filter applied in CTE above)
    LEFT JOIN NoteLatest AS n
        ON  n.main_ref_id = CAST(cl.membership_id AS VARCHAR(20))
        AND n.rn = 1
    LEFT JOIN BRONZE.dbo.sub_ref_type AS srt
        ON  srt.sub_ref_type_id = n.sub_ref_type_id

    -- Latest cover per membership
    INNER JOIN MemberCoverLatest AS mc
        ON  mc.membership_id = cl.membership_id
        AND mc.rn = 1

    WHERE cl.status_date >= @DateStart
      AND cl.status_date <  @DateEnd
      AND cl.create_operator IN ('HICAPS', 'IBA')
      AND cg.assessing_code_type IS NOT NULL
      AND NOT (cg.item_number LIKE '6%' AND cg.service_type = 'OPTICAL');

END;
GO
```

# [dbo].[usp_Load_Earned_Contributions]

```sql
/****** Object:  StoredProcedure [dbo].[usp_Load_Earned_Contributions]    Script Date: 29/06/2026 1:38:34 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[usp_Load_Earned_Contributions]
    @StartMonth  date = '2026-05-01',        -- First CurrentMonth to process; defaults to current month (single-month run)
    @CurrRI      date = '2026-04-01' -- Current Rate Increase date; update annually
AS
BEGIN
    SET NOCOUNT ON;

    -- 1. 日期变量初始化
    DECLARE @EndMonth date = DATEFROMPARTS(YEAR(GETDATE()), MONTH(GETDATE()), 1);

    IF @StartMonth IS NULL
        SET @StartMonth = @EndMonth;  -- 默认只跑上个月（现有行为）

    DECLARE @LoopMonth      date;
    DECLARE @PreviousMonth  date;
    DECLARE @ReportMonth    date;
    DECLARE @ReportMonthEnd date;
    DECLARE @PrevRI         date = DATEADD(YEAR, -1, @CurrRI);  -- 不随循环变化，声明一次

    -- 2. 清空目标表（只执行一次）
    TRUNCATE TABLE dbo.Earned_Contributions;

    -- 3. 预计算费率（不依赖循环月份，提到循环外）
    SELECT
        cover_state, cover_type, product_id,
        ISNULL(CAST(MAX(CASE WHEN fee_from_date <= @CurrRI THEN product_fee_amount * 52.0 / 12 END) AS decimal(18,2)), 0) AS curr_mrate, --changed to <= on 13/05 per requested
        ISNULL(CAST(MAX(CASE WHEN fee_from_date <= @PrevRI THEN product_fee_amount * 52.0 / 12 END) AS decimal(18,2)), 0) AS prev_mrate  --changed to <= on 13/05 per requested
    INTO #product_rates
    FROM BRONZE.dbo.product_fee
    GROUP BY cover_state, cover_type, product_id;

    -- 4. 按月循环
    SET @LoopMonth = @StartMonth;

    WHILE @LoopMonth <= @EndMonth
    BEGIN
        SET @PreviousMonth  = DATEADD(MONTH, -1, @LoopMonth);
        SET @ReportMonth    = @PreviousMonth;
        SET @ReportMonthEnd = EOMONTH(@ReportMonth);

        -- 核心计算逻辑
        WITH Advances_Arrears AS (
            SELECT
                a.membership_id,
                @ReportMonth AS Month_Year,
                (ISNULL(b.member_arrears, 0) - ISNULL(c.member_arrears, 0)) AS Net_Arrears,
                (ISNULL(b.member_advance, 0) - ISNULL(c.member_advance, 0)) AS Net_Advance
            FROM (SELECT DISTINCT membership_id FROM BRONZE.dbo.group_key_full_by_branch WHERE rundate BETWEEN @PreviousMonth AND @LoopMonth) a
            LEFT OUTER JOIN BRONZE.dbo.group_key_full_by_branch b
                ON a.membership_id = b.membership_id AND b.rundate = @LoopMonth
            LEFT OUTER JOIN BRONZE.dbo.group_key_full_by_branch c
                ON a.membership_id = c.membership_id AND c.rundate = @PreviousMonth
        ),

        Receipts_raw AS (
            SELECT
                r.membership_id,
                ISNULL(r.con_discount_pct, 0) / 100 AS con_discount,
                CAST((r.receipt_amount + r.rebate_amount) AS decimal(18,2)) AS received_amount,
                CAST((ISNULL(r.receipt_amount_bonus, 0) + ISNULL(r.rebate_amount_bonus, 0)) AS decimal(18,2)) AS bonus_amount
            FROM BRONZE.dbo.receipt AS r
            INNER JOIN BRONZE.dbo.receipt_status rs
                ON r.receipt_id = rs.receipt_id AND r.membership_id = rs.membership_id
            WHERE r.create_datetime >= @ReportMonth
              AND r.create_datetime < DATEADD(DAY, 1, @ReportMonthEnd)
              AND rs.receipt_status_type IN ('P', 'U', 'A')
        ),

        Monthly_receipts AS (
            SELECT
                membership_id,
                MAX(con_discount) AS con_discount,
                SUM(received_amount) AS received_amount,
                SUM(bonus_amount) AS bonus_amount
            FROM Receipts_raw
            GROUP BY membership_id
        ),

        All_contributions AS (
            SELECT
                COALESCE(mr.membership_id, aa.membership_id) AS membership_id,
                @ReportMonth AS Month_Year,
                ISNULL(mr.received_amount, 0) AS received_amount,
                ISNULL(mr.bonus_amount, 0)    AS bonus_amount,
                ISNULL(aa.Net_Advance, 0)     AS Net_Advance,
                ISNULL(aa.Net_Arrears, 0)     AS Net_Arrears,
                ISNULL(mr.con_discount, 0)    AS con_discount
            FROM Monthly_receipts mr
            FULL OUTER JOIN Advances_Arrears aa
                ON mr.membership_id = aa.membership_id
        ),

        Max_cover_version AS (
            SELECT cp.membership_id, MAX(cp.cover_version) AS max_cv
            FROM BRONZE.dbo.cover_product cp
            INNER JOIN BRONZE.dbo.cover c
                ON c.membership_id = cp.membership_id AND c.cover_version = cp.cover_version
            WHERE c.cover_from_date <= @LoopMonth
            GROUP BY cp.membership_id
        ),

        Member_details AS (
            SELECT
                cp.membership_id, cp.cover_version, cp.product_id,
                p.product_type,
                c.cover_type, c.cover_state
            FROM BRONZE.dbo.cover_product cp
            INNER JOIN Max_cover_version mcv
                ON cp.membership_id = mcv.membership_id AND cp.cover_version = mcv.max_cv
            LEFT OUTER JOIN BRONZE.dbo.product p
                ON cp.product_id = p.product_id
            LEFT OUTER JOIN BRONZE.dbo.cover c
                ON c.membership_id = cp.membership_id AND c.cover_version = cp.cover_version
        ),

        DRL AS (
            SELECT
                m.membership_id,
                (m.member1_loading / 100) AS mem1_loading,
                (m.member2_loading / 100) AS mem2_loading,
                ISNULL(m.overall_agediscount, 0) / 100 AS overall_agediscount,
                ISNULL(gr.grp_discount_amount, 0) / 100 AS corporate_discount
            FROM BRONZE.dbo.memship m
            LEFT OUTER JOIN BRONZE.dbo.memship_app_agent ma ON m.membership_id = ma.membership_id
            LEFT OUTER JOIN BRONZE.dbo.grouping gr ON ma.agent_group_id = gr.group_id
            WHERE m.memship_status NOT IN ('L', 'P')
        ),

        Combined_data AS (
            SELECT
                ac.*,
                (received_amount + bonus_amount + Net_Arrears - Net_Advance) AS earned_contributions,
                ISNULL(MAX(CASE WHEN md.product_type = 'H' THEN md.product_id END), 0) AS H_prod,
                ISNULL(MAX(CASE WHEN md.product_type = 'A' THEN md.product_id END), 0) AS A_prod,
                ISNULL(MAX(CASE WHEN md.product_type = 'B' THEN md.product_id END), 0) AS B_prod,
                md.cover_type,
                md.cover_state,
                CASE WHEN md.cover_type IN ('A', 'B', 'P', 'S') THEN d.mem1_loading ELSE (d.mem1_loading + d.mem2_loading) / 2 END AS overall_lhc,
                d.overall_agediscount,
                d.corporate_discount
            FROM All_contributions ac
            LEFT OUTER JOIN Member_details md
                ON md.membership_id = ac.membership_id
            LEFT OUTER JOIN DRL d
                ON ac.membership_id = d.membership_id
            GROUP BY
                ac.membership_id, ac.Month_Year, ac.received_amount, ac.bonus_amount,
                ac.Net_Advance, ac.Net_Arrears, md.cover_type, md.cover_state,
                ac.con_discount, d.overall_agediscount, d.corporate_discount,
                d.mem1_loading, d.mem2_loading
        ),

        Latest_payment_date AS (
            SELECT membership_id, MAX(old_paid_to) AS old_paidto
            FROM BRONZE.dbo.receipt
            WHERE create_datetime >= EOMONTH(@LoopMonth, -24)
            GROUP BY membership_id
        ),

        All_Data AS (
            SELECT
                a.*, c.old_paidto,
                CASE WHEN c.old_paidto < @CurrRI THEN 1 ELSE 0 END AS RP_flag,
                ISNULL(b1.curr_mrate, 0) AS curr_H_rate, ISNULL(b1.prev_mrate, 0) AS prev_H_rate,
                ISNULL(b2.curr_mrate, 0) AS curr_A_rate, ISNULL(b2.prev_mrate, 0) AS prev_A_rate,
                ISNULL(b3.curr_mrate, 0) AS curr_B_rate, ISNULL(b3.prev_mrate, 0) AS prev_B_rate
            FROM Combined_data a
            LEFT OUTER JOIN Latest_payment_date c ON a.membership_id = c.membership_id
            LEFT OUTER JOIN #product_rates b1 ON a.cover_state = b1.cover_state AND a.cover_type = b1.cover_type AND b1.product_id = a.H_prod
            LEFT OUTER JOIN #product_rates b2 ON a.cover_state = b2.cover_state AND a.cover_type = b2.cover_type AND b2.product_id = a.A_prod
            LEFT OUTER JOIN #product_rates b3 ON a.cover_state = b3.cover_state AND a.cover_type = b3.cover_type AND b3.product_id = a.B_prod
        )

        -- 插入当月数据
        INSERT INTO dbo.Earned_Contributions (
            Month_Year, membership_id, cover_state, cover_type,
            H_prod, A_prod, B_prod,
            earned_contributions, bonus_amount,
            H_earned, A_earned, B_earned,
            H_bonus, A_bonus, B_bonus)
        SELECT
            Month_Year, membership_id, cover_state, cover_type,
            H_prod, A_prod, B_prod,
            earned_contributions, bonus_amount,
            -- H_earned
            CASE WHEN ISNULL(RP_flag, 1) = 0
                 THEN CAST(((curr_H_rate * (1 + overall_lhc - con_discount) * (1 - corporate_discount - overall_agediscount)) / ((curr_H_rate * (1 + overall_lhc - con_discount) + curr_A_rate + curr_B_rate) * (1 - corporate_discount - overall_agediscount) + 0.0001)) * earned_contributions AS decimal(18,2))
                 ELSE CAST(((prev_H_rate * (1 + overall_lhc - con_discount) * (1 - corporate_discount - overall_agediscount)) / ((prev_H_rate * (1 + overall_lhc - con_discount) + prev_A_rate + prev_B_rate) * (1 - corporate_discount - overall_agediscount) + 0.0001)) * earned_contributions AS decimal(18,2))
            END,
            -- A_earned
            CASE WHEN ISNULL(RP_flag, 1) = 0
                 THEN CAST(((curr_A_rate * (1 - corporate_discount - overall_agediscount)) / ((curr_H_rate * (1 + overall_lhc - con_discount) + curr_A_rate + curr_B_rate) * (1 - corporate_discount - overall_agediscount) + 0.0001)) * earned_contributions AS decimal(18,2))
                 ELSE CAST(((prev_A_rate * (1 - corporate_discount - overall_agediscount)) / ((prev_H_rate * (1 + overall_lhc - con_discount) + prev_A_rate + prev_B_rate) * (1 - corporate_discount - overall_agediscount) + 0.0001)) * earned_contributions AS decimal(18,2))
            END,
            -- B_earned
            CASE WHEN ISNULL(RP_flag, 1) = 0
                 THEN CAST(((curr_B_rate * (1 - corporate_discount - overall_agediscount)) / ((curr_H_rate * (1 + overall_lhc - con_discount) + curr_A_rate + curr_B_rate) * (1 - corporate_discount - overall_agediscount) + 0.0001)) * earned_contributions AS decimal(18,2))
                 ELSE CAST(((prev_B_rate * (1 - corporate_discount - overall_agediscount)) / ((prev_H_rate * (1 + overall_lhc - con_discount) + prev_A_rate + prev_B_rate) * (1 - corporate_discount - overall_agediscount) + 0.0001)) * earned_contributions AS decimal(18,2))
            END,
            -- H_bonus
            CASE WHEN ISNULL(RP_flag, 1) = 0
                 THEN CAST(((curr_H_rate * (1 + overall_lhc - con_discount) * (1 - corporate_discount - overall_agediscount)) / ((curr_H_rate * (1 + overall_lhc - con_discount) + curr_A_rate + curr_B_rate) * (1 - corporate_discount - overall_agediscount) + 0.0001)) * bonus_amount AS decimal(18,2))
                 ELSE CAST(((prev_H_rate * (1 + overall_lhc - con_discount) * (1 - corporate_discount - overall_agediscount)) / ((prev_H_rate * (1 + overall_lhc - con_discount) + prev_A_rate + prev_B_rate) * (1 - corporate_discount - overall_agediscount) + 0.0001)) * bonus_amount AS decimal(18,2))
            END,
            -- A_bonus
            CASE WHEN ISNULL(RP_flag, 1) = 0
                 THEN CAST(((curr_A_rate * (1 - corporate_discount - overall_agediscount)) / ((curr_H_rate * (1 + overall_lhc - con_discount) + curr_A_rate + curr_B_rate) * (1 - corporate_discount - overall_agediscount) + 0.0001)) * bonus_amount AS decimal(18,2))
                 ELSE CAST(((prev_A_rate * (1 - corporate_discount - overall_agediscount)) / ((prev_H_rate * (1 + overall_lhc - con_discount) + prev_A_rate + prev_B_rate) * (1 - corporate_discount - overall_agediscount) + 0.0001)) * bonus_amount AS decimal(18,2))
            END,
            -- B_bonus
            CASE WHEN ISNULL(RP_flag, 1) = 0
                 THEN CAST(((curr_B_rate * (1 - corporate_discount - overall_agediscount)) / ((curr_H_rate * (1 + overall_lhc - con_discount) + curr_A_rate + curr_B_rate) * (1 - corporate_discount - overall_agediscount) + 0.0001)) * bonus_amount AS decimal(18,2))
                 ELSE CAST(((prev_B_rate * (1 - corporate_discount - overall_agediscount)) / ((prev_H_rate * (1 + overall_lhc - con_discount) + prev_A_rate + prev_B_rate) * (1 - corporate_discount - overall_agediscount) + 0.0001)) * bonus_amount AS decimal(18,2))
            END
        FROM All_Data;

        SET @LoopMonth = DATEADD(MONTH, 1, @LoopMonth);
    END

    DROP TABLE #product_rates;

END;
GO
```

# [dbo].[usp_Load_Episode_Base_Data]

```sql
/****** Object:  StoredProcedure [dbo].[usp_Load_Episode_Base_Data]    Script Date: 29/06/2026 1:38:34 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[usp_Load_Episode_Base_Data]
AS
BEGIN
    SET NOCOUNT ON;
    SET ANSI_WARNINGS OFF;
    
    DECLARE @RowCount INT;
    
    -- ========================================================================
    -- Step 1: 清空工作表
    -- ========================================================================
    TRUNCATE TABLE SILVER.dbo.Episode_Detail;
    
    -- ========================================================================
    -- Step 2: 加载基础 Claim 数据
    -- ========================================================================
    INSERT INTO SILVER.dbo.Episode_Detail (
        EpisodeClaimID,
        EpisodeClaimLineID,
        benefit
    )
    --SELECT 
    --    claim_id,
    --    claim_line_id,
    --    benefit
    --FROM BRONZE.dbo.ClaimDetailGenAndHosp
    --WHERE line_status = 'Paid'
    --    AND status_date >= DATEADD(YEAR, -6, GETDATE())
    --    AND claim_type IN ('Hospital', 'Medical');
    
	SELECT 
		cl.claim_id,
		cl.claim_line_id,
		COALESCE(ch.benefit, cg.benefit) AS benefit
	FROM BRONZE.dbo.claim_line AS cl
		INNER JOIN BRONZE.dbo.claim_line_status_type AS cls 
			ON cl.claim_line_status_type = cls.claim_line_status_type
		LEFT OUTER JOIN BRONZE.dbo.claim_hospitalitem AS ch 
			ON cl.claim_id = ch.claim_id 
			AND cl.claim_line_id = ch.claim_line_id
		LEFT OUTER JOIN BRONZE.dbo.claim_generalitem AS cg 
			ON cl.claim_id = cg.claim_id 
			AND cl.claim_line_id = cg.claim_line_id
	WHERE cls.description = 'Paid'
		AND cl.status_date >= DATEADD(YEAR, -6, GETDATE())
		AND (
			-- Hospital claims
			(ch.service_type IS NOT NULL)
			OR 
			-- Medical claims (from general items)
			(ch.service_type IS NULL AND cg.service_type = 'MEDICAL')
		);


    SET @RowCount = @@ROWCOUNT;
    
    -- ========================================================================
    -- Step 3: 关联 Episode ID（从 claim_line）
    -- ========================================================================
    UPDATE ew
    SET ew.EpisodeIDTmp = cl.episode_id
    FROM SILVER.dbo.Episode_Detail ew
    LEFT JOIN BRONZE.dbo.claim_line cl
        ON ew.EpisodeClaimID = cl.claim_id
        AND ew.EpisodeClaimLineID = cl.claim_line_id;
    
    -- ========================================================================
    -- Step 4: 关联 Episode ID（从 fact_claim_data）
    -- ========================================================================
    UPDATE ew
    SET ew.EpisodeIDTmp2 = fcd.episode_id
    FROM SILVER.dbo.Episode_Detail ew
    --LEFT JOIN SILVER.dbo.fact_claim_data fcd
	LEFT JOIN SILVER.dbo.Claim_Episode_Staging fcd
        ON ew.EpisodeClaimID = fcd.claim_id
        AND ew.EpisodeClaimLineID = fcd.claim_line_id;
    
    -- ========================================================================
    -- Step 5: 合并 Episode ID（优先使用 EpisodeIDTmp）
    -- ========================================================================
    UPDATE SILVER.dbo.Episode_Detail
    SET EpisodeID = COALESCE(EpisodeIDTmp, EpisodeIDTmp2);
    
    -- ========================================================================
    -- Step 6: 删除没有 Episode ID 的记录（INNER JOIN 效果）
    -- ========================================================================
    DELETE FROM SILVER.dbo.Episode_Detail
    WHERE EpisodeID IS NULL;
    
    -- ========================================================================
    -- Step 7: 关联 Episode 主表字段
    -- ========================================================================
    UPDATE ew
    SET ew.separation_mode = e.separation_mode,
        ew.membership_id = e.membership_id,
        ew.admission_date = e.admission_date,
        ew.discharge_date = e.discharge_date,
        ew.drg_number = e.drg_number,
        ew.drg_version = e.drg_version,
        ew.person_id = e.person_id,
        ew.admission_code = e.admission_code,
        ew.separation_code = e.separation_code,
        ew.episode_type = e.episode_type
    FROM SILVER.dbo.Episode_Detail ew
    INNER JOIN BRONZE.dbo.episode e
        ON ew.EpisodeID = e.episode_id;
    
    -- ========================================================================
    -- Step 8: 添加 Procedure 信息（ACHI 代码）
    -- ========================================================================
    UPDATE ew
    SET ew.ProcedureType = edp.diagnosis_procedure_type,
        ew.ProcedureNumber = edp.diagnosis_procedure_number,
        ew.EpisodeProcedureServiceType = edp.service_type,
        ew.EpisodeProcedureItemNumber = edp.item_number,
        ew.ProcedureConditionOnsetFlag = edp.condition_onset_flag
    FROM SILVER.dbo.Episode_Detail ew
    LEFT JOIN BRONZE.dbo.episode_diagnosis_procedure edp
        ON ew.EpisodeID = edp.episode_id
        AND edp.diagnosis_procedure_type = 'P'
        AND edp.diagnosis_procedure_number = '1';
    
    -- ========================================================================
    -- Step 9: 添加 ACHI 代码详细信息（最新版本）
    -- ========================================================================
    WITH LatestProcedure AS (
        SELECT 
            m1.icd_item,
            REPLACE(m1.icd_item, '.', '') AS ACHI_CodeProcedure,
            m1.block,
            ROW_NUMBER() OVER (PARTITION BY m1.icd_item, m1.icd_type 
                              ORDER BY m1.edition DESC) as rn
        FROM BRONZE.dbo.medical_item_icd_10am m1
    )
    UPDATE ew
    SET ew.ACHI_CodeProcedure = lp.ACHI_CodeProcedure,
        ew.Procedure_Block = lp.block
    FROM SILVER.dbo.Episode_Detail ew
    LEFT JOIN LatestProcedure lp
        ON ew.EpisodeProcedureItemNumber = lp.icd_item
        AND lp.rn = 1;
    
    -- ========================================================================
    -- Step 10: 添加 Diagnosis 信息（ICD-10 代码）
    -- ========================================================================
    UPDATE ew
    SET ew.DiagnosisType = edp.diagnosis_procedure_type,
        ew.DiagnosisNumber = edp.diagnosis_procedure_number,
        ew.EpisodeDiagnosisServiceType = edp.service_type,
        ew.EpisodeDiagnosisItemNumber = edp.item_number,
        ew.DiagnosisConditionOnsetFlag = edp.condition_onset_flag
    FROM SILVER.dbo.Episode_Detail ew
    LEFT JOIN BRONZE.dbo.episode_diagnosis_procedure edp
        ON ew.EpisodeID = edp.episode_id
        AND edp.diagnosis_procedure_type = 'D'
        AND edp.diagnosis_procedure_number = '1';
    
    -- ========================================================================
    -- Step 11: 计算 ChapterItemNumberTmp
    -- ========================================================================
    UPDATE SILVER.dbo.Episode_Detail
    SET ChapterItemNumberTmp = COALESCE(EpisodeDiagnosisItemNumber, EpisodeProcedureItemNumber);
    
    -- ========================================================================
    -- Step 12: 添加 ICD-10 详细信息和 Block 分类
    -- ========================================================================
    WITH LatestICD AS (
        SELECT 
            m1.icd_item,
            REPLACE(m1.icd_item, '.', '') AS ACHI_Code,
            m1.icd_type,
            m1.edition AS Edition,
            m1.long_description,
            m1.short_description,
            m1.block,
            m1.code_id_clean,
            ROW_NUMBER() OVER (PARTITION BY m1.icd_item, m1.icd_type 
                              ORDER BY m1.edition DESC) as rn
        FROM BRONZE.dbo.medical_item_icd_10am m1
    )
    UPDATE ew
    SET ew.ICD_Type = COALESCE(im.description, 'No Type'),
        ew.ACHI_Code = li.ACHI_Code,
        ew.Edition = li.Edition,
        ew.long_description = li.long_description,
        ew.short_description = li.short_description,
        ew.block = li.block,
        ew.code_id_clean = li.code_id_clean
    FROM SILVER.dbo.Episode_Detail ew
    LEFT JOIN LatestICD li
        ON ew.ChapterItemNumberTmp = li.icd_item
        AND li.rn = 1
    --LEFT JOIN SILVER.dbo.etl_icd_mapping im
	  LEFT JOIN SILVER.dbo.ICD_Code_Mapping im
        ON li.icd_type = im.icd_type;
    
    -- ========================================================================
    -- 完成
    -- ========================================================================
    SET @RowCount = (SELECT COUNT(*) FROM SILVER.dbo.Episode_Detail);
    
    -- 返回加载的记录数
    SELECT @RowCount AS RecordsLoaded;
END;
GO
```

# [dbo].[usp_Load_GOV_Glossary_Cleaned]

```sql
/****** Object:  StoredProcedure [dbo].[usp_Load_GOV_Glossary_Cleaned]    Script Date: 29/06/2026 1:38:34 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE   PROCEDURE [dbo].[usp_Load_GOV_Glossary_Cleaned]
AS
BEGIN
    SET NOCOUNT ON;

    -- Drop Silver table if exists (for full refresh)
    IF OBJECT_ID('[SILVER].[GOV].[glossary_cleaned]', 'U') IS NOT NULL -- testing comment
        DROP TABLE [SILVER].[GOV].[glossary_cleaned];

    -- Step 1: Stage raw data
    SELECT
        [Nick_Name],
        [Name],
        CAST([Definition] AS NVARCHAR(MAX)) AS Definition
    INTO #stage_raw
    FROM [BRONZE].[dbo].[Westfund_Enterprise_Glossary]
    WHERE [Name] IS NOT NULL;

    -- Step 2: Replace HTML entities and control characters
    SELECT
        [Nick_Name],
        [Name],
        REPLACE(REPLACE(REPLACE(REPLACE(REPLACE(REPLACE(REPLACE(REPLACE(REPLACE(
            Definition,
            '&nbsp;',  ' '),
            '&amp;',   '&'),
            '&lt;',    '<'),
            '&gt;',    '>'),
            '&quot;',  '"'),
            '&#39;',   ''''),
            '&apos;',  ''''),
            CHAR(13),  ' '),
            CHAR(9),   ' ')
        AS Definition
    INTO #stage_entities
    FROM #stage_raw;

    -- Step 3: Use WHILE loop to strip ALL < > tags regardless of nesting depth
    SELECT Nick_Name, [Name], Definition
    INTO #stage_loop
    FROM #stage_entities;

    DECLARE @dirty INT = 1;

    WHILE @dirty > 0
    BEGIN
        UPDATE #stage_loop
        SET Definition =
            LEFT(Definition, CHARINDEX('<', Definition) - 1)
            + SUBSTRING(Definition, CHARINDEX('>', Definition) + 1, LEN(Definition))
        WHERE CHARINDEX('<', Definition) > 0
          AND CHARINDEX('>', Definition) > CHARINDEX('<', Definition);

        SET @dirty = @@ROWCOUNT;
    END;

    -- Step 4: Final whitespace cleanup and write to Silver
    SELECT
        Nick_Name,
        [Name],
        LTRIM(RTRIM(
            REPLACE(REPLACE(REPLACE(REPLACE(REPLACE(REPLACE(REPLACE(
                Definition,
            '  ', ' '), '  ', ' '), '  ', ' '),
            '  ', ' '), '  ', ' '), '  ', ' '), '  ', ' ')
        )) AS Definition
    INTO [SILVER].[GOV].[glossary_cleaned]
    FROM #stage_loop;

    -- Cleanup
    DROP TABLE #stage_raw;
    DROP TABLE #stage_entities;
    DROP TABLE #stage_loop;

END;
GO
```

# [dbo].[usp_Load_Hospital_Lookup]

```sql
/****** Object:  StoredProcedure [dbo].[usp_Load_Hospital_Lookup]    Script Date: 29/06/2026 1:38:34 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[usp_Load_Hospital_Lookup]
AS
BEGIN
    SET NOCOUNT ON;

    -- 清空目标表
    TRUNCATE TABLE dbo.Hospital_Lookup;

    ;WITH ProductCounts AS (
        SELECT 
            cp.membership_id,
            cp.cover_version,
            COUNT(*) AS product_count
        FROM BRONZE.dbo.cover_product cp
        JOIN SILVER.dbo.[Product] p
            ON cp.[product_id] = p.[Product Id]
        WHERE p.[Product Type] = 'H'
        GROUP BY cp.membership_id, cp.cover_version
    ),
    HospitalProducts AS (
        SELECT 
            cp.membership_id,
            cp.cover_version,
            p.[Product Code]
        FROM BRONZE.dbo.cover_product cp
        JOIN SILVER.dbo.[Product] p
            ON cp.[product_id] = p.[Product Id]
        JOIN ProductCounts pc
            ON cp.membership_id = pc.membership_id
           AND cp.cover_version = pc.cover_version
        WHERE p.[Product Type] = 'H'
          AND (
                (pc.product_count > 1 AND p.[Product Description] LIKE '%Hospital%')
                OR (pc.product_count = 1)
              )
    )

    INSERT INTO dbo.Hospital_Lookup (
        claim_id,
        membership_id,
        service_date,
        cover_type,
        cover_state,
        Ancillary_Product_Code
    )
    SELECT DISTINCT
        cl.claim_id,
        cl.membership_id,
        cl.service_date,
        co.cover_type,
        co.cover_state,
        hospital.[Product Code] AS Hospital_Product_Code
    FROM BRONZE.dbo.claim_line AS cl
    INNER JOIN BRONZE.dbo.cover AS co
        ON cl.membership_id = co.membership_id
    LEFT JOIN HospitalProducts AS hospital
        ON hospital.membership_id = co.membership_id
       AND hospital.cover_version = co.cover_version
    WHERE cl.service_date >= '2023-07-01'
      AND co.cover_version = (
            SELECT MAX(cover_version)
            FROM BRONZE.dbo.cover
            WHERE cover_from_date <= cl.service_date
              AND membership_id = co.membership_id
        );
END
GO
```

# [dbo].[usp_Load_Retained_Member]

```sql
/****** Object:  StoredProcedure [dbo].[usp_Load_Retained_Member]    Script Date: 29/06/2026 1:38:34 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE   PROCEDURE [dbo].[usp_Load_Retained_Member]
AS
BEGIN
    SET NOCOUNT ON;

    TRUNCATE TABLE SILVER.dbo.Retained_Member;

    WITH

    -- --------------------------------------------------------
    -- Layer 1: CurrentProductFee
    -- Gets the latest product fee version per cover_state / cover_type / product_id
    -- --------------------------------------------------------
    CurrentProductFee AS (
        SELECT
            pf.cover_state,
            pf.cover_type,
            pf.product_id,
            pf.product_fee_amount
        FROM BRONZE.dbo.product_fee pf
        WHERE pf.product_fee_version = (
            SELECT MAX(pf2.product_fee_version)
            FROM BRONZE.dbo.product_fee pf2
            WHERE pf2.fee_from_date  <= GETDATE()
              AND pf2.cover_state     = pf.cover_state
              AND pf2.cover_type      = pf.cover_type
              AND pf2.product_id      = pf.product_id
        )
    ),

    -- --------------------------------------------------------
    -- Layer 2: MemberProducts
    -- Pivots product types (A=ANC, H=HOS, B=AMB) for latest cover version
    -- --------------------------------------------------------
    MemberProducts AS (
        SELECT
            c.membership_id,
            MAX(CASE WHEN p.product_type = 'A' THEN pf.product_fee_amount END) AS [ANC$],
            MAX(CASE WHEN p.product_type = 'H' THEN pf.product_fee_amount END) AS [HOS$],
            MAX(CASE WHEN p.product_type = 'B' THEN pf.product_fee_amount END) AS [AMB$]
        FROM BRONZE.dbo.cover c
        JOIN BRONZE.dbo.cover_product cp
            ON  cp.membership_id  = c.membership_id
            AND cp.cover_version  = c.cover_version
        JOIN BRONZE.dbo.product p
            ON  p.product_id      = cp.product_id
        LEFT JOIN CurrentProductFee pf
            ON  pf.cover_state    = c.cover_state
            AND pf.cover_type     = c.cover_type
            AND pf.product_id     = cp.product_id
        WHERE c.cover_version = (
            SELECT MAX(c2.cover_version)
            FROM BRONZE.dbo.cover c2
            WHERE c2.cover_from_date  <= GETDATE()
              AND c2.membership_id     = c.membership_id
        )
        GROUP BY c.membership_id
    ),

    -- --------------------------------------------------------
    -- Layer 3: ProductPremium
    -- Calculates AnnualPremium = (ANC$ + HOS$ + AMB$) * 52
    -- --------------------------------------------------------
    ProductPremium AS (
        SELECT
            membership_id,
            (ISNULL([ANC$], 0) + ISNULL([HOS$], 0) + ISNULL([AMB$], 0)) * 52 AS Annual_Premium
        FROM MemberProducts
    ),

    -- --------------------------------------------------------
    -- Layer 4: LatestPromoSalesChannelOperator
    -- Gets the latest sales channel per member (based on latest status_date <= today)
    -- --------------------------------------------------------
    LatestPromoSalesChannelOperator AS (
        SELECT
            pm.membership_id,
            ps.description AS sales_channel_description
        FROM BRONZE.dbo.promotion_reference mg
        JOIN BRONZE.dbo.promotion gr
            ON gr.promotion_id         = mg.promotion_id
        JOIN BRONZE.dbo.promotion_sales_channel ps
            ON ps.sales_channel_id     = mg.sales_channel_id
        LEFT JOIN BRONZE.dbo.operator o
            ON o.oper_name             = mg.operator
        JOIN BRONZE.dbo.person_membership pm
            ON pm.person_id            = mg.main_ref
        WHERE mg.status_date IN (
            SELECT MAX(mg2.status_date)
            FROM BRONZE.dbo.promotion_reference mg2
            WHERE mg2.main_ref         = mg.main_ref
              AND mg2.status_date     <= GETDATE()
        )
          AND pm.relationship = 1
    )

    INSERT INTO SILVER.dbo.Retained_Member (
        Membership_Id,
        First_Name,
        Surname,
        Status,
        Product,
        Product_Code,
        Operator,
        Date,
        Retained_Or_Terminated,
        Annual_Premium,
        Branch,
        Sales_Channel
    )
    SELECT
        m.membership_id                                             AS Membership_Id,
        pc.first_name                                               AS First_Name,
        pc.surname                                                  AS Surname,
        pc.status_flag                                              AS Status,
        mc.Product_Description                                      AS Product,
        mc.FixCode                                                  AS Product_Code,

        -- Replaces: ApplyMap('OperatorName', create_operator)
        o.first_name + ' ' + o.surname                             AS Operator,

        CAST(n.create_datetime AS DATE)                             AS Date,
        srt.description                                             AS Retained_Or_Terminated,
        pp.Annual_Premium,

        mb.description                                              AS Branch,

        -- Replaces: if(isnull(Sales_Channel), 'No Channel', Sales_Channel)
        ISNULL(lpsc.sales_channel_description, 'No Channel')        AS Sales_Channel

        -- Sales_Channel_Group: requires Sales Channel Mapping Excel data
        -- To enable: import Excel sheet [Sales Channel Groups] into a lookup table, then:
        --   LEFT JOIN sales_channel_group_mapping scm
        --       ON scm.[Sales Channel] = ISNULL(lpsc.sales_channel_description, 'No Channel')
        --   ISNULL(scm.[Group], 'No Group') AS Sales_Channel_Group

    FROM BRONZE.dbo.note n

    JOIN BRONZE.dbo.sub_ref_type srt
        ON srt.sub_ref_type_id      = n.sub_ref_type_id

    -- Replaces: ApplyMap('OperatorName', create_operator)
    JOIN BRONZE.dbo.operator o
        ON o.oper_name              = n.create_operator

    JOIN BRONZE.dbo.memship m
        ON m.membership_id          = CAST(n.main_ref_id AS INT)

    JOIN BRONZE.dbo.PersonContact pc
        ON pc.membership_id         = m.membership_id
        AND pc.relationship         = 1

    JOIN ProductPremium pp
        ON pp.membership_id         = m.membership_id

    JOIN BRONZE.dbo.MemberCover mc
        ON mc.membership_id         = m.membership_id

    JOIN BRONZE.dbo.MemberBranch mb
        ON mb.membership_id         = m.membership_id

    LEFT JOIN LatestPromoSalesChannelOperator lpsc
        ON lpsc.membership_id       = m.membership_id

    WHERE n.sub_ref_type_id IN (123, 122);

END;
GO
```

# [dbo].[usp_Load_Retention_Tasklist]

```sql
/****** Object:  StoredProcedure [dbo].[usp_Load_Retention_Tasklist]    Script Date: 29/06/2026 1:38:34 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE   PROCEDURE [dbo].[usp_Load_Retention_Tasklist]
AS
BEGIN
    SET NOCOUNT ON;

    TRUNCATE TABLE SILVER.dbo.Retention_Tasklist;

    INSERT INTO SILVER.dbo.Retention_Tasklist (
        Membership_Id,
        Task_Type,
        Memship_Status,
        Tasklist_Date,
        Tasklist_Week_End,
        Tasklist_Month_Year,
        Completed_Flag,
        Reason_For_Cancel,
        Note_Text,
        Branch_Tasklist,
        Days_On_Task_List,
        Flag_14_Days
    )
    SELECT
        mt.membership_id,
        mt.task_type,
        ISNULL(ms.description, m.memship_status)                                                AS Memship_Status,
        CAST(mt.create_datetime AS DATE)                                                         AS Tasklist_Date,

        -- WeekEnd: Saturday of the week (week starts Sunday), DATEFIRST-independent
        DATEADD(DAY, 6 - (DATEDIFF(DAY, '1900-01-07', CAST(mt.create_datetime AS DATE)) % 7),
                CAST(mt.create_datetime AS DATE))                                                AS Tasklist_Week_End,

        FORMAT(CAST(mt.create_datetime AS DATE), 'MMM yyyy')                                    AS Tasklist_Month_Year,
        mt.completed_flag,
        mt.reason_for_cancel,
        CAST(mt.note_text AS NVARCHAR(MAX))                                                     AS Note_Text,
        ISNULL(g.description, 'Unknown')                                                        AS Branch_Tasklist,
        DATEDIFF(DAY, CAST(mt.create_datetime AS DATE), CAST(GETDATE() AS DATE))                AS Days_On_Task_List,
        CASE
            WHEN DATEDIFF(DAY, CAST(mt.create_datetime AS DATE), CAST(GETDATE() AS DATE)) > 14
                THEN 'Greater than 14 Days'
            ELSE 'Less than 14 Days'
        END                                                                                     AS Flag_14_Days

    FROM BRONZE.dbo.memship_tasklist mt

    JOIN BRONZE.dbo.memship m
        ON m.membership_id = mt.membership_id

    -- Replaces: ApplyMap('MemshipStatusMapping', memship_status)
    LEFT JOIN BRONZE.dbo.memship_status ms
        ON ms.memship_status = m.memship_status

    -- Replaces: ApplyMap('GroupMap', branch_group_id, 'Unknown')
    LEFT JOIN BRONZE.dbo.grouping g
        ON g.group_id = mt.branch_group_id
        AND g.group_type = 'C'

    WHERE mt.task_type = 'R'
      AND mt.completed_flag = 'N';

END;
GO
```

# [dbo].[usp_Load_RPA_Consolidated]

```sql
/****** Object:  StoredProcedure [dbo].[usp_Load_RPA_Consolidated]    Script Date: 29/06/2026 1:38:34 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE   PROCEDURE [dbo].[usp_Load_RPA_Consolidated]
AS
BEGIN
    SET NOCOUNT ON;

    -- ----------------------------------------------------------------
    -- Step 1: Truncate and reload (full refresh pattern).
    -- Prerequisite: run create_table_RPA_Consolidated.sql first.
    -- ----------------------------------------------------------------
    TRUNCATE TABLE SILVER.dbo.RPA_Consolidated;

    -- ----------------------------------------------------------------
    -- Step 3: Build intermediate CTEs and insert.
    -- ----------------------------------------------------------------
    WITH Reason_Clean AS (
        -- Reads rpa.HPC_ClaimsInfo ONCE.
        -- claimType is selected here so Claims_Flag can reference it.
        -- All HPC fields needed downstream are also selected here to
        -- avoid a second JOIN to this table later.
        SELECT
            execution_id,
            process_time,
            abbyTransactionID,
            timestampUsername,
            claimNumber,
            memberID,
            claimType,
            claim_status,
            process_status,
            manualReviewerUsername,
            sendToOpenAI,
            process_remarks,
            CASE
                WHEN process_remarks LIKE '%Error in selecting dependent%'
                 AND process_remarks LIKE '%System Exception%'
                    THEN 'Error in selecting dependent, System Exception'
                WHEN process_remarks LIKE 'Error in selecting dependent%'
                    THEN 'Error in selecting dependent'
                WHEN process_remarks LIKE 'Discount detected%'
                    THEN 'Discount detected'
                WHEN process_remarks LIKE 'Balance due detected%'
                    THEN 'Balance due detected'
                WHEN process_remarks LIKE 'Orthodontic%'
                    THEN 'Orthodontic Item Numbers identified'
                WHEN process_remarks LIKE 'System%'
                    THEN 'System Exception'
                ELSE process_remarks
            END AS Reasons_Display,
            REPLACE(CAST(process_remarks AS VARCHAR(MAX)), 'Rejected Claim in Manual Validation - ', '') AS Sub_Reason
        FROM [BRONZE].[dbo].[rpa.HPC_ClaimsInfo]
        -- IS NOT NULL added explicitly: NULL <> '' evaluates to UNKNOWN,
        -- which would silently drop NULL rows without this guard.
        WHERE execution_id IS NOT NULL
          AND execution_id <> ''
    ),

    Claims_Flag AS (
        -- Tags Remedial Massage records before the cutoff as 'Exclude'.
        SELECT
            *,
            CASE
                WHEN claimType = 'Remedial Massage' AND process_time < '2025-03-11'
                    THEN 'Exclude'
                ELSE 'Include'
            END AS Flag
        FROM Reason_Clean
    ),

    HIPPOLatestStatus AS (
        -- Most recent HIPPO status per claim.
        -- Range predicate on status_date keeps the filter sargable (index seek).
        SELECT
            mg.claim_id,
            cst.description AS Claim_Status_Type,
            ROW_NUMBER() OVER (
                PARTITION BY mg.claim_id
                ORDER BY mg.claim_status_version DESC
            ) AS rn
        FROM [BRONZE].[dbo].[claim_status] mg
        LEFT JOIN [BRONZE].[dbo].[claim_status_type] cst
            ON cst.claim_status_type = mg.claim_status_type
        WHERE mg.status_date >= '2024-01-01'
    ),

    Claim_Lines_RPA AS (
        -- Seven RPA reporting tables unified into one column schema.
        -- Columns absent in a given source are filled with NULL.
        SELECT
            execution_id, process_time, abbyTransactionID, timestampUsername,
            claimNumber                                AS ClaimID,
            CONCAT(claimNumber, '-', Line_Item_Number) AS ClaimKey,
            memberName, lineID, Provider_Number, Provider_Name,
            Description, Net_Price, Postcode, Service_Date, Reject_Page,
            Item_Number, Line_Item_Number, Hippo_Service_Desc, Hippo_Service_Type,
            'Acupuncture' AS Source
        FROM [BRONZE].[dbo].[rpa.Reporting_Acupuncture]

        UNION ALL

        SELECT
            execution_id, process_time, abbyTransactionID, timestampUsername,
            claimNumber, CONCAT(claimNumber, '-', Line_Item_Number),
            MemberName, lineID, Provider_Number, Provider_Name,
            Description, Net_Price,
            NULL AS Postcode,          -- not captured for Chiro
            Service_Date, Reject_Page, Item_Number, Line_Item_Number,
            Hippo_Service_Desc, Hippo_Service_Type, 'Chiropractic'
        FROM [BRONZE].[dbo].[rpa.Reporting_Chiro]

        UNION ALL

        SELECT
            execution_id, process_time, abbyTransactionID, timestampUsername,
            claimNumber, CONCAT(claimNumber, '-', Line_Item_Number),
            MemberName, lineID, Provider_Number,
            NULL AS Provider_Name,     -- not captured for CHM
            Description, Net_Price, Postcode, Service_Date, Reject_Page,
            Item_Number, Line_Item_Number, Hippo_Service_Desc, Hippo_Service_Type,
            'Chinese Herbal Medicine'
        FROM [BRONZE].[dbo].[rpa.Reporting_CHM]

        UNION ALL

        SELECT
            execution_id, process_time, abbyTransactionID, timestampUsername,
            claimNumber, CONCAT(claimNumber, '-', Line_Item_Number),
            MemberName, lineID, Provider_Number,
            NULL AS Provider_Name,     -- not captured for Dental
            Description, Net_Price,
            NULL AS Postcode,          -- not captured for Dental
            Service_Date, Reject_Page, Item_Number, Line_Item_Number,
            Hippo_Service_Desc, Hippo_Service_Type, 'Dental'
        FROM [BRONZE].[dbo].[rpa.Reporting_Dental]

        UNION ALL

        SELECT
            execution_id, process_time, abbyTransactionID, timestampUsername,
            claimNumber, CONCAT(claimNumber, '-', Line_Item_Number),
            MemberName, lineID, Provider_Number,
            NULL AS Provider_Name,     -- not captured for Massage
            Description, Net_Price, Postcode, Service_Date, Reject_Page,
            Item_Number, Line_Item_Number, Hippo_Service_Desc, Hippo_Service_Type,
            'Massage'
        FROM [BRONZE].[dbo].[rpa.Reporting_Massage]

        UNION ALL

        SELECT
            execution_id, process_time, abbyTransactionID, timestampUsername,
            claimNumber, CONCAT(claimNumber, '-', Line_Item_Number),
            MemberName, lineID,
            NULL AS Provider_Number,   -- not captured for Pharmacy
            NULL AS Provider_Name,
            Description, Net_Price,
            NULL AS Postcode,
            NULL AS Service_Date,      -- not captured for Pharmacy
            Reject_Page,
            NULL AS Item_Number,       -- not captured for Pharmacy
            Line_Item_Number, Hippo_Service_Desc, Hippo_Service_Type, 'Pharmacy'
        FROM [BRONZE].[dbo].[rpa.Reporting_Pharmacy]

        UNION ALL

        SELECT
            execution_id, process_time, abbyTransactionID, timestampUsername,
            claimNumber, CONCAT(claimNumber, '-', Line_Item_Number),
            MemberName, lineID, Provider_Number,
            NULL AS Provider_Name,     -- not captured for Physio
            Description, Net_Price,
            NULL AS Postcode,
            Service_Date, Reject_Page, Item_Number, Line_Item_Number,
            Hippo_Service_Desc, Hippo_Service_Type, 'Physiotherapy'
        FROM [BRONZE].[dbo].[rpa.Reporting_Physio]
    ),

    Claim_Lines_HPC AS (
        -- ClaimInfo rows from HPC — claim-level, no line items.
        SELECT
            execution_id, process_time, abbyTransactionID, timestampUsername,
            claimNumber        AS ClaimID,
            NULL               AS ClaimKey,
            NULL               AS memberName,
            NULL               AS lineID,
            NULL               AS Provider_Number,
            NULL               AS Provider_Name,
            NULL               AS Description,
            NULL               AS Net_Price,
            NULL               AS Postcode,
            NULL               AS Service_Date,
            NULL               AS Reject_Page,
            NULL               AS Item_Number,
            NULL               AS Line_Item_Number,
            NULL               AS Hippo_Service_Desc,
            NULL               AS Hippo_Service_Type,
            'ClaimInfo'        AS Source,
            memberID           AS Membership_ID_HPC,
            process_remarks    AS Reasons_HPC,
            Reasons_Display,
            Sub_Reason,
            claim_status       AS Status_HPC,
            process_status     AS ProcessStatus_HPC,
            manualReviewerUsername  AS Manual_Reviewer_Username,
            sendToOpenAI       AS Send_To_Open_AI,
            Flag
        FROM Claims_Flag
    ),

    All_Lines AS (
        -- Merge RPA and HPC rows; pre-compute Sydney date once here.
        SELECT
            execution_id, process_time, abbyTransactionID, timestampUsername,
            ClaimID, ClaimKey, memberName, lineID,
            Provider_Number, Provider_Name, Description, Net_Price,
            Postcode, Service_Date, Reject_Page, Item_Number, Line_Item_Number,
            Hippo_Service_Desc, Hippo_Service_Type, Source,
            CAST(
                process_time AT TIME ZONE 'UTC'
                             AT TIME ZONE 'AUS Eastern Standard Time'
            AS DATE)       AS process_date_sydney,
            NULL           AS Membership_ID_HPC,
            NULL           AS Reasons_HPC,
            NULL           AS Reasons_Display,
            NULL           AS Sub_Reason,
            NULL           AS Status_HPC,
            NULL           AS ProcessStatus_HPC,
            NULL           AS Manual_Reviewer_Username,
            NULL           AS Send_To_Open_AI,
            NULL           AS Flag
        FROM Claim_Lines_RPA

        UNION ALL

        SELECT
            execution_id, process_time, abbyTransactionID, timestampUsername,
            ClaimID, ClaimKey, memberName, lineID,
            Provider_Number, Provider_Name, Description, Net_Price,
            Postcode, Service_Date, Reject_Page, Item_Number, Line_Item_Number,
            Hippo_Service_Desc, Hippo_Service_Type, Source,
            CAST(
                process_time AT TIME ZONE 'UTC'
                             AT TIME ZONE 'AUS Eastern Standard Time'
            AS DATE)       AS process_date_sydney,
            Membership_ID_HPC, Reasons_HPC, Reasons_Display, Sub_Reason,
            Status_HPC, ProcessStatus_HPC, Manual_Reviewer_Username,
            Send_To_Open_AI, Flag
        FROM Claim_Lines_HPC
    )

    INSERT INTO SILVER.dbo.RPA_Consolidated
    SELECT
        al.execution_id,
        al.process_time,
        al.process_date_sydney,
        DATENAME(MONTH, al.process_date_sydney)
            + ' ' + CAST(YEAR(al.process_date_sydney) AS VARCHAR(4))   AS MonthYear,
        al.abbyTransactionID,
        al.timestampUsername,
        al.ClaimID,
        al.ClaimKey,
        al.memberName,
        al.lineID,
        al.Provider_Number,
        al.Provider_Name,
        al.Description,
        al.Net_Price,
        al.Postcode,
        al.Service_Date,
        al.Reject_Page,
        al.Item_Number,
        al.Line_Item_Number,
        al.Hippo_Service_Desc,
        al.Hippo_Service_Type,
        al.Source,
        -- COALESCE: HPC row already has these populated; RPA rows fall back to hpc2 JOIN
        COALESCE(al.Membership_ID_HPC,  hpc2.memberID)          AS Membership_ID_HPC,
        COALESCE(al.Reasons_HPC,        hpc2.process_remarks)   AS Reasons_HPC,
        al.Reasons_Display,
        al.Sub_Reason,
        COALESCE(al.Status_HPC,         hpc2.claim_status)      AS Status_HPC,
        COALESCE(al.ProcessStatus_HPC,  hpc2.process_status)    AS ProcessStatus_HPC,
        al.Manual_Reviewer_Username,
        al.Send_To_Open_AI,
        al.Flag,
        hippo.Claim_Status_Type                                  AS MaxClaimStatus_HIPPO,
        fee.critical_red_lower_limit                             AS Red_Lower_Limit,
        fee.critical_red_upper_limit                             AS Red_Upper_Limit,
        fee.warning_blue_lower_limit                             AS Blue_Lower_Limit,
        fee.warning_blue_upper_limit                             AS Blue_Upper_Limit
    FROM All_Lines al

    -- Provides memberID/status for RPA line rows that have no ClaimInfo entry
    LEFT JOIN [BRONZE].[dbo].[rpa.HPC_ClaimsInfo] hpc2
        ON al.ClaimID = hpc2.claimNumber

    LEFT JOIN [BRONZE].[dbo].[item_averagefee] fee
        ON  al.Hippo_Service_Type = fee.service_type
        AND TRY_CONVERT(VARCHAR(50), al.Item_Number) = TRY_CONVERT(VARCHAR(50), fee.item_number)

    -- Latest HIPPO status only
    LEFT JOIN (
        SELECT * FROM HIPPOLatestStatus WHERE rn = 1
    ) hippo
        ON al.ClaimID = TRY_CONVERT(VARCHAR(50), hippo.claim_id);

END;
GO
```

# [dbo].[usp_Process_Episode_Classification]

```sql
/****** Object:  StoredProcedure [dbo].[usp_Process_Episode_Classification]    Script Date: 29/06/2026 1:38:34 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[usp_Process_Episode_Classification]
AS
BEGIN
    SET NOCOUNT ON;
    SET ANSI_WARNINGS OFF;
    
    DECLARE @RowCount INT;
    
    -- ========================================================================
    -- Step 0: 创建最终输出表（如果不存在）
    -- ========================================================================
    IF OBJECT_ID('dbo.Episode_Classification', 'U') IS NULL
    BEGIN
        CREATE TABLE dbo.Episode_Classification (
            -- 主键字段
            EpisodeClaimID BIGINT NOT NULL,
            EpisodeClaimLineID BIGINT NOT NULL,
            EpisodeID INT NOT NULL,
            
            -- Episode 基础信息
            ChapterItemNumber VARCHAR(50) NULL,
            membership_id VARCHAR(50) NULL,
            admission_date DATE NULL,
            discharge_date DATE NULL,
            drg_number VARCHAR(50) NULL,
            drg_version VARCHAR(50) NULL,
            episode_type VARCHAR(50) NULL,
            admission_code VARCHAR(50) NULL,
            separation_code VARCHAR(50) NULL,
            person_id VARCHAR(50) NULL,
            
            -- ICD 代码信息
            ICD_Type VARCHAR(100) NULL,
            block INT NULL,
            ACHI_Code VARCHAR(50) NULL,
            long_description VARCHAR(500) NULL,
            short_description VARCHAR(255) NULL,
            
            -- 最终分类
            Chapter VARCHAR(50) NULL,
            
            -- Procedure 和 Diagnosis
            EpisodeProcedureItemNumber VARCHAR(50) NULL,
            EpisodeDiagnosisItemNumber VARCHAR(50) NULL,
            Procedure_Block INT NULL,
            code_id_clean VARCHAR(50) NULL
        );
        
        PRINT 'Table Episode_Classification created.';
    END
    
    -- ========================================================================
    -- Step 1: 计算 ChapterTmp（基于 Block 范围）
    -- ========================================================================
    UPDATE SILVER.dbo.Episode_Detail
    SET ChapterTmp = 
        CASE 
            WHEN block BETWEEN 1 AND 86 THEN 'Neurological'
            WHEN block BETWEEN 110 AND 129 THEN 'Endocrine'
            WHEN block BETWEEN 160 AND 256 THEN 'Opthalmological'
            WHEN block BETWEEN 300 AND 334 THEN 'Otological'
            WHEN block BETWEEN 370 AND 422 THEN 'Digestive Disorders'
            WHEN block BETWEEN 450 AND 490 THEN 'Other'
            WHEN block BETWEEN 520 AND 572 THEN 'Respiratory'
            WHEN block BETWEEN 600 AND 777 THEN 'Cardiovascular'
            WHEN block BETWEEN 800 AND 817 THEN 'Haematological'
            WHEN block BETWEEN 850 AND 1011 THEN 'Digestive Disorders'
            WHEN block BETWEEN 1040 AND 1129 THEN 'Genitourinary'
            WHEN block BETWEEN 1160 AND 1203 THEN 'Male Genital Organs'
            WHEN block BETWEEN 1240 AND 1299 THEN 'Gynaecological'
            WHEN block BETWEEN 1330 AND 1347 THEN 'Maternity'
            WHEN block BETWEEN 1360 AND 1580 THEN 'Musculoskeletal'
            WHEN block BETWEEN 1600 AND 1718 THEN 'Integumentary'
            WHEN block BETWEEN 1740 AND 1759 THEN 'Genitourinary'
            WHEN block BETWEEN 1786 AND 1799 THEN 'Neoplasm'
            WHEN block BETWEEN 1820 AND 1923 THEN 'Other'
            WHEN block BETWEEN 1940 AND 2016 THEN 'Other'
            ELSE 'Other'
        END;
    
    -- ========================================================================
    -- Step 2: 确定最终的 ChapterItemNumber（特殊逻辑）
    -- ========================================================================
    UPDATE SILVER.dbo.Episode_Detail
    SET ChapterItemNumber = 
        CASE 
            WHEN ChapterTmp IN ('Interventions not elsewhere classified', 'Unknown', 'Dental',
                               'Procedures on breast', 'Imaging services', 'Other') 
                THEN EpisodeDiagnosisItemNumber
            WHEN LEFT(UPPER(EpisodeDiagnosisItemNumber), 1) IN ('C', 'Z', 'E', 'J') 
                THEN EpisodeDiagnosisItemNumber
            ELSE ChapterItemNumberTmp
        END;
    
    -- ========================================================================
    -- Step 3: 提取 ICD10 前缀和数字部分
    -- ========================================================================
    UPDATE SILVER.dbo.Episode_Detail
    SET ICD10Prefix = UPPER(LEFT(ChapterItemNumber, 1)),
        CodeNum = TRY_CAST(SUBSTRING(ChapterItemNumber, 2, 2) AS INT);
    
    -- ========================================================================
    -- Step 4: 关联基础诊断分类（A-Z）
    -- ========================================================================
    UPDATE ew
    SET ew.DxCategory = icm.Category
    FROM SILVER.dbo.Episode_Detail ew
    LEFT JOIN BRONZE.dbo.icd10_category_map icm
        ON ew.ICD10Prefix = icm.ICD10Prefix;
    
    -- ========================================================================
    -- Step 5: 应用 R 分类细分
    -- ========================================================================
    UPDATE ew
    SET ew.R_Category = rcm.Category
    FROM SILVER.dbo.Episode_Detail ew
    LEFT JOIN BRONZE.dbo.icd10_r_category_map rcm
        ON ew.CodeNum BETWEEN rcm.RangeStart AND rcm.RangeEnd;
    
    -- ========================================================================
    -- Step 6: 应用 Z 分类细分
    -- ========================================================================
    UPDATE ew
    SET ew.Z_Category = zcm.Category
    FROM SILVER.dbo.Episode_Detail ew
    LEFT JOIN BRONZE.dbo.icd10_z_category_map zcm
        ON ew.CodeNum BETWEEN zcm.RangeStart AND zcm.RangeEnd;
    
    -- ========================================================================
    -- Step 7: 应用 S 分类细分
    -- ========================================================================
    UPDATE ew
    SET ew.S_Category = scm.Category
    FROM SILVER.dbo.Episode_Detail ew
    LEFT JOIN BRONZE.dbo.icd10_s_category_map scm
        ON ew.CodeNum BETWEEN scm.RangeStart AND scm.RangeEnd;
    
    -- ========================================================================
    -- Step 8: 应用 H 分类细分
    -- ========================================================================
    UPDATE ew
    SET ew.H_Category = hcm.Category
    FROM SILVER.dbo.Episode_Detail ew
    LEFT JOIN BRONZE.dbo.icd10_h_category_map hcm
        ON ew.CodeNum BETWEEN hcm.RangeStart AND hcm.RangeEnd;
    
    -- ========================================================================
    -- Step 9: 应用 D 分类细分
    -- ========================================================================
    UPDATE ew
    SET ew.D_Category = dcm.Category
    FROM SILVER.dbo.Episode_Detail ew
    LEFT JOIN BRONZE.dbo.icd10_d_category_map dcm
        ON ew.CodeNum BETWEEN dcm.RangeStart AND dcm.RangeEnd;
    
    -- ========================================================================
    -- Step 10: 应用 Q 分类细分
    -- ========================================================================
    UPDATE ew
    SET ew.Q_Category = qcm.Category
    FROM SILVER.dbo.Episode_Detail ew
    LEFT JOIN BRONZE.dbo.icd10_q_category_map qcm
        ON ew.CodeNum BETWEEN qcm.RangeStart AND qcm.RangeEnd;
    
    -- ========================================================================
    -- Step 11: 应用 T 分类细分
    -- ========================================================================
    UPDATE ew
    SET ew.T_Category = tcm.Category
    FROM SILVER.dbo.Episode_Detail ew
    LEFT JOIN BRONZE.dbo.icd10_t_category_map tcm
        ON ew.CodeNum BETWEEN tcm.RangeStart AND tcm.RangeEnd;
    
    -- ========================================================================
    -- Step 12: 计算 ChapterTmp2（优先诊断分类）
    -- ========================================================================
    UPDATE SILVER.dbo.Episode_Detail
    SET ChapterTmp2 = 
        CASE 
            WHEN LEN(LTRIM(RTRIM(ISNULL(DxCategory, '')))) > 0 THEN DxCategory
            WHEN LEN(LTRIM(RTRIM(ISNULL(ChapterTmp, '')))) > 0 THEN ChapterTmp
            ELSE 'Other'
        END;
    
    -- ========================================================================
    -- Step 13: 计算 ChapterTmp3（应用子分类）
    -- ========================================================================
    UPDATE SILVER.dbo.Episode_Detail
    SET ChapterTmp3 = 
        CASE 
            WHEN ICD10Prefix = 'R' THEN R_Category
            WHEN ICD10Prefix = 'Z' THEN Z_Category
            WHEN ICD10Prefix = 'H' THEN H_Category
            WHEN ICD10Prefix = 'T' THEN T_Category
            WHEN ICD10Prefix = 'D' THEN D_Category
            WHEN ICD10Prefix = 'Q' THEN Q_Category
            WHEN ICD10Prefix = 'S' THEN S_Category
            ELSE ChapterTmp2
        END;
    
    -- ========================================================================
    -- Step 14: 计算 ChapterLevel1
    -- ========================================================================
    UPDATE SILVER.dbo.Episode_Detail
    SET ChapterLevel1 = 
        CASE 
            WHEN ChapterTmp3 = 'Other' AND ICD10Prefix <> 'T' THEN ChapterTmp2
            ELSE ChapterTmp3
        END;
    
    -- ========================================================================
    -- Step 15: 计算 Chapter_Level_2（基于 Procedure Block）
    -- ========================================================================
    UPDATE SILVER.dbo.Episode_Detail
    SET Chapter_Level_2 = 
        CASE 
            WHEN ChapterLevel1 = 'Other' AND Procedure_Block BETWEEN 1 AND 86 THEN 'Neurological'
            WHEN ChapterLevel1 = 'Other' AND Procedure_Block BETWEEN 110 AND 129 THEN 'Endocrine'
            WHEN ChapterLevel1 = 'Other' AND Procedure_Block BETWEEN 160 AND 256 THEN 'Opthalmological'
            WHEN ChapterLevel1 = 'Other' AND Procedure_Block BETWEEN 300 AND 334 THEN 'Otological'
            WHEN ChapterLevel1 = 'Other' AND Procedure_Block BETWEEN 370 AND 422 THEN 'Digestive Disorders'
            WHEN ChapterLevel1 = 'Other' AND Procedure_Block BETWEEN 450 AND 490 THEN 'Other'
            WHEN ChapterLevel1 = 'Other' AND Procedure_Block BETWEEN 520 AND 572 THEN 'Respiratory'
            WHEN ChapterLevel1 = 'Other' AND Procedure_Block BETWEEN 600 AND 777 THEN 'Cardiovascular'
            WHEN ChapterLevel1 = 'Other' AND Procedure_Block BETWEEN 800 AND 817 THEN 'Haematological'
            WHEN ChapterLevel1 = 'Other' AND Procedure_Block BETWEEN 850 AND 1011 THEN 'Digestive Disorders'
            WHEN ChapterLevel1 = 'Other' AND Procedure_Block BETWEEN 1040 AND 1129 THEN 'Genitourinary'
            WHEN ChapterLevel1 = 'Other' AND Procedure_Block BETWEEN 1160 AND 1203 THEN 'Male Genital Organs'
            WHEN ChapterLevel1 = 'Other' AND Procedure_Block BETWEEN 1240 AND 1299 THEN 'Gynaecological'
            WHEN ChapterLevel1 = 'Other' AND Procedure_Block BETWEEN 1330 AND 1347 THEN 'Maternity'
            WHEN ChapterLevel1 = 'Other' AND Procedure_Block BETWEEN 1360 AND 1580 THEN 'Musculoskeletal'
            WHEN ChapterLevel1 = 'Other' AND Procedure_Block BETWEEN 1600 AND 1718 THEN 'Integumentary'
            WHEN ChapterLevel1 = 'Other' AND Procedure_Block BETWEEN 1740 AND 1759 THEN 'Genitourinary'
            WHEN ChapterLevel1 = 'Other' AND Procedure_Block BETWEEN 1786 AND 1799 THEN 'Neoplasm'
            WHEN ChapterLevel1 = 'Other' AND Procedure_Block BETWEEN 1820 AND 1923 THEN 'Other'
            WHEN ChapterLevel1 = 'Other' AND Procedure_Block BETWEEN 1940 AND 2016 THEN 'Other'
            ELSE ChapterLevel1
        END;
    
    -- ========================================================================
    -- Step 16: 计算最终 Chapter（应用特殊规则）
    -- ========================================================================
    UPDATE SILVER.dbo.Episode_Detail
    SET Chapter = 
        CASE 
            -- 特殊 Procedure 项目号规则
            WHEN Chapter_Level_2 = 'Other' AND EpisodeProcedureItemNumber = '96273-00' THEN 'Infectious Diseases'
            WHEN Chapter_Level_2 = 'Other' AND EpisodeProcedureItemNumber = '95550-03' THEN 'Musculoskeletal'
            WHEN Chapter_Level_2 = 'Other' AND EpisodeProcedureItemNumber = '96187-00' THEN 'Neurological'
            WHEN Chapter_Level_2 = 'Other' AND EpisodeProcedureItemNumber = '39015-02' THEN 'Neurological'
            WHEN Chapter_Level_2 = 'Other' AND EpisodeProcedureItemNumber = '11718-00' THEN 'Cardiovascular'
            WHEN Chapter_Level_2 = 'Other' AND EpisodeProcedureItemNumber IN ('38285-00','38286-00') THEN 'Respiratory'
            
            -- 特殊 ChapterItemNumber 规则
            WHEN Chapter_Level_2 = 'Other' AND ChapterItemNumber IN ('S01.39', 'T85.62') THEN 'Otological'
            WHEN Chapter_Level_2 = 'Other' AND ChapterItemNumber IN ('R57.0', 'T80.1') THEN 'Cardiovascular'
            WHEN Chapter_Level_2 = 'Other' AND ChapterItemNumber = 'R57.1' THEN 'Haematological'
            WHEN Chapter_Level_2 = 'Other' AND ChapterItemNumber IN ('R61.0', 'T85.81', 'T85.72', 'T85.1','T85.0', 'R51','R56.8', 'R55', 'R53') THEN 'Neurological'
            
            -- COVID-19 和感染相关
            WHEN Chapter_Level_2 = 'Other' AND ChapterItemNumber IN ('U07.2','U07.12','Y95','T89.02', 'T87.4','T79.3', 'T80.2', 'R50.9') THEN 'Infectious Diseases'
            WHEN Chapter_Level_2 = 'Other' AND ChapterItemNumber = 'U73.9' THEN 'Infectious Diseases'
            
            -- 消化系统
            WHEN Chapter_Level_2 = 'Other' AND ChapterItemNumber IN ('Z41.82','T85.5', 'T86.5','T86.4', 'T85.82', 'T85.73', 'T85.69') THEN 'Digestive Disorders'
            
            -- 泌尿生殖系统
            WHEN Chapter_Level_2 = 'Other' AND ChapterItemNumber IN ('Z52.4', 'T86.1', 'T85.71','T85.63', 'T80.8', 'Z46.6') THEN 'Genitourinary'
            
            -- 行为健康
            WHEN Chapter_Level_2 = 'Other' AND ChapterItemNumber = 'U79.1' THEN 'Behavioural'
            
            -- 眼科
            WHEN Chapter_Level_2 = 'Other' AND ChapterItemNumber IN ('T86.85','T85.2', 'T85.9', 'T85.3') THEN 'Opthalmological'
            
            -- 呼吸系统
            WHEN Chapter_Level_2 = 'Other' AND ChapterItemNumber IN ('T85.77', 'T85.65', 'T70.1') THEN 'Respiratory'
            
            -- 内分泌系统
            WHEN Chapter_Level_2 = 'Other' AND ChapterItemNumber IN ('T85.4', 'T85.85','T85.75') THEN 'Endocrine'
            
            -- 骨骼肌肉系统
            WHEN Chapter_Level_2 = 'Other' AND ChapterItemNumber IN ('T79.6', 'T86', 'T85.78','T85.69','Z47.0') THEN 'Musculoskeletal'
            
            -- 皮肤
            WHEN Chapter_Level_2 = 'Other' AND ChapterItemNumber IN ('T89.03','T89.01','T89.00','R63.4','S01.0','R60.0', 'T87.69', 'T86.86', 'T85.83', 'T79.7','T79.6', 'T79.3') THEN 'Integumentary'
            
            -- 默认使用 Chapter_Level_2
            ELSE Chapter_Level_2
        END;
    
    -- ========================================================================
    -- Step 17: 输出到最终表（DISTINCT去重）
    -- ========================================================================
    
    -- 清空最终表
    TRUNCATE TABLE SILVER.dbo.Episode_Classification;
    
    -- 插入去重后的最终结果
    INSERT INTO SILVER.dbo.Episode_Classification (
        EpisodeClaimID,
        EpisodeClaimLineID,
        EpisodeID,
        ChapterItemNumber,
        membership_id,
        admission_date,
        discharge_date,
        drg_number,
        drg_version,
        episode_type,
        admission_code,
        separation_code,
        person_id,
        ICD_Type,
        block,
        ACHI_Code,
        long_description,
        short_description,
        Chapter,
        EpisodeProcedureItemNumber,
        EpisodeDiagnosisItemNumber,
        Procedure_Block,
        code_id_clean
    )
    SELECT DISTINCT
        EpisodeClaimID,
        EpisodeClaimLineID,
        EpisodeID,
        ChapterItemNumber,
        membership_id,
        admission_date,
        discharge_date,
        drg_number,
        drg_version,
        episode_type,
        admission_code,
        separation_code,
        person_id,
        ICD_Type,
        block,
        ACHI_Code,
        long_description,
        short_description,
        Chapter,
        EpisodeProcedureItemNumber,
        EpisodeDiagnosisItemNumber,
        Procedure_Block,
        code_id_clean
    FROM SILVER.dbo.Episode_Detail;
    
    SET @RowCount = @@ROWCOUNT;
    
    -- ========================================================================
    -- 完成
    -- ========================================================================
    
    -- 返回最终记录数
    SELECT @RowCount AS FinalRecordsOutput;
END;
GO
```

# [obs].[usp_CheckHealth]

```sql
/****** Object:  StoredProcedure [obs].[usp_CheckHealth]    Script Date: 29/06/2026 1:38:34 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
-- ============================================================
-- obs.usp_CheckHealth (DAILY OVERWRITE VERSION)
-- Step 1: Volume check -> obs.ETL_VolumeLog
-- Step 2: Schema drift check -> obs.ETL_SchemaDrift
-- Step 3: Save new schema snapshot -> obs.ETL_SchemaSnapshot (keep last 5 days)
--
-- Behaviour: running multiple times on the same calendar day
-- overwrites that day's records. Different days accumulate.
-- Manual execution only - read-only against dbo/GOV
-- ============================================================

CREATE PROCEDURE [obs].[usp_CheckHealth]
AS
BEGIN
    SET NOCOUNT ON;

    DECLARE @TableSchema NVARCHAR(50), @TableName NVARCHAR(200), @SQL NVARCHAR(MAX);
    DECLARE @RowCnt BIGINT;
    DECLARE @Today DATE = CAST(GETDATE() AS DATE);

    -- ========================================================
    -- STEP 1: VOLUME CHECK
    -- ========================================================

    -- Remove today's existing volume records (if re-running today)
    DELETE FROM obs.ETL_VolumeLog
    WHERE CAST(CheckTime AS DATE) = @Today;

    DECLARE vol_cursor CURSOR FOR
        SELECT TABLE_SCHEMA, TABLE_NAME
        FROM INFORMATION_SCHEMA.TABLES
        WHERE TABLE_SCHEMA IN ('dbo', 'GOV')
          AND TABLE_TYPE = 'BASE TABLE';

    OPEN vol_cursor;
    FETCH NEXT FROM vol_cursor INTO @TableSchema, @TableName;

    WHILE @@FETCH_STATUS = 0
    BEGIN
        SET @SQL = N'SELECT @cnt = COUNT(*) FROM '
                   + QUOTENAME(@TableSchema) + '.' + QUOTENAME(@TableName);

        EXEC sp_executesql @SQL, N'@cnt BIGINT OUTPUT', @cnt = @RowCnt OUTPUT;

        INSERT INTO obs.ETL_VolumeLog (TableSchema, TableName, RowCount_, CheckTime)
        VALUES (@TableSchema, @TableName, @RowCnt, GETDATE());

        FETCH NEXT FROM vol_cursor INTO @TableSchema, @TableName;
    END

    CLOSE vol_cursor;
    DEALLOCATE vol_cursor;

    -- ========================================================
    -- STEP 2: SCHEMA DRIFT CHECK
    -- ========================================================

    -- Remove today's drift records (if re-running today),
    -- so re-running doesn't create duplicate drift entries
    DELETE FROM obs.ETL_SchemaDrift
    WHERE CAST(DetectedTime AS DATE) = @Today;

    -- Most recent snapshot per column
    -- (excludes today's snapshot, since that's about to be replaced below)
    SELECT *,
           ROW_NUMBER() OVER (PARTITION BY TableSchema, TableName, ColumnName
                               ORDER BY CheckTime DESC) AS rn
    INTO #LastSnapshot
    FROM obs.ETL_SchemaSnapshot
    WHERE CAST(CheckTime AS DATE) < @Today;

    SELECT TableSchema, TableName, ColumnName, DataType, MaxLength, IsNullable
    INTO #PrevCols
    FROM #LastSnapshot
    WHERE rn = 1;

    -- Current structure
    SELECT TABLE_SCHEMA AS TableSchema, TABLE_NAME AS TableName,
           COLUMN_NAME AS ColumnName, DATA_TYPE AS DataType,
           CHARACTER_MAXIMUM_LENGTH AS MaxLength, IS_NULLABLE AS IsNullable
    INTO #CurrCols
    FROM INFORMATION_SCHEMA.COLUMNS
    WHERE TABLE_SCHEMA IN ('dbo', 'GOV');

    -- New columns added (only flag if table was previously snapshotted)
    INSERT INTO obs.ETL_SchemaDrift (TableSchema, TableName, ColumnName, ChangeType, OldValue, NewValue)
    SELECT c.TableSchema, c.TableName, c.ColumnName, 'ADDED', NULL, c.DataType
    FROM #CurrCols c
    LEFT JOIN #PrevCols p
        ON c.TableSchema = p.TableSchema AND c.TableName = p.TableName AND c.ColumnName = p.ColumnName
    WHERE p.ColumnName IS NULL
      AND EXISTS (SELECT 1 FROM #PrevCols px WHERE px.TableSchema = c.TableSchema AND px.TableName = c.TableName)
      AND NOT EXISTS (
          SELECT 1 FROM obs.ETL_SchemaDrift d
          WHERE d.TableSchema = c.TableSchema AND d.TableName = c.TableName
            AND d.ColumnName = c.ColumnName AND d.ChangeType = 'ADDED'
      );

    -- Columns removed
    INSERT INTO obs.ETL_SchemaDrift (TableSchema, TableName, ColumnName, ChangeType, OldValue, NewValue)
    SELECT p.TableSchema, p.TableName, p.ColumnName, 'REMOVED', p.DataType, NULL
    FROM #PrevCols p
    LEFT JOIN #CurrCols c
        ON p.TableSchema = c.TableSchema AND p.TableName = c.TableName AND p.ColumnName = c.ColumnName
    WHERE c.ColumnName IS NULL
      AND NOT EXISTS (
          SELECT 1 FROM obs.ETL_SchemaDrift d
          WHERE d.TableSchema = p.TableSchema AND d.TableName = p.TableName
            AND d.ColumnName = p.ColumnName AND d.ChangeType = 'REMOVED'
      );

    -- Data type changed
    INSERT INTO obs.ETL_SchemaDrift (TableSchema, TableName, ColumnName, ChangeType, OldValue, NewValue)
    SELECT c.TableSchema, c.TableName, c.ColumnName, 'TYPE_CHANGED',
           p.DataType + ISNULL('(' + CAST(p.MaxLength AS VARCHAR) + ')',''),
           c.DataType + ISNULL('(' + CAST(c.MaxLength AS VARCHAR) + ')','')
    FROM #CurrCols c
    INNER JOIN #PrevCols p
        ON c.TableSchema = p.TableSchema AND c.TableName = p.TableName AND c.ColumnName = p.ColumnName
    WHERE (c.DataType <> p.DataType OR ISNULL(c.MaxLength,-1) <> ISNULL(p.MaxLength,-1))
      AND NOT EXISTS (
          SELECT 1 FROM obs.ETL_SchemaDrift d
          WHERE d.TableSchema = c.TableSchema AND d.TableName = c.TableName
            AND d.ColumnName = c.ColumnName AND d.ChangeType = 'TYPE_CHANGED'
      );

    DROP TABLE #LastSnapshot;
    DROP TABLE #PrevCols;
    DROP TABLE #CurrCols;

    -- ========================================================
    -- STEP 3: SAVE NEW SNAPSHOT + CLEAN UP (keep last 5 days)
    -- ========================================================

    -- Remove today's existing snapshot (if re-running today)
    DELETE FROM obs.ETL_SchemaSnapshot
    WHERE CAST(CheckTime AS DATE) = @Today;

    INSERT INTO obs.ETL_SchemaSnapshot (TableSchema, TableName, ColumnName, DataType, MaxLength, IsNullable, CheckTime)
    SELECT TABLE_SCHEMA, TABLE_NAME, COLUMN_NAME, DATA_TYPE, CHARACTER_MAXIMUM_LENGTH, IS_NULLABLE, GETDATE()
    FROM INFORMATION_SCHEMA.COLUMNS
    WHERE TABLE_SCHEMA IN ('dbo', 'GOV');

    -- Keep only the most recent 5 distinct days of snapshots
    DELETE FROM obs.ETL_SchemaSnapshot
    WHERE CAST(CheckTime AS DATE) NOT IN (
        SELECT DISTINCT TOP 5 CAST(CheckTime AS DATE)
        FROM obs.ETL_SchemaSnapshot
        ORDER BY CAST(CheckTime AS DATE) DESC
    );

END
GO
```
